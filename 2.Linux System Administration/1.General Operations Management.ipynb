### 1.1 사용자 관리  
##### 1.1.1 root 관리  

* 사용자의 개요  

1. 사용자의 분류  

1-1. root  
시스템운영에 있어서 모든 권한을 행사하므로 Privileged User 또는 Super User 라고 한다. 보통 root라는 계정 자체를 슈퍼 유저라 인식하는데, 리눅스 시스템 내부에서는 사용자의 UID를 0번 부터 정숫값으로 배분된다. 이때, root는 0이 할당된다.  

1-2. 일반 사용자  
Normal User, Unprivileged User라 부르는데, 시스템에 대해 제한적인 권한을 행사한다. UID는 1번부터 할당되는데 RHEL 7 버전 부터는 1000번 부터 할당한다.  
일반 계정도 UID를 0번으로 변환하면 슈퍼유저로 전환할 수 있다.   

2. 시스템 계정  
리눅스 사용자 정보는 /etc/passwd 파일에 기록되는데, 이 파일에 등록된 계정을 보면 root 이외에 bin, daemon, adm, game등 이러한 계정또한 시스템 계정이라고 하며, 말 그대로 시스템의 필요에 의해 생성된 계정이다. 예를 들어 root가 만든 계정이면 root 권한이여서 실행도 root만 실행 가능하다.  

3. root 계정 관리  
root를 운영 측면에서 보면 다음과 같은 5가지의 주의 사항이 요구된다.  
3-1. root 계정 이외에 다른 슈퍼유저가 존재하는 지를 점검한다 즉, root 이외에 다른 사용자의 UID가 0번 인지를 점검한다.  

3-2. PAM를 이용하여 root 계정으로 직접 로그인하는 것을 막는다. 특히, 텔넷, ssh를 통한 접근은 확실히 통제하도록 한다.  

3-3. 관리자라고 하더라도 일반 계정으로 로그인하도록 하며, root 권한이 필요한 경우에 su 명령어를 사용한다.  

3-4. 환경 변수인 TMOUT를 설정하여 무의미하게 장시간 로그인하는 것을 막는다.  

3-5. 일반 사용자에게 특정 명령어 권한만 할당해줄 경우에는 su보다는 sudo를 이용하도록 한다.  

##### 1.1.2 사용자 계정 관리  

1. 사용자 계정 생성 : user add  
계정이란 시스템에 ID와 암호를 생성하여 사용권을 부여하는 것을 말한다. 이 떄 사용자의 ID를 생성하는 명령이 useradd 이다. root 권한자가 root 이외의 사용자를 생성할 때 사용하는 명령어로 adduser라고 입력해도 된다.  

사용법  
useradd [option] 사용자명  

* -p: 사용자의 암호를 추가 시에 지정할 수 있다.  
* -d: 홈 디렉터리를 지정할 때 쓰인다.  
* -g: 그룹을 지정할 때 사용하는데, 지정할 그룹이 미리 생성되어 있어야 한다.  
* -G: 기본 그룹 이외에 추가로 그룹에 속하게 할 경우에 쓴다.  
* -c: 사용자 생성 시 사용자에 대한 설명을 설정한다.  
* -s: 사용자 생성 시 사용자에 대한 설명을 설정한다.  
* -D: /etc/default/useradd 에 설정한 유저추가와 관련된 기본 사항들을 보여준다.  
* -m: 사용자를 생성할 때 홈 디렉터리를 생성해주는 옵션이다.  
* -k: 사용자 생성 시에 제공되는 환경파일 이외의 디렉터리를 지정할 때 쓰인다.  
* -f: 사용자의 패ㅡㅅ워드 만기일을 날짜수로 지정한다.  
* -e: 계정의 만기일을 YYYY-MM-DD 형식으로 지정한다.  
* -u: 사용자 추가 시에 UID 값을 지정한다.  
* -h: useradd 명령어의 사용법 및 주요 옵션을 화면에 출력한다.  

2. 사용자 암호 부여 : passwd  
등록된 사용자의 암호를 지정하거나 변경하는 명령어이다.  

사용법  
passwd [사용자명]  

3. 사용자 전환 : su  
로그인되어 있는 상태에서 다른 사용자의 권한으로 셸을 실행할 수 있도록 전환하는 명령이다.  

사용법  
su [option] [사용자명]  

* -, -l: su 명령을 옵션 없이 실행하면 로그인한 사용자의 환경 변수 값을 가지고 다른 사람의 권한만 갖는다.  
* -c: 사용자를 전환하지 않고 다른 사용자의 권한으로 명령을 일시적으로 수행할 때 사용한다.  

4. 사용자 관련 파일 및 디렉터리  
4-1. /etc/passwd  
시스템에 로그인하여 자원을 이용할 수 있는 사용자의 목록을 저장하고 있는 정보파일  

항목 설명  
* (사용자명) : 사용자명으로 일명 ID라 부른다  
* x : 사용자의 패스워드가 기록되는 부분이다.  
* 1000 : UserID로 리눅스에서는 사용자를 숫자값으로 관리한다.  
* 1000 : GroupID로 사용자가 속한 그룹의 숫자값이다.  
* System Engineer : 사용자에 대한 추가적인 정보가 기록되는 부분이다.  
* /home/(사용자명) : 사용자의 작업 디렉터리인 홈 디렉터리를 나타낸다.  
* /bin/bash : 로그인시 사용하는 셸을 나타낸다.  

4-2. /etc/shadow  
/etc/passwd 파일의 두 번째 필드인 패스워드 부분을 암호화 하여 관리한다  

필드 설명  
* 1번째 필드 : 사용자 이름, 즉 ID  
* 2번째 필드 : 암호화된 사용자의 패스워드가 기록되는 곳.  
* 3번째 필드 : 가장 최근에 패스워드를 바꾼 날로 1970년 1월 1일 부터 계산한 날 수  
* 4번째 필드 : 패스워드를 바꾼 후 최소로 사용행 하는 날짜 수로 다른 패스워드로 변경할 수 있는 최소 날 수  
* 5번째 필드 : 현재 사용 중인 패스워드의 최대 사용가능한 기간  
* 6번째 필드 : 패스워드 사용기한이 만료되기 전에 사용자에게 바꿀 것을 경고하는 기간 
* 7번째 필드 : 패스워드 사용기한이 만료된 뒤에 실제 계정 사용이 불가능하기까지 기간이다.  
* 8번째 필드 : 계정 만기일로 계정 사용 불가능하게 된 날 또는 불가능하게 되는 날  
* 9번째 필드 : 다른 기능을 사용하기 위해 남겨둔 예약된 공간  

4-3. /etc/deault/useradd  
계정 생성 시에 기본적으로 적용되는 설정이 들어있는 파일  

사용법  
useradd -D  

항목 설명  
* GROUP=100 : GID가 100인 users라는 그룹에 속하게 된다는 뜻  
* HOME=/home : 홈 디렉터리의 위치로 사용자를 생성하면 '/home/사용자명'이 된다.  
* INACTIVE=-1 : 패스워드의 사용기한이 지난 뒤 실제로 계정의 로그인을 막는 유예기간을 설정하는 것, -1은 설정되지 않음, 0은 유예기간 없음, 3이라고 설정하면 유효기간이 지난 후 3일 내에는 로그인 할 수 있다는 뜻  
* EXPIRE= : 계정 유효기간을 지정하는 부분으로 YYYY-MM-DD와 같은 날짜 형식으로 지정  
* SHELL=/bin/bash : 사용자 생성 시에 할당되는 셸을 의미  
* SKEL=/etc/skel : 사용자 생성 시에 제공되는 파일 및 디렉터리가 들어 있는 디렉터리  
* CREATE_MAIL_SPOOL=yes : 사용자 생성 시에 메일 파일을 생성할 것인지를 지정하는 항목  

4-4. useradd 명령을 /etc/deault/useradd 파일 관리  

사용법  
useradd -D [option]  

* -D: 사용자 추가 시에 기본적으로 반영되는 정보파일인 /etc/default/useradd의 내용을 출력 
* -g: 기본 그룹을 지정, 여기에서 지정하는 값은 /etc/group 내에 있어야 한다 
* -b: 사용자 홈 디렉터리의 상위 디렉터리를 지정 
* -f: INACTIVE 행과 관련된 옵션으로 패스워드 유효기간이 만료된 후에 언제까지 해당 계정이 로그인 할 수 있는지 유예기간을 날짜 단위로 설정하낟 
* -e: 사용자의 계정이 만료일을 지정하는 옵션 
* -s: 사용자의 기본 셸을 지정할 때 사용하는 옵션  

4-5. /etc/login.defs  
메일 디렉터리, 패스워드 관련 설정, UID의 최솟값과 최댓값, GID의 최솟값 및 최댓값, 홈 디렉터리 생성 여부, 기본 UMASK 값, 패스워드에 적용하는 암호화 알고리즘 등이 정의되어 있다.  

주요 항목  
* MAIL_DER : 사용자의 메일 디렉터리를 지정 
* PASS_MAX_DAYS : 패스워드 최대 사용 기간을 지정한다. 
* PASS_MIN_DAYS : 패스워드 변경 후에 사용하는 최소 기간을 설정 
* PASS_MIN_LEN : 패스워드 최소 길이를 지정  
* PASS_WARN_AGE : 패스워드 만기일 도래 전에 경고를 보여주는 날짜를 지정  
* UID_MIN : UID의 최솟값을 지정 
* UID_MAX : UID의 최댓값을 지정  
* SYS_UID_MIN : 시스템 계정의 UID 최솟값을 지정  
* SYS_UID_MAX : 시스템 계정의 UID 최댓값을 지정  
* GID_MIN : GID의 최솟값을 지정 
* GID_MAX : GID의 최댓값을 지정  
* SYS_GID_MIN : 시스템 계정의 GID 최솟값을 지정  
* SYS_GID_MAX : 시스템 계정의 GID 최댓값을 지정  
* CREATE_HOME : 사용자 추가 시에 홈 디렉터리 생성 여부를 지정  
* USERGROUPS_ENAB : 래드햇 계열 리눇에서 옵션 없이 사용자를 추가하면 본인아이디와 동일한 그룹을 생성하여 포함시킨다  
* ENCRYPT_METHOD : 사용자 패스워드에 사용할 암호화 알고리즘을 지정하는 설정  

4-6. /etc/skel  
사용자 생성 시에 기본적으로 제공되는 파일이나 디렉터리가 위치한다.  

5.  usermod : 사용자 계정 관리  
사용자의 셸, 홈 디렉터리, 그룹, UID, GID 등을 변경하는 명령어이다.  

사용법  
usermod [option] 사용자명  

* -d: 사용자의 홈 디렉터리를 변경한다. 변경할 디렉터리는 미리 생성되어야 한다. 
* -m: 사용자의 홈 디렉터리 변경 시 기존에 사용하던 파일 및 디렉터리를 옮겨주는 옵션  
* -g: 사용자의 그룹을 변경한다 
* -s: 사용자의 쉘을 변경한다 
* -u: 사용자의 UID 값을 변경한다 
* -e: 계정만기일을 변경한다.  
* -f: 패스워드 만기일이 지난 후 패스워드에 Lock을 설정할 유예 기간을 지정한다. 
* -c: 사용자의 간단한 정보를 입력하거나 변경한다. 
* -G: 추가로 다른 그룹에 속하게 할 때 쓰인다. 
* -a: -G 옵션과 같이 사용하는 옵션으로 기존의 2차 그룹 이외에 추가로 2차 그룹을 지정할 때 사용한다. 
* -p: /etc/shadow의 2번째 필드인 암호화된 패스워드 값을 변경할 때 쓴다. 
* -l: 사용자아이디를 변경한다. 
* -L: 사용자의 패스워드에 잠금을 걸어 로그인을 막는다 
* -U: 사용자의 패스워드에 설정된 잠금을 푼다.  

6. userdel: 사용자 계정 삭제
사용자 계정을 삭제하는 명령어이다.  

사용법  
userdel [option] 사용자명  

* -r: 사용자의 홈 디렉터리 및 메일 관련 파일까지 제거한다.  

7. passwd: 사용자 패스워드 관리1  
기본적으로 패스워드를 부여하거나 변경하는 명령이지만, 계정을 사용하지 못하게 하거나 패스워드 만기일 및 유효 기간 등을 설정할 수 있다.  

사용법  
passwd [option] 사용자명  

* -S: 사용자에 대한 패스워드 정보를 알 수 있다. 
* -l: 사용자의 패스워드에 잠금을 걸어 로그인을 막는다 
* -u: 사용자에게 설정된 패스워드 잠금을 푼다 
* -d: 사용자의 패스워드를 제거한다. 패스워드 없이 로그인 가능  
* -n: 패스워드 변경까지의 최소 날짜를 지정한다.  
* -x: 현재 패스워드의 유효 기간을 지정한다. 
* -w: 패스워드 만료 전 경로 날짜를 지정한다 
* -i: 패스워드 만료된 뒤에 사용자 계정 사용이 실제 로그인이 불가능하게 되기까지의 유예 기간을 설정한다 
* -e: 다음 로그인 시에 반드시 패스워드를 변경하도록 할 때 사용한다. 이 옵션을 사용하면 /etc/shadow의 세번째 필드의 값이 0으로 설정된다.  

8. chage : 사용자 패스워드 관리2
사용자의 패스워드에 대한 정보를 출력하고 설정하는 명령으로 /etc/shadow의 날찌 관련 필드 설정을 모두 할 수 있는 명령이다.  

사용법  
chage [option] 사용자명  

* -l: 사용자의 패스워드에 대한 정보를 보여준다. 
* -d: /etc/shadow의 3번째 필드에 해당하는 값인 최근 패스워드를 바꾼날을 수정한다 
* -m: 패스워드 변경의 최소 날짜를 지정한다. 
* -M: 패스워드를 변경 없이 사용 가능한 최대 날짜를 지정한다 
* -I: 패스워드 최대 사용 기간 만료 후에 실제 패스워드에 잠금을 설정하기까지의 유예 기간을 지정  
* -E: 계정이 만기되는 날을 지정 
* -W: 패스워드 만료 전 변경을 요구하는 경고 날짜를 지정 
* -h: 명령어의 사용법 및 주요 옵션 정보를 간략히 출력한다.  

9. chpasswd  
사용자의 패스워드를 변경하는 명령어로 보통 다수 사용자의 패스워드를 한 번에 변경할 때 사용한다  

사용법  
chpasswd [option] [ < 파일]  

* -e: 암호화된 패스워드 값을 사용할 때 지정 
* -c: 패스워드에 사용될 암호화 알고리즘을 지정할 때 사용하는 옵션이다. 
* -m: 암호화 알고리즘으로 MD5를 사용할 때 지정하는 옵션  

##### 1.1.3 그룹 계정 관리  

* 그룹의 조회  

1. /etc/group  
4개의 필드로 구성되어 있고, tail 명령으로 확인한다.  

항목 설명  
* 첫번째 필드 (GroupName) : 그룹 이름을 나타낸다. 보통 groupadd 명령으로 생성한 그룹명 
* 두번째 필드 (Password) : 그룹 패스워드를 나타내는 부분이다. 별도로 관리되는 경우에는 'x'라고만 표기된다. 
* 세번째 필드 (GID) : 리눅스에서 그룹에 부여한 숫자값  
* 네번째 필드 (Member_List) : 해당 그룹에 속한 사용자의 아이디가 기록된다.  

2. /etc/gshadow  
그룹의 암호가 관리되고 root 사용자만 확인할 수 있으며, tail 명령으로 확인한다.  

항목 설명  
* 첫번째 필드 (GroupName) : 그룹 이름을 나타낸다.  
* 두번째 필드 (Password) : 그룹 패스워드가 기록된다. 설정하지 않으면 '!'로 표시됨  
* 세번째 필드 (Admin) : 그룹 관리자가 기록되는 영역으로 지정되는 사용자는 해당 그룹에 속해 있지 않아도 된다.  
* 네번째 필드 (Member_List) : 그룹의 멤버를 나타낸다. 보통 2차 그룹으로 속한 사용자의 아이디가 기록된다.  

3. groupadd : 그룹 생성  
새로운 그룹을 생성하는 명령으로 root만 이용할 수 있는 명령어이다.  

사용법

* -r: 생성하는 그룹을 시스템 그룹으로 생성할 때 사용한다. 0~999가 GID로 할당됨

6. gpasswd  
그룹의 패스워드를 설정하거나 그룹 관리자를 지정하는 명령어이며 그룹 관리자는 해당 그룹에 속하지 않아도 지정할 수 있다.

newgrp 그룹명

### 1.2  파일 시스템 관리  
##### 1.2.1 파일 및 디렉터리 관리  

* 소유권  
어떤 파일이나 디렉터리를 소유하여 지배하는 권리  

* 허가권 
어떤 파일이나 디렉터리에 접근 권한을 설정하는 권리로 일반적으로 소유권을 가진 사용자가 허가권을 설정한다.  

ls-l 명령으로 확인할 수 있는데 -,r,w,x 가 섞여 있는 부분이 허가권, 그다음이 소유권을 가진 사용자, 그다음이 그룹 소유권을 나타낸다.  

r = 파일의 내용을 볼 수 있는 권한  
w = 파일의 내용을 수정할 수 있는 권한  
x = 실행 파일을 실행시킬 수 있는 권한  

* 특수 권한  
1. 특수 권한의 개요  
일부의 권한 허용 등을 위한 특별한 권한이다.  

1-1. Set-UID  
보통 실행 파일에 사용되며 Set-UID가 부여된 파일을 실행 시, 해당 파일을 실행하는 동안에는 실행시킨 사용자의 권한이 아닌 해당 파일의 소유자 권한으로 인식한다. 설정하면 소유자 권한 부분 x자리에 s로 표기되며 실행 권한이 없는 파일에 부여하면 S 

1-2. Set-GID  
실행 파일에 설정되어 있을 때 이 파일을 실행하는 동안에 실행시킨 사용자 그룹 권한이 아닌 해당 파일을 소유한 그룹 권한으로 인식한다. 설정하면 소유자 권한 부분 x자리에 x로 표기되며 실행 권한이 없는 파일에 부여하면 S  

1-3. Sticky-Bit  
디렉터리에 설정되는 특수 권한으로 일종의 공유 디렉터리로 사용된다. Group, Other 계층 사용자에게 사용할 수 있으나 두 계층 함께 사용은 불가능하다. 이 권한 설정은 무조건 other 계층에 설정하도록 되어 있으며 설정하면 x자리에 t로 표시된다. 실행 권한이 없는 경우 T  

2. 특수 권한의 확인  

* write : Set-GID에 설정된 실행 파일에 있는 명령어로 특정 사용자에게 메시지를 전달하는 명령어 이다. 로그인한 사용자가 이용하고 있는 터미널에 메시지를 전송하는 명령어이다. 

Sticky-Bit는 /tmp 디렉터리에서 확인할 수 있는데 /tmp는 임시저장 디렉터리로 X 윈도 실행시에 필요한 소켓 등 일시적으로 시스템이나 응용 프로그램에 필요한 파일들이 저장되는 디렉터리이다. 주로 특별히 사용자 권한을 지정하기 어려운 응용 프로그램들이 일시적으로 파일을 생성하고 삭제하도록 이용된다.  

3. 특수 권한의 설정  

* chmod : Set-UID, Set-GID, Sticky-Bit도 허가권 설정 명령어 사용  
문자 모드인 경우에는 Set-UID,Set-GID는 s를 사용, Sticky-Bit는 t를 사용한다.  
Set-UID = 4, Set-GID = 2, Sticky-Bit = 1의 값을 갖는다.  

* 소유권 및 허가권 관련 명령어  
1. chmod  
파일이나 디렉터리에 접근할 수 있는 허가권을 설정하는 명령이다. 알파벳 기호를 사용하는 문자모드와 8진수 값을 사용하는 숫자모드로 지정한다.  

* 문자 모드 : 권한,사용자,설정등 3가지로 나눌 수 있는데 권한의 표시는 r,w,x를 사용하고, 사용자의 지정은 u,g,o,a를 사용하며 설정 기호는 +,-,=를 사용한다.  

u = 파일 소유자의 접근 권한을 지정할 때 사용  
g = 그룹에 속한 사용자들에 대한 접근 권한을 지정할 때 사용  
o = 다른 사용자들에 대한 접근 권한을 지정할 떄 사용  
a = 모든 사용자를 포함하여 접근 권한을 지정할 때 사용  

\+ = 설정  
\- = 해제  
\= = 특정 권한만 지정할 때 이용  

읽기 권한 = 4  
숫자 권한 = 2  
실행 권한 = 1  
모든 권한 = 7  
읽기,쓰기 권한 = 6  
읽기, 실행 = 5  
어떠한 권한 실행 x = 0  

사용법  
chmod [option] mode file(s)  

* -R: 하위 디렉터리를 포함하여 디렉터리 내부의 모든 파일의 접근 권한을 변경한다.  
* -c: 변겨된 정보를 출력한다. 
* -f: 중요한 오류 메시지가 아니면 출력하지 않는다  
* -v: 명령 진행 결광 대한 정보를 자세히 출력한다 
* --version: 명령어의 버전 정보를 출력한다  

2. chown  
파일이나 디렉터리의 소유권 및 그룹 소유권을 변경하는 명령  

사용법  
chown [option] owner[:group] file(s)  

* -R: 하위 디렉터리를 포함하여 디렉터리 내부의 모든 파일의 소유 권한을 변경한다.  
* -c: 변경된 정보를 출력한다 
* -f: 중요한 오류 메시지가 아니면 출력하지 않는다  
* -v: 명령 진행 결과에 대한 정보를 자세히 출력한다.  
* --version: 명령어의 버전 정보를 출력한다.  

3. chgrp  
파일이나 디렉터리의 소유 그룹을 바꾸는 명령으로 특정한 사용자가 여러 그룹에 속한 경우에 본인 소유의 파일을 본인이 속한 그룹 내에서 소유권을 변경시킬 수 있다.  

* -R: 하위 디렉터리를 포함하여 디렉터리 내부의 모든 파일의 그룹 소유권을 변경한다.  
* -c: 변경된 정보를 출력한다 
* -f: 중요한 오류 메시지가 아니면 출력하지 않는다 
* -v: 명령 진행 결과에 대한 정보를 자세히 출력한다 
* -h: 심볼릭 링크 파일의 그룹 소유권을 변경한다.  
* --version: 명령어의 버전 정보를 출력한다  

4. umask  
파일이나 디렉터리 생성 시 부여되는 기본 허가권 값을 지정하는 명령이다. 파일 생성 시에는 666, 디렉터리인 경우에는 777에서 설정한 umask 값을 뺀 값을 기본 허가권으로 지정한다.  

사용법

* -S: umask 값을 문자로 표기한다

하드링크 또는 심볼릭 링크를 생성하는 명령이다.(디폴트는 하드링크)  

사용법  
ln [option] 원본_대상파일명  

* -s: 심볼릭 링크를 생성시에 사용하는 옵션  
* -v: 링크 만드는 정보를 자세히 출력 
* -f: 링크 파일 존재 시에 삭제하고 생성  

3. 하드링크와 심볼릭 링크  
* 하드 링크  
하나의 동일한 파일을 디스크의 다른 곳에 배치(복사)하여 여러 이름으로 사용되는 형식이며 파일에만 부여할 수 있다.  

[하드 링크의 특징] 
1. ls -i 명령으로 I-node 번호를 확인하면 원본과 링크 파일의 번호가 같다. 
2. 원본과 링크 파일의 내용과 파일의 크기가 같다. 
3. ls -l 명령 시에 출력되는 링크의 숫자가 올라간다. 
4. 원본이나 링크 파일 중에 어떠한 파일을 수정해도 같이 반영된다. 
5. 원본 파일을 삭제해도 링크 파일은 아무런 영향을 받지 않는다. 
6. 하드 링크는 파일만 설정할 수 있고, 동일한 파일 시스템에서만 사용할 수 있다. 
7. 파티션이나 디스크 드라이브를 가로질러 사용할 수는 없다.  

* 심볼릭 링크  
하나의 파일을 여러 이름으로 가리키게 하는 것으로 원본과 생성된 링크 파일은 완전히 다른 파일로 관리되며 파일이나 디렉터리에 모두 사용할 수 있다.  

[심볼릭 링크의 특징]  
1. ls -i 명령으로 I-node 번호를 확인하면 원본과 링크 파일의 번호가 다르다.  
2. 생성된 링크 파일의 크기가 매우 작다 
3. ls -l 명ㄹ여 시에 출력되는 권한 영역의 맨 앞쪽에 '1'이라고 표시된다. 
4. 원본이나 링크 파일 중에 어떠한 파일을 수정해도 같이 반영된다. 
5. 원본 파일을 삭제하면 링크 파일은 아무런 구실을 하지 못한다. 
6. 디렉터리에 링크 파일을 생성하면 윈도우의 바로가기나 단축아이콘의 기능과 같다. 
7. 생성되는 링크 파일의 퍼미션 값이 777로 표시되나, 이 값은 원본 파일의 퍼미션과는 무관하다.  




##### 1.2.2 파일 시스템 관리 및 복구  

* 파일 시스템 생성  
1. 파일 시스템 생성의 예  

1-1. fdisk -l 명령으로 디스크 인식 여부를 확인  
1-2. fdisk 장치명 명령 실행으로 원하는 만큼 용량을 할당하여 파티션 작업을 한다.  
1-3. 시스템 재부팅 partprobe, reboot 명령으로 재부팅  
1-4. mkfs.xfs [파티션 이름] 으로 파일 시스템을 생성한다.  
1-5. mkdir 명령으로 디렉터리 생성  
1-6. mount -f xfs 명령으로 마운트 작업  
1-7. mount -> df -h 명령 순으로 마운트 및 용량 확인  
1-8. vi /etc/fstab 명령으로 파일 등록  

관련 명령어  

1. fdisk  
디스크 파티션을 확인하고 추가/삭제하는 명령  

사용법  
fdisk [option] [device]  

* -l [device]: 지정한 장치 파일의 파티션 테이블 정보를 출력한다  
* -s partition: 특정 파티션의 크기를 출력하는데 512 바이트 크기의 섹터 수이다.  
* -v, -V: fdisk 의 버전을 출력한다.  

2. mkfs  
새로운 파일 시스템을 만드는 명령으로 root 사용자만 사용할 수 있다.  

사용법  
mkfs [-t fs_type] [option] 장치명  

* -t fs_type: 파일 시스템의 유형을 지정하는 옵션이다.  
* -V: 인자값 없이 사용하면 버전 정보를 출력하고, 인자값과 함께 사용하면 상세 정보를 출력한다.  

3. mke2fs  
ext2, ext3, ext4 파일 시스템을 만드는 명령어로 mkfs 명령어를 실행하면 실제 사용되는 명령어이며 파일 시스템의 유형을 지정하지 않으면 ext2로 생성된다.  

사용법  
mke2fs [option] 장치명  

* -j : 저널링 파일 시스템인 ext3로 만든다.  
* -t fs_type: 파일 시스템의 유형을 지정하는 옵션으로 ext3, ext4 등으로 지정하면 된다.  
* -b block_size: 블록 사이즈를 지정하는 옵션이다.  
* -R raid_options: RAID 관련 옵션으로 -R 다음에 argument=값 형태로 지정한다.  

4. mkfs.xfs  
XFS 파일 시스템을 만드는 명령이다. mkfs 명령으로 -t xfs를 지정하면 이 명령이 실행된다.  

사용법  
mkfs.xfs [option] 장치명  

* -b block_size: 블록 사이즈를 지정하는 옵션으로 최소 512 바이트, 최대 65536 바이트  
* -f: 지정한 디바이스에 파일 시스템이 존재하는 경우 강제로 수행한다.  

5. mount  
보조기억장치나 파일 시스템이 다른 디스크를 /의 하위디렉터리로 연결하여 사용 가능하게 해주는 명령으로 옵션 없이 사용하면 현재 마운트된 장치나 디스크를 확인할 수 있다.  

사용법  
mount [option] [device] [directory]  

* -a: /etc/fstab에 명시된 파일 시스템을 마운트 할 때 쓰이는 옵션  
* -f fs_type: 파일 시스템의 유형을 지정하는 옵션  
* -o 항목: 마운트 할때 추가적인 설정을 적용할 때 사용하는 옵션   

[주요 파일 시스템의 유형]  
* msdos: MS-DOS의 파일 시스템인 FAT-16을 마운트할 때 지정한다. 
* vfat: 마이크로소프트사의 파일 시스템인 FAT-32를 마운트 할때 지정한다.  
* ntfs: 마이크로소프트사의 윈도우 NT, 2000, XP 이후 버전에서 사용하는 파일 시스템을 마운트 할때 지정  
* ext2, ext3, ext4: ext2, ext3, ext4 파일 시스템을 마운트 할 때 지정 
* xfs: xfs 파일 시스템을 마운트할 때 지정  
* iso9660: CD-ROM이나 DVD를 마운트 할 때 지정  
* smbfs: 네트워크 파일 시스템인 삼바 파일 시스템을 마운트 할 때 지정 
* cifs: 삼바 파일 시스템이 확장된 파일 시스템  
* nfs: 네트워크 파일 시스템인 NFS로 공유된 영역을 마운트 할 때 지정  
* udf: DVD 파일 시스템이다.  

[-o의 주요 항목]  
* ro: 읽기전용으로 마운트한다  
* rw: 읽기/쓰기 모드로 마운트 한다. 
* remount: 해당 파티션을 다시 마운트한다. 파티션 정보를 바꾸었을 때 사용  
* loop: loop 디바이스로 마운트 할때 쓴다.  
* noatime: 파일이 변경 되기 전까지는 Access Time이 변경되지 않는다. 
* username=사용자명: 삼바처럼 사용자 계정이 필요한 경우 사용자명을 입력한다 
* password=암호: 보통 username과 같이 사용되며 패스워드가 설정된 경우 사용한다. 
* acl: ext3 파일 시스템에서 지원하는 접근 제어 리스트를 사용가능 하도록 마운트 할때 사용한다. 
* uquota: 사용자 쿼터 옵션을 부여하면서 마운트 할 때 사용한다  
* gquota: 그룹 쿼터 헤드 옵션을 부여하면서 마운트할 때 사용한다.  

6. umount  
마운트된 파일 시스템을 해제시켜주는 명령이다.  

사용법  
umount [option] 디바이스명 or 마운트된_디렉터리명  

* -a: /etc/mtab에 명시된 파일 시스템을 언마운트 할 때 쓰이는 옵션이다.  
* -t fs_type: 언마운트할 파일 시스템을 지정할 때 사용하는옵션  

7. eject  
이동식 보조기억장치의 미디어를 꺼낼 때 사요한다. 이 명령을 사용하면 자동으로 언마운트 작업을 수행한다.  

사용법  
eject [장치명 or 마운트된_디렉터리명]  


관련 파일  
1. /etc/fstab  
파일 시스템에 대한 다양한 정보를 담고 있는 파일로 부팅 시에 마운트할 파티션 정보가 기록되어 있다.  

[fstab 필드 구성]  
* 첫번째: 장치명이 기록되는 영역이었으나, 최근에는 볼륨라벨이나 UUID가 대신 사용되고 있다.  
* 두번째: 마운트될 디렉터리를 나타낸다 
* 세번째: 파일 시스템의 유형을 나타낸다. 
* 네번째: 마운트 될 때의 옵션을 나타낸다. 
* 다섯번째: dump 명령을 통한 백업 시 레벨 덤프 사용주기를 결정하는 부분으로 0이면 dump를 사용하지 않고, 1이면 매일 수행, 2이면 이틀에 한번 수행한다. 
* 여섯번째: 부팅 시 파일 시스템을 점검하는 fsck 명령의 순서를 정한다. 0으로 설정하면 부팅 시에 해당 파일 시스템은 검사하지 않는다.  

[fstab 4번째 필드의 주요 옵션]  
* defaults: rw, suid, dev, exec, auto, nouser, async가 적용된다.  
* auto: 시스템 부팅 시 미디어가 존재하면 자동으로 마운트 해주고, -a 옵션을 이용한 마운트 및 활성화 등을 가능하게 한다.  
* nauto: 시스템 부팅 시 자동으로 마운트가 되지 않도록 하는 옵션으로 명시적으로만 마운트가 가능하다. -a 옵션을 이용한 마운트는 되지 않는다.  
* user: 로컬 시스템에서 일반 사용자가 마운트 할 수 있는 권한을 부여한다. 
* owner: 장치 소유자가 마운트할 수 있는 권한을 부여한다. 
* nofail: 해당 장치 파일이 존재하지 않는 경우에도 에러 보고를 하지 않는다. 
* uquota, usrquota, quota: 사용자의 용량을 제한하는 Disk quota를 사용할 때 해당 영역에 설정한다.  
* gquota, grpquota: Disk quota를 그룹별로 사용할 때 해당 영역에 설정한다 
* noquota: 해당 파티션에서 사용자들의 Quota를 설정하지 않는다. 
* nosuid: 해당 파티션에서 SUID 및 SGID 사용을 허용하지 않는다 
* nodev: 해당 파티션에서 문자나 특별한 장치를 허용하지 않는다 
* noexec: 해당 파티션에서 실행 파일이 실행되지 않도록 설정한다.  
* suid: 해당 파티션에서 SUID 및 SGID의 사용을 허가한다.  
* ro: 해당 파티션을 읽기 전용으로 설정한다 
* rw: 해당 파티션을 읽고 쓰기 모드로 설정
* async: 파일을 비동기적으로 관리하도록 설정한다. 
* acl: Access Control Lists를 사용한다 

2. /etc/mtab  
현재 시스템에 마운트된 파일 시스템 정보를 담고 있는 심볼릭 파일로써 실제 정보는 /proc/self/mounts 파일에서 가져온다  

3. blkid  
블록 장치의 속성 정보를 출력하는 명령으로 UUID, LABEL 등을 확인할 수 있다.  

사용법  
blkid [option] [device]  

* -L: 라벨명으로 블록 장치를 찾을 때 사용한다 
* -U: UUID 명으로 블록 장치를 찾을 때 사용한다  

4. lsblk  
블록 장치의 목록을 출력하는 명령  

사용법  
lsblk [option] [장치명]  

* -m: 블록 장치의 소유자, 소유 그룹, 허가권 정보를 출력한다 
* -f: 블록 장치의 이름, 파일 시스템 타입, 라벨, UUID, 마운트 포인트 정보를 출력한다  
* -o: 명시한 항목에 대한 정보를 추가로 출력한다  

5. findfs  
라벨명이나 UUID 명으로 파일 시스템을 찾는 명령  

사용법  
findfs LABEL=라벨명  
findfs UUID=UUID명  

6. findmnt  
파일 시스템 정보를 출력하는 명령으로 보통 마운트된 파일 시스템을 트리 구조 형태로 출력한다  

사용법  
findmt [option] [장치명]  

* -D: df 명령의 형식으로 출력하는 옵션으로 파일 시스템 타입, 전체 용량, 사용량, 사용 가능한 용량 등을 출력  
* -s: /etc/fstab 파일에서 관련 정보를 찾아서 출력한다  
* -t: 특정 파일 시스템을 지정해서 출력한다 
* -l: 출력되는 포맷을 지정하는 옵션으로 기본적으로 자동 적용되는 옵션이다.  

파일 시스템 점검 및 관리  
1. fsck  
ext 계열 리눅스 파일 시스템을 검사하고 수리하는 명령이다.  

사용법  
fsck [option] 장치명  

* -r: 명령 수행에 대한 확인 질문을 한다.
* -A: /etc/fstab 파일에 정의된 모든 파일 시스템을 검사한다  
* -P: -A 옵션을 사용할 때, 루트 파일 시스템을 다른 파일 시스템과 병렬로 함께 체크한다  
* -R: -A 옵션을 사용할 때, 루트 파일 시스템은 체크하지 않고 건너뛴다.  
* -N: 실행을 하지 않고 단지 어떤 것이 실행되어지는지만 보여준다.  
* -T: 검사를 시작할 때 제목을 보여주지 않는다.  
* -s: fsck 동작을 시리얼화한다. 
* -V: 실행되는 각 파일 시스템용 명령을 포함해 자세한 출력을 수행한다. 
* -version: 버전 정보를 보여준다 
* -t fs_type: 점검할 파일 시스템의 유형을 지정한다. 파일 시스템 앞에 no를 붙이면 지정한 파일 시스템을 제외한 나머지를 검사하고, -A 옵션이 설정된 경우에는 /etc/fstab 파일에서 파일 시스템 유형이 맞는 것만 검사한다.  

2. e2fsck  
리눅스 파일 시스템인 ext2, ext3, ext4를 검사하고 수리하는 명령이고 XFS 파일 시스템에서는 사용할 수 없다.  

사용법  
e2fsck [option] 장치명  

* -n: 특정 상황에 대한 물음에 대해 'no'라고 인식하여 처리한다  
* -y: 특정 상황에 대한 물음에 대해 'yes'라고 인식하여 처리한다  
* -c: 배드 블록을 검사한다  
* -f: 꺠끗한 파일 시스템까지 강제적으로 검사한다  

3. xfs_repair  
XFS 파일 시스템을 점검 및 복구할 때 사용하는 명령어이다.  

사용법  
xfs_repair [option] 장치명  

* -n: 파일 시스템을 점검만 하고, 복구 작업은 수행하지 않는다  
* -L: 관련 장치의 로그를 비운다. 데이터 손실이 있을 수 있으므로 최후의 수단으로만 사용한다  
* -v: 점검 및 복구와 관련된 시간 정보를 추가로 출력한다  
* -V: 버전 정보를 출력한다  

4. df  
현재 마운트된 디스크의 크기, 사용량, 남아있는 용량등에 대한 정보를 출력한다  

사용법  
df [option]  

* -h: 용량의 단위 표시를 표시하는데, 1KB = 1024 byte로 계산  
* -H: 용량의 단위 표시를 표시하는데, 1KB = 1000 byte로 계산  
* -k: 킬로바이트 단위로 보여준다  
* -m: 메가바이트 단위로 보여준다  
* -T: 각 파티션에 대한 파일 시스템의 유형을 보여준다  
* -i: 아이노드의 사용량을 보여준다. 'No space left on device'라는 메시지는 아이노드를 다 써버린 경우이다.  

5. du  
파일이나 디렉터리들이 디스크에서 차지하고 있는 크기를 출력한다  

사용법  
du [option] [FILE(s)]  

* -h: 용량의 단위 표시를 해준다  
* -b: 바이트 단위로 보여준다  
* -k: 킬로바이트 단위로 보여준다  
* -m: 메가바이트 단위로 보여준다  
* -a: 디렉터리에 존재하는 모든 파일에 대해 각각의 크기를 보여준다  
* -s: 파일들의 전체 크기를 합한 값만 보여준다  
* --inodes: 아이노드 사용량을 출력한다  

6. dd  
디스크 이미지를 이미지 형태로 백업하거나 파일의 포맷 즉, 형식을 바꾸는 명령으로 텍스트 파일의 대소문자변환과 부팅 디스크를 만들 때도 쓰인다. 또한 스왑 파일을 만들거나 디바이스 초기화 시킬 때도 사용한다  

사용법  
dd if=입력파일 [conv=conversion type] of=출력파일 [bs=] [count=] [skip=]  

* if= : 입력파일을 지정한다. 장치명이나 이미지 파일도 지정할 수 있다.  
* of= : 출력파일을 지정한다. 장치명이나 이미지 파일도 지정할 수 있다.  
* conv= : 변환하는 데이터 유형을 지정한다 
* bs= : 출력되는 블록 사이즈를 지정한다 한 번에 변환하는 블록 사이즈를 지정하는 항목이다  
* count= : 변환되는 블록의 수를 지정한다  
* skip= : 입력 파일의 시작되는 블록 번호를 지정한다. 보통 데이터를 나누어 이미지를 만들 때 사용한다 
* oflag= : 표준 출력에 사용되는 FLAG를 지정한다  

7. partprobe  
변경된 파티션 정보를 반영시킬 때 사용하는 명령이다. 이 명령을 실행하면 재부팅 하지 않고도 바로 적용된다.  

사용법  
partprobe [option]  

* -d: 실제 적용되지는 않고 실행 가능 여부만 점검한다  
* -s: 파티션 관련 정보를 간단히 출력한다  
* -h: 명령어 사용법 관련 정보를 출력한다  
* -v: 명령어의 버전 정보를 출력한다  

8. stat  
파일 또는 파일 시스템 관련 정보를 출력하는 명령  

사용법  
stat [options] 파일명  

* -f: 파일 대신에 파일이 속한 파일 시스템 관련 정보를 출력한다  
* -L: 심볼릭 링크 파일인 경우에는 원본 파일의 정보를 출력한다  
* --printf=형식: %n(파일이름), %U(소유자이름), %G(소유자그룹), $C(SELinux 관련 정보), %z(마지막 변경된 시간 정보) 등을 사용해서 형식을 지정할 수 있다.

2. 관련 명령어  

2-1. mkswap  
스왑 파티션이나 스왑 파일을 생성하는 명령이다  

사용법  
mkswap [option] 스왑_파일 [size]  
mkswap [option] 스왑_파티션  

* -c: 스왑 파티션 생성하기 전에 배드 블록을 검사하는 옵션이다.  

2-2. swapon  
스왑 파티션이나 스왑 파일을 활성화하는 명령으로 스왑의 상태 확인도 가능하다  

사용법  
swapon [option] 스왑_파일  
swapon [option] 스왑_파티션  

* -a: /etc/fstab 파일에 등록된 스왑 영역을 전부 활성화 시킨다 (nauto 옵션 설정인 경우는 제외)  
* -s: 스왑 영역의 상태를 출력한다  

2-3. swapoff  
활성화된 스왑 파티션이나 스왑 파일을 중지시킨다  

사용법  
swapoff [option] 스왑_파일  
swapoff [option] 스왑_파티션  

* -a: 모든 스왑 영역을 중지시킨다

* 시스템 부팅시마다 사용할 경우에는 /etc/fstab 파일에 등록  

Disk Quota  

1. 디스크 쿼터의 개요  
사용자가 많은 서버의 경우에 사용자의 디스크 사용량 제한이 필요한데 이러한 제한을 디스크 쿼터라고 한다  

2. 관련 명령어  

2-1. xfs_quota  
XFS 파일 시스템의 쿼터를 관리해주는 명령어이다.  

사용법  
xfs_quota [option]  

* -x: 전문가 모드를 활성화시키는 옵션으로 관리자 명령을 실행할 때 기본적으로 사용해야하는 옵션이다.  
* -c 명령: 쿼터 관련 명령을 지정할 때 사용하는 옵션이다.  

2-2. edquota  
사용자나 그룹에 쿼터를 설정할 때 사용하는 명령으로 실행시키면 vi 편집기가 실행되고, 기본 단위가 KB이므로 10MB를 제한하려면 10000이라고 입력해야 한다.  

사용법  
edquota [option]  

* -u: 사용자에 대한 쿼터를 설정할 때 사용하는 옵션이다. 
* -g: 그룹에 대한 쿼터를 설정할 때 사용하는 옵션이다.  
* -t: Soft limit를 초과한 후부터 적용되는 시간 제한을 설정하는 옵션이다.  
* -p: 특정 사용자의 쿼터를 다른 사용자에게 동일한 설정으로 적용할 때 사용하는 옵션이다.  
* -h: 명령어의 사용법 및 주요 옵션 정보를 간략히 출력한다  

2-3. repquota  
파일 시스템에 설정된 쿼터 정보를 출력하는 명령이다.  

사용법  
repquota [option] [디렉터리명]  

* -a: 쿼터가 설정된 모든 파티션의 정보를 출력한다 (디렉터리명을 지정할 필요가 없다)  
* -u: 사용자 쿼터 정보를 출력하는데, 디퐅트값  
* -g: 그룹 쿼터 정보를 출력한다  

2-4. quota  
사용자나 그룹 단위로 쿼터 설정 정보를 출력하는 명령이다.  

사용법  
quota [option] [user | group]  

* -u: 사용자 쿼터 정보를 출력하는데, 디폴트값  
* -g: 그룹 쿼터 정보를 출력  
* -h: 명령어의 사용법 및 주요 옵션 정보를 간략히 출력  

2-5. setquota  
쿼터를 설정하는 명령으로 vi 편집기를 이용하는 대신에 직접 명령행에서 설정할 수 있다.  

사용법  
setquota [option] 사용자명 block_soft block_hard inode_soft inode_hard 파일 시스템  
setquota -t block_grace inode_grace 파일 시스템  

* -u: 사용자 쿼터를 설정할 때 사용하는데, 디폴트값  
* -g: 그룹 쿼터를 설정할 때 사용한다  
* -t: 유예 기간을 설정할 때 사용하는데, 기본 단위는 초이다.

##### 1.2.3 파일 시스템 관련 명령어  

1. 디렉터리 관련 명령어  

1-1. pwd  
현재 위치한 작업 디렉터리의 경로를 출력하는 명령이다.  

1-2. cd  
디렉터리를 이동하는 명령  

* cd home : 상대 경로 이동법  
* cd /home : 절대 경로 이동법  
* cd : 홈 디렉터리로 이동  
* cd ~ : 홈 디렉터리로 이동  
* cd / : 최상위 디렉터리로 이동  
* cd . : 현재 디렉터리로 이동  
* cd .. 한단계 위의 디렉터리로 이동  
* cd - : 이동하기 바로 직전의 디렉터리로 이동  

1-3. mkdir  
디렉터리를 생성하는 명령  

사용법  
mkdir [option] directory_name(s)  

* -p: 경로로 지정된 부모 디렉터리가 존재하지 않으면 함께 생성  
* -m: 디렉터리 생성과 동시에 퍼미션을 설정  
* -v: 생성된 디렉터리를 메시지로 출력  

1-4. rmdir  
디렉터리를 제거하는 명령  

사용법  
rmdir [option] directory_name(s)  

* -p: 대상이 되는 디렉터리가 제거된 후에 그 결과로 부모 디렉터리도 빈 디렉터리가 되면 부모 디렉터리도 제거한다  

1-5. tree  
특정 디렉터리의 목록 정보를 트리 구조 형태로 출력하는 명령  

사용법  
tree [option] [디렉터리명]  

* -l: 심볼릭 링크가 있는 경우에 실제 디렉터리에 대한 정보도 함께 출력  
* -a: 숨겨진 파일이나 디렉터리 정보도 함께 출력한다  
* -d: 디렉터리 정보만 출력  
* -f: 파일인 경우 지정한 디렉터리를 기준으로 전체 경로 형식으로 출력한다  
* -p: 허가권 정보도 함께 출력한다  
* -s: 파일의 크기 정보도 함께 출력한다  
* -h: 파일의 크기 정보를 사람이 읽기 쉬운 형태의 단위를 붙여서 출력한다  
* -L 레벨값: 디렉터리 탐색의 깊이를 지정한다  
* -C: 항상 파일의 종류에 따라 색깔로 분류해서 출력해준다.  

2. 파일 관련 명령어  

2-1. ls  
디렉터리 안의 내용을 출력하는 명령어  

사용법  
ls [option] [파일 또는 디렉터리명]  

* -a: 숨겨진 파일이나 디렉터리도 출력  
* -A: 현재 디렉터리 및 부모 디렉터리를 제외하고 '.'으로 시작하는 파일은 숨겨진 파일이나 디렉터리를 출력한다  
* -F: 파일의 형태에 따라 관련 정보를 출력해준다. 실행 가능한 파일은 '*', 디렉터리는 './', 심볼릭 링크는 '@'를 붙여서 표시한다  
* -l: 파일이나 디렉터리 정보를 한 줄씩 하나씩 길게 보여준다.  
* -d: 대사이 디렉터리인 경우에 디렉터리 안의 내용 대신에 해당 디렉터리에 대한 정보를 출력한다 
* -t: 출력되는 정보가 최근 수정된 파일 부터 시간순으로 출력된다.  
* -u: 파일 및 디렉터리의 최종 수정된 시간 대신 사용된 시간으로 출력된다.  
* -i: 아이노드 번호를 파일 또는 디렉터리명 앞에 출력한다  
* -r: 파일 및 디렉터리 정보를 알파벳 역순으로 출력한다  
* -R: 존재하는 하위 디렉터리별로 관련 내용을 재귀적으로 전부 출력한다  
* -S: 파일 크기가 큰 것 부터 출력  
* --color: 파일의 종류에 따라 색을 보여준다  
* -1: 한 줄에 한 파일씩 출력한다  
* -h: 보통 -l 옵션과 같이 사용되며, 보기 쉽게 단위를 표시한다  
* -Z: SELinux 관련된 보안문책을 출력한다  

2-2. cp  
파일이나 디렉터리를 복사하는 명령이다.  

사용법  
cp [options] source(s) destination  

* -r, -R: 복사하려는 대상이 디렉터리인 경우에 하위 디렉터리까지 같이 복사한다  
* -i: 복사를 통해 생성할 파일이 존재할 경우에 덮어 쓸 것인지를 물어본다  
* -f: 같은 파일의 이름이 존재하는 경우에 강제로 덮어 쓴다  
* -b: 덮어쓰기가 진행될 때 백업 파일을 만든다. 파일명 뒤에 ~ 가 붙는다  
* -s: 복사하는 대신에 심볼릭 링크를 만든다  
* -p: 소유권, 그룹, 허가권 등 모든 정보를 복사한다  
* -v: 파일을 복사하기 전에 각각 파일명을 표시한다  
* -d: cp 명령은 복사할 파일이 심볼릭 링크 파일인 경우에는 그 원본 파일을 찾아 복사한다  
* -a: -dR 옵션과 같다. 특정 파일의 속성, 링크 정보들을 그대로 유지하고 그 하위 디렉터리까지 전부 복사한다  
* -l: 복사 대신에 하드 링크 파일을 만든다  
* -u: 복사되는 대상이 최신의 파일이라면 복사되지 않는다  

2-3. rm  
파일을 삭제하는 명령으로 옵션에 따라 디렉터리도 삭제한다  

사용법  
rm [options] file(s) or directory  

* -i: 파일 삭제할 때 질의를 사용한다  
* -f: 존재 여부나 어떠한 메시지 없이 강제로 지운다  
* -r,-R: 하위 디렉터리를 포함하여 디렉터리 안의 모든 내용을 지운다  
* --: 파일의 이름이 '-'로 시작하는 파일을 지울 때 사용한다  

2-4. mv  
파일이나 디렉터리의 이름을 변경하거나 이동시키는 명령  

사용법  
mv [options] source(s) destination  

* -i: 이동할 디렉터리에 동일한 파일명이 존재하면 덮어쓸 것인지를 물어본다  
* -f: 같은 파일의 이름이 존재해도 강제로 덮어쓴다  
* -b: 덮어쓰기 실행전에 백업 파일을 만든다. 원래 파일명에 ~가 붙는다  
* -v: 파일을 이동하는 내용을 출력한다  
* -u: 이동할 디렉터리에 동일한 파일명이 있을때, 최신의 파일이면 이동하지 않는다.  

2-5. touch  
파일의 접근시간, 수정시간과 같이 타임스탬프를 변경하거나 파일의 크기가 0인 파일을 생성하는 명령  

사용법  
touch [option] 파일명  

* -a: 접근 시간을 바꾼다  
* -m: 수정 시간을 바꾼다  
* -t: 현재 시간 대신 MMDDhhmm 형식으로 지정한 시간을 타임스탬프로 변경한다  
* -r: 지정된 파일의 Access time 및 Modify time으로 특정 파일의 시간을 변경한다.  

2-6. file  
파일의 종류를 출력하는 명령으로 관련 정보는 /usr/share/magic 이라는 파일을 참조한다  

사용법  
file [option] 파일명  

* -b: 정보 출력할 때 파일 이름을 출력하지 않는다  
* -i: 특정 파일의 문자 집합 정보를 출력한다  

2-7. locate  
시스템 전체에서 찾고자 하는 패턴에 맞는 파일이나 디렉터리를 경로와 함께 출력한다  

사용법  
locate [option] pattern  

* -c: 관련 정보의 경로 출력 대신에 개수를 출력한다  
* -i: 대소문자를 구분하지 않는다  

2-8. find  
현재 디렉터리부터 하위 디렉터리까지 주어진 여러 조건을 가지고 파일 및 디렉터리를 찾아서 그 경로를 출력하는 명령이다.  

사용법  
find [path] options experssion  

* -name: 주어진 이름의 파일이나 디렉터리를 찾는다  
* -user: 지정한 사용자의 파일이나 디렉터리를 찾는다  
* -group: 지정한 그룹 소유의 파일이나 디렉터리를 찾는다  
* -uid: 지정한 UID 소유의 파일이나 디렉터리를 찾는다  
* -gid: 지정한 GID 소유의 파일이나 디렉터리를 찾는다  
* -perm: 지정한 허가권 값을 갖는 파일이나 디렉터리를 찾는다  
* -type: 지정한 파일 ㅠ형을 찾는다  
* -atime n: n일 이전에 접근한 파일을 찾는다 +n은 n일 이전에 액세스한 파일, -n은 n일 내에 액세스한 파일  
* -ctime n: Change Time 기준으로 찾는다  
* -mtime n: Modify Time 기준으로 찾는다  
* -exec: 찾아진 파일 등과 연계하여 어떠한 명령을 내릴 수 있는 옵션이다.  
* -ok: 확인메시지를 출력하여 명령 실행 시 'y'나 'n'을 입력받는다  
* -print: 화면으로 출력하기 한다. 생략 가능  
* -fprint: 출력 결과를 파일로 저장할 때 사용하는 옵션 '>' 기호를 이용하면 된다  
* -ls: 맨 뒤에 이 옵션을 사용하면 상세 출력한다  
* -size n: n블록 길이의 파일을 찾는다 +n은 n블록보다 큰 파일을 의미, -n은 n블록 보다 작은 파일  
* -inum n: node 번호가 n인 파일을 찾는다  
* -iname: 대소문자를 구별하지 않고 이름으로 찾을 때 사용  
* -maxdepth n: 검색 시 주어진 정수값 n만큼의 서브 디렉터리만 검색한다 
* -prune: 찾고자하는 대상이 디렉터리인 경우 하위 디렉터리의 검색을 하지 않는다  
* -empty: 파일의 크기가 0이거나 비어있는 디렉터리를 찾는다  
* -newer: 지정한 파일보다 뒤에 수정된 파일을 찾는다  
* -cnewer: 지정한 파일보다 뒤에 변화된 파일을 찾는다
