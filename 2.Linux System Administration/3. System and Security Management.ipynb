### 3.1 시스템 분석  
##### 3.1.1 시스템 로그 분석 및 관리  

* 시스템 로그의 개요  
시스템에서 일어나는 모든 사건이나 이벤트 등은 각 서비스별로 기록되는데, 이러한 기록들을 로그라 부른다. 최근 리눅스 배포판에서는 rsyslog 패키지를 사용한다  



* rsyslog  
1. rsyslog 의 개요  
rsyslogd 데몬이 동작하면서 로그를 기록한다    

[주요 파일]  
* /etc/rsyslog.conf: rsyslogd 데몬의 환경 설정 파일이다.
* /etc/sysconfig/rsyslog: rsyslogd 데몬이 실행과 관련된 옵션이 설정되는 파일  
* /usr/sbin/rsyslogd: 실제 rsyslogd 데몬 실행 명령이다.  
* /usr/lib/systemd/system/rsyslog.service: systemctl 명령으로 제어되는 유닛 파일이다.  

[주요 동작 명령어]  
* systemctl stop rsyslog: rsyslog 데몬의 동작을 중단시킨다 
* systemctl start rsyslog: rsyslog 데몬을 시작한다  
* systemctl -l status rsylog: rsyslog 데몬의 상태 정보를 출력한다  

[rsyslog 데몬 동작의 확인]  
* ps aux |grep rsyslog  

[rsyslog 데몬의 재시작]  
* systemctl restart rsyslog  

[부팅 시 rsyslog 데몬의 활성화]  
* systemctl enable rsyslog  

2. /etc/rsyslog.conf 파일  

[기본 구성 형식]  
* facility.priority action  

[facility의 종류]  
* cron: cron, at과 같은 스케줄링 프로그램이 발생한 메시지  
* auth, security: login과 같이 인증 프로그램 유형이 발생한 메시지  
* authpriv: ssh와 같이 인증이 필요한 프로그램 유형이 발생한 메시지  
* deamon: telnet, ftp등과 같이 여러 데몬이 발생한 메시지  
* kern: 커널이 발생한 메시지  
* lpr: 프린트 유형의 프로그램이 발생한 메시지 
* mail: mail 시스템이 발생한 메시지  
* mark: syslogd에 의해 만들어지는 날짜 유형  
* news: 유즈넷 뉴스 프로그램 유형이 발생한 메시지 
* syslog: syslog 프로그램 유형이 발생한 메시지 
* user: 사용자 프로세스 
* uucp: UUCP 시스템이 발생한 메시지 
* local0 ~ local7: 여분으로 남겨둔 유형  
* \*: 모든 facility를 의미  

[priority의 종류]  
* none: 지정한 facility를 제외. 앞부분에 모든 facility에 관한 설정을 하고, ; 기호 뒤에 특정한 facility를 제외할 때 사용
* debug: 프로그램을 디버깅할 때 발생하는 메시지 
* info: 통계, 기본 정보 메시지  
* notice: 특별한 주의를 필요하나 에러는 아닌 메시지 
* warning, warn: 주의가 필요한 경고 메시지 
* error, err: 에러가 발생하는 경우의 메시지 
* crit: 크게 급하지는 않지만 시스템에 문제가 생기는 단계의 메시지 
* alert: 즉각적인 조정을 해야 하는 상황  
* emerg, panic: 모든 사용자들에게 전달해야 할 위험한 상황 
* \*: 모든 priority를 의미  

[action의 종류]  
* file: 지정한 파일에 로그를 기록, /로 시작해서 절대 경로를 기재  
* @host: 지정한 호스트로 메시지를 UDP 기반으로 전달
* @@host: 지정한 호스트로 메시지를 TCP 기반으로 전달 
* user: 지정한 사용자가 로그인한 경우 해당 사용자의 터미널로 전달 
* \*: 현재 로그인 되어 있는 모든 사용자의 화면으로 전달 
* 콘솔 또는 터미널: 지정한 터미널로 메시지를 전달

* logrotate
1. logrotate의 개요  
로그 파일이 계속해서 덧붙여지면서 쌓이면 파일의 크기도 계속 커지게 되는데 이를 방지하기 위해 로그 파일을 여러 개로 분할하는 프로그램이다.  

사용법  
logrotate [option] config_file  

* -f: 강제로 환경 설정 파일을 읽어 들여서 실행한다  

2. /etc/logrotate.conf의 주요 설정  
* weekly: 로그 파일을 일주일마다 로테이트 한다  
* rotate 4: 최대 4번까지 rotate를 하는 설정이다.  
* create: 로테이트를 한 후에 비어있는 로그 파일을 생성하도록 설정하는 항목이다.  
* dateext: 로테이션으로 생성되는 로그 파일에 해당 날짜를 덧붙여서 생성하는 항목이다.  
* compress: 로테이트 한 후에 생성된 로그 파일에 대해 압축할 때 사용하는 항목이다.  
* include /etc/logrotste.d: /etc/logrotate.d 디렉터리 안에 설정된 파일에 대해서도 로테이트를 적용하는 설정이다.  
* nomissingok: 로그 파일이 존재하지 않는 경우에 에러 메시지를 출력한다
* missingok: 로그 파일이 존재하지 않는 경우에 에러 메시지를 출력하지 않고 다음 파일로 이동한다  

3. 관련 파일: /var/lib/logrotate.status: 각 로그 파일별로 로테이션된 날짜가 기록된 파일이다.  

##### 3.1.2 로그 관련 파일  
* 로그 관련 주요 파일  
1. /var/log/messages  
시스템에서 발생하는 표준 메시지가 기록되는 파일로 root 사용자만 읽을 수 있다.  

2. /var/log/secure  
인증을 기반으로 접속과 관련된 로그가 기록되는 파일이다.  

3. /var/log/dmesg  
시스템이 부팅할 때 출력되었던 로그가 기록되는데, 보통 커널 부트 메시지 로그라고 한다  

4. /var/log/maillog  
sendmail, postfix, dovecot 등과 같은 메일 관련 작업이 기록되는 로그 파일이다.  

5. /var/log/xferlog  
FTP 접속과 관련된 작업이 기록되는 파일로 로그 포맷은 총 14개의 항목으로 구성되어 있다.  

[로그 포맷]  
* current-time: 접속한 현재 시간을 나타낸다  

* transfer-time: 전송된 총 시간을 뜻하는데 초 단위로 표시된다  
* remote-host: 원격 호스트의 IP 주소가 기록된다.  
* file-size: 전송된 크기를 나타내며 단위는 바이트다  
* filename: 전송된 파일 이름이 절대 경로로 표시된다  
* transfer-type: 전송 형태를 뜻하는데 하나의 문자로 표시된다.  
* special-action-flag: 특정한 action이 발생하는 것을 나타내는데 하나의 문자로 표기된다. C는 압축된경우, U는 압축되지 않은 경우, T는 tar로 묶인 경우 _는 어떠한 action이 발생하지 않은 경우이다.  
* direction: 전송된 지시를 나타내는 것으로 o는 다운로드 된 경우, i는 업로드 된 경우이다.  
* access-mode: 사용자가 어떤 형태로 login 했는지 나타낸다 a는 익명, g는 패스워드를 소유한 손님 user, r는 인증한 로컬 사용자이다.  
* username: 사용자의 계정이 기록된다.  
* service-name: 발생한 service의 이름이 명기되는데 보통 ftp라고 기록된다  
* authentication-method: 인증에 사용되는 방법을 뜻하는데 0은 기본인증, 1은 RFC931  
* authentication-user-id: 인증 방법에 의해 되돌려지는 사용자 계정이 기록된다.  
* completion-status: 전송 상태를 하나의 문자로 나타내는데 c는 완전한 전송, i는 불완전한 전송이다.  

6. /var/log/cron  
cron 관련 정보가 기록되는 파일이다  

7. /var/log/boot.log  
부팅 시 발생하는 메시지가 기록되는 파일로 보통 부팅 시 동작하는 데몬 관련 정보가 기록된다  

8. /var/log/lastlog  
telnet이나 ssh를 이용해서 접속한 각 사용자의 마지막 정보가 기록되는 파일이다.  

9. /var/log/wtmp  
콘솔, telnet, ftp등 이용하여 접속한 사용자 기록, 시스템을 재부팅한 기록 등의 로그가 쌓임 

10. /var/log/btmp  
wtmp와 반대되는 로그로 접속이 실패한 경우를 기록되며 바이너리 파일로 되어 있고, lastb로 확인한다  


* 관련 명령어  
1. last  
사용자의 로그인 정보, 재부팅한 정보를 바이너리 파일로 생성하여 저장한다  

사용법  
last [option] [사용자명] [ttyn]  

* -f 파일명: 로그 로테이션 설정이 되어 있는 경우, 기본 로그 파일 이외의 다른 로그 파일의 기록을 볼 경우에 사용한다  
* -n 숫자: 가장 최근부터 해당 숫자값 만큼만 출력한다
* -t YYYYMMDDHHMMSS: 지정한 시간 이전에 로그인한 기록을 출력한다
* -R: IP 주소나 호스트명을 출력하지 않는다 
* -a: 호스트명이나 IP 주소 필드를 맨 마지막에 출력한다
* -d: 호스트 이름이 존재하는 경우에는 IP 주소를 호스트 이름으로 변환하여 출력한다
* -F: 로그인 및 로그아웃 시간을 출력한다  
* -i: 접속한 호스트의 IP 주소로만 출력한다
* -w: 사용자의 전체 이름이나 전체 도메인 이름을 전부 출력해준다.  


2. lastlog  
각각의 사용자가 마지막으로 로그인한 정보를 출력하는 명령으로 바이너리 파일인 /var/log/loastlog의 내용을 출력  

사용법  
lastlog [option]  

* -u 사용자명: 특정 사용자에 대한 정보만 출력한다
* -t 날짜수: 오늘부터 지정한 날짜 수 만큼 거슬러 올라가 그 이후에 로그인한 사용자의 정보를 출력한다
* -b 날짜수: 오늘을 기준으로 지정한 날짜수 이전에 최종적으로 로그인한 사용자 정보를 출력한다

3. lastb  
last와 반대되는 개념으로 로그인에 실패 정보를 /var/log/btmp에 기록하는데 이 파일의 내용을 출력하는 명령이다.  

사용법  
lastb [option] [사용자명]  

* -f 파일명: 로그 로테이션 설정이 되어 있는 경우, 기본 로그 파일 이외의 다른 로그 파일의 기록을 볼 경우에 사용한다  
* -n 숫자: 가장 최근부터 해당 숫자값 만큼만 출력한다
* -YYYYMMDDHHMMSS: 지정한 시간 이전에 로그인한 기록을 출력한다
* -R: IP 주소나 호스트명을 출력하지 않는다
* -a: 호스트명이나 IP 주소 필드를 맨 마지막에 출력한다
* -d: 호스트 이름이 존재하는 경우에는 IP 주소를 호스트 이름으로 변환하여 출력한다  
* -F: 로그인 및 로그아웃 시간을 출력한다
* -i: 접속한 호스트의 IP 주소로만 출력한다
* -w: 사용자의 전체 이름이나 전체 도메인 이름을 전부 출력한다  

4. dmesg  
커널 링 버퍼의 내용을 출력하고 제어하는 명령이다.  

사용법  
dmesg [option]  
* -c: 커널 링 버퍼에 저장된 메시지를 출력한 후에 지운다  

5. logger  
명령행에서 로그 시스템에 메시지를 전송할 떄 사용하는 명령으로 /var/log/message 파일에 기록된다  

사용법  
logger [option] [message]  

* -i: PID를 기록할 떄 사용한다
* -f: 저장되는 파일명 지정할 떄 사용한다
* -t: 지정한 태그를 함께 기록할 때 사용한다
* -s: 표준 출력에도 메시지를 출력한다  

### 3.2 시스템 보안 관리  
##### 3.2.1 시스템 보안 관리  

* 리눅스와 보안 개요  
1. 물리적 보안 설정  
보안에 있어 가장 큰 선결 조건은 물리적 보안이다. CCTV 같은 감시 체계를 갖추고, 잠금장치가 있는 랙을 이용하여 물리적 접근을 최대한 막아야 한다.  

다음으로 BIOS 보안도 설정하면 좋은데 ROM BIOS 또는 CMOS Setup 이라고도 불리며 하드웨어를 점검하여 소프트웨어인 운영체제에 연결하는 기능을한다. 부팅시 [Delete] or [F2]를 눌러 접근할 수 있다.  

2. 불필요한 서비스 제거  
응용 프로그램의 버그를 통한 해킹 시도가 빈번하므로 불필요한 서비스들을 제거해서 위험 요소를 최대한 줄일 필요가 있다.  

3. 시스템 정보 감추기  
로컬이나 외부에서 접속 할때 /etc/issue 및 /etc/issue.net 메시지 파일의 내용 변경을 통해 시스템 정보 노출을 최소화 한다  

4. root 패스워드 변경 제한  
grub에 패스워드를 설정하여 임의로 root 패스워드를 변경하는 것을 막을 수 있다.  

5. 사용자 권리  
Jhon the ripper와 같은 도구를 사용해서 손쉬운 암호를 설정한 사용자를 찾아 변경하도록 한다. 또한 su와 같은 명령 제한을 위한 PAM을 이용하거나, 일부 명령 할당을 위한 sudo 등의 보안 프로그램을 이용하는 것이 좋다  

6. 보안이 강화된 서빗로 대체 이용  
텔넷보다는 ssh를 이용하도록 한다  

7. 파일 시스템 관리  
* Set-UID, Set-GID, Sticky-Bit 등과 같은 특수 권한은 검색을 통해 필요 없는 경우에는 제거하는 것이 좋다.  

* 파티션이 분할된 영역인 경우에는 /etc/fstab 파일에 Set-UID나 Set-GID의 설정을 제한하는 nosuid 옵션을 지정함으로써 막을 수 있다.

* 커널 매개 변수와 sysctl  
1. 커널 매개 변수의 개요  
핵심적인 역할을 수행하는 커널과 관련된 다양한 매개 변수를 제공하고 있다.  

커널 제어를 위한 매개 변수들을 /proc/sys 디렉터리에 보관하여 관리하고 있는데,
* fs: 커널에서 최대로 사용 가능한 파일의 수 지정과 같은 파일 시스템 관련 매개 변수가 들어있는 디렉터리 
* kernel: IPC와 관련된 내용들을 설정하는 매개 변수가 들어있다  
* net: TCP, ICMP등 네트워크 관련 매개 변수가 들어있다.  
* vm: 스왑, 페이지등 메모리 관련 변수가 들어있는 디렉터리이다.  

2. 커널 매개 변수값의 확인및 변경  
* cat /proc/sys/fs/file-max: 사용가능한 파일의 수가 있으며 일반적으로 매개 변수에 해당하는 수치값이 명시되어 있다.  

* 커널 매개 변수의 값 변경: syscrl 명령어를 이용하거나 /proc/sys 디렉터리 하위에 있는 항목들을 echo 명령과 > 기호를 사용해서 변경할 수 있다.  

* IPC  
시스템 성능 향상과 관련된 항목으로 프로세스간 통신이라는 뜻이며 프로세스들 사이에 서로 데이터를 주고받는 행위 또는 그에 대한 방법이나 경로를 통칭하는 말이다.  

1. 공유 메모리  
통신하는 프로세스들 사이에 특정 변수를 공유하고 이 공유 변수를 이용해서 정보를 교환하는 방법이다. 사용자가 단위를 지정하여 한개 이상의 영역을 나누어 설정할 수 있으며 한 개의 공유 메모리 영역을 세그먼트라고 부른다  

2. 메시지 시스템 기법  
운영체제가 프로세스 간의 메시지를 교환할 수 있도록 관리하는 방법으로 메시지를 메시지 큐에 저장하여 전달하는 방법이다.  

3. 세마포  
운영체제 상에서 프로세스들이 공유 자원에 대한 접속을 제어하기 위해 사용되는 신호를 의미하며 프로세스 간의 동기화를 구현하기 위해 사용된다.  

4. 상호 배제(Mutex)  
여러 프로세스들이 하나의 공유 자원을 상호 베타적으로 사용할 수 있으나, 동시에 사용할 수 없도록 하는 것이다.

1. 파일 시스템 관련  
파일 시스템 관련 커널 매개 변수는 /proc/sys/fs 디렉터리에 들어있다.  

* file-max: 커널에서 최대로 사용 가능한 파일의 숫자값이 들어있다. 
* file-nr: 파일을 열 때 사용하는 파일 디스크립터 관련 매개변수이다.  

2. IPC 지원 관련  
IPC 관련 매개 변수는 /proc/sys/kernel 디렉터리에 들어 있다.  

2-1. 공유 메모리 관련변수  
* 페이지: 주소 공간을 일정한 크기로 나눈 블록 
* 세그먼트: 고정된 단위가 아니고 논리적으로 관련 있는 정보의 단위로 프로그램에서 분할하여 사용  

* shmmax: 프로세스가 사용할 수 있는 공유 메모리의 최대 크기로 단위는 바이트이다  
* shmmni: 시스템 전체에서 사용할 수 있는 공유 메모리 세그먼트의 최대 개수이다  
* shmall: 모든 프로세스가 사용할 수 있는 공유 메모리 페이지의 최대 개수이다  

  
* 리눅스에서 기본 페이지 값을 확인하는 명령어 
getconf PAGE_SIZE  

2-2. 세마포 관련 변수  

세마포 오퍼레이션: 메모리에 대한 동시성을 제어할 필요가 생길때가 있는데 이 경우에 메모리에 대한 읽기 및 쓰기에 대한 접근권한의 제어를 위해 발생한다  

* sem: 이 변수 안에는 4개의 설정값이 들어있다.  
* SEMMSL: 한개의 세마포 세트에 존재하는 세마포의 최대 개수 
* SEMMNS: 전체 시스템 최대 세마포 수 
* SEMOPM: 세마포 콜에서 처리하는 최대 운영 수 
* SEMMNI: 세마포 세트의 수  

2-3. 스와핑 관련 변수  
운영체제에서는 성능적인 향상을 위해 사용하는 기술로 스와핑과 캐시를 손꼽을 수 있다. 이 두 기술과 연관된 커널 매개변수는 /proc/sys/vm 디렉터리에 저장하고 있다.  

스와핑은 주기억장치보다 더 큰 응용 프로그램이나 데이터 파일을 처리하기 위해 디스크의 일부를 메모리처럼 사용하는 기술이다.  

* swappiness: 스왑 성향을 0-100 수치로 정의, 값이 작을수록 적은양의 스왑이 발생  

2-4. 캐시 관련 변수  
캐시는 주기억장치와 보조기억장치 간의 속도 차이로 인한 병목 현상을 해소하기 위해 운영체제에서 관리하는 일종의 시스템 버퍼이다.  

리눅스 커널 2.6.16 버전 부터는 czche로 사용되는 메모리 영역을 즉시 회수할 수 있는 커널 매개 변수를 제공하는데 이 변수가 /proc/sys/vm/drop_caches 이다.  

* drop_caches  
1로 설정: page chche 제거  
2로 설정: inode 및 dentry cache 제거  
3으로 설정: page, inode, dentry cache 모두 제거  

캐시 메모리 회수는 free 명령으로 확인할 수 있다.  

3. 네트워크 관련  
네트워크 관련 매개변수는 /proc/sys/net 디렉터리에 위치한다.  

[주요 하위 디렉터리]

* sysctl  
1. 개요  
커널 매개변수의 값을 제어하여 시스템을 최적화할 수 있는 명령으로 /proc/sys 디렉터리에 존재하는 매개변수를 제어하는 명령이다.

##### 3.2.2 관련 명령어  

* ssh  
1. ssh의 특징  
* 패킷을 암호화하여 다른 원격 로그인 프로그램인 텔넷이나 rlogin에 비해 안전하다
* rlogin처럼 패스워드 입력 없이 로그인할 수 있다 
* rsh 처럼 원격 셸을 지원한다
* 원격 복사(scp)를 지원한다
* 안전한 파일 전송(sftp)을 지원한다  

2. ssh 설치  
* openssh: openssh 서버와 openssh 클라이언트에 모두 필요한 핵심적인 파일들이 들어있는 패키지로 ssh-keygen 명령이 들어있다 
* openssh-server: openssh 서버 관련 패키지로 데몬인 sshd, sftp 서버 등이 들어 있다 
* openssh-clients: openssh 클라이언트 관련 패키지로 ssh, scp, sftp 등의 명령어가 들어있다  

3. ssh 서버 설정  
ssh 서버의 환경 설정 파일은 /etc/ssh/sshd_config 이고 실행 명령은 'systemctl start sshd.service' 이다.  

4. ssh 클라이언트 명령어: ssh  
openssh-clients 패키지를 설치하면 ssh 명령을 이용해서 접속할 수 있다.  

사용법  
ssh [option] 호스트명 or IP_주소  
ssh 계정이름@호스트네임  
ssh 호스트네임 명령  

* -l: 다른 계정으로 접속할 때 사용한다. 이 옵션 대신 서버 주소 앞에 @를 붙일 수도 있다  
* -p: ssh 서버의 포트 번호가 22번이 나닌 경우 바뀐 포트를 지정할 때 사용한다  

6. 인증키를 이용한 ssh 서버 접속: ssh-keygen  
ssh 클라이언트에서 ssh-keygen 명령을 이용해서 비밀키와 공개키를 생성하고, ssh 서버에 공개키를 복사하면 된다  

사용법  
ssh-keyhen [option]  

* -t: 사용할 암호화 알고리즘을 지정하는 옵션이다 (rsa, dsa 등을 사용할 수 있음)  

7. 원격지 공개키 복사: ssh-copy-id  
ssh-keygen으로 생성된 공개키 파일을 원격지 서버의 홈 디렉터리 내에 인증키 파일로 복사하는 명령이다.  

사용법  
ssh-copy-id [option] 공개키_파일_원격지_주소  

* -i: 공개키 파일을 지정할 때 사용한다 
* -f: 강제로 인증키 파일을 추가한다
* -p: 기본 포트인 22번이 아닌 경우 특정 포트 번호를 지정할 떄 사용한다  

* PAM  
1. PAM의 개요  
사용자를 인증하고 그 사용자의 서비스에 대한 접근을 제어하는 모듈화된 방법을 말한다  

2. PAM의 구성  
2-1. 기본 구성  
PAM에서 제공하는 라이브러리는 /usr/lib64/security에 위치하고 동적으로 로드 가능한 오브젝트 파일 형태로 되어 있다.  

2-2. PAM 설정 파일의 구성  
2-2-1. type: PAM에 어떤 타입의 인증이 사용될 것인지를 알려주는 항목이다  
* account: 사용자가 해당 서비스에 접근이 허용되는지 여부, 패스워드 기간 만료 여부 검사  
* auth: 기본적으로 패스워드를 통해 인증하지만, 생체인증과 같이 다른 인증을 통해 자신임을 주장하는 사용자가 맞는지를 검사한다  
* password: 사용자가 그들의 인증 방법을 변경하도록 할 때 제공하는 방법으로 패스워드를 사용한다  
* session: 사용자가 인증받기 전후에 해야 할 것을 나타낸다  

2-2-2. control: 통제를 담당하는 부분으로 PAM에 무엇을 해야 할지를 알려준다 
* requisite: 이 모듈을 이용하는 인증이 실패할 경우에 즉시 거부한다  
* required: PAM 모듈의 인증이 실패해도 이후 모듈 실행을 계속하지만, 최종적으로 인증 결과에 영향을 미친다  
* sufficient: 이전에 요청된 모듈이 실패하더라도 이 모듈에 의해서 인증이 성공할 경우에 인증을 승인한다  
* optional: 서비스에 대한 응용 프로그램의 성공/실패가 중요하지 않다는 것을 의미한다  
* include: 이 항목 다음에는 모듈명 대신에 다른 PAM 관련 서비스가 오는데, 해당 서비스 인증을 통과해야 가능하도록 설정할 떄 사용한다 
* substack: 이 컨트롤의 인수로 지정된 구성 파일에서 지정된 유형의 모든 행을 포함한다  

3. PAM의 주요 모듈  
* pam_securetty.so: 접속하는 계정이 root인 경우 /etc/securetty 파일에 기록된 터미널을 통하는 경우에만 허가하도록 한다  
* pam_lisfile.so: 임의의 파일에 대해 서비스를 허가하거나 거부하는 방법을 제공한다  
* pam_nologin.so: /etc/nologin 파일이 존재하면 root만 로그인할 수 있다  
* pam_deny.so: 접근을 무조건 거부할 때 사용하고, 응용 프로그램에게 항상 실패를 리턴한다  
* pam_wheel.so: root 권한을 얻을 수 있는 사용자를 wheel 이라는 그룹으로 묶어서 사용하도록 지원하는 모듈이다  
* pam_rootok.so: root가 암호 입력 없이 해당 서비스에 대한 접근을 허용할 때 사용되는 모듈이다  

* sudo  
1. sudo의 개요  
특정 사용자 또는 특정 그룹에 root 사용자 권한을 가질 수 있도록 일부 명령 또는 모든 명령을 실행할 수 있도록 해주는 도구이다.

sudo 명령

* 파일 시스템 보안  

1. lsattr  
파일에 설정된 속성을 확인할 떄 사용하는 명령이다  

사용법  
lsattr [option] [파일명]  

* -R: 하위 디렉터리까지 한 번에 속성을 확인할 때 사용한다 
* -a: .으로 시작되는 숨김 파일과 디렉터리까지 확인할 떄 사용한다  

2. chattr  
파일의 속성을 변경하는 명령으로 root 사용자만 사용 가능하다  

사용법  
chattr [option] mode 파일명  

* -R: 하위 디렉터리까지 한 번에 속성을 변경할 때 사용한다  

[mode]  
기호와 속성으로 이루어지는데 기호는 +,-,=를 사용한다  
+: 해당 속성을 부여, -: 해당 속성 해제, =: 해당 속성만 부여하고 해제  

* A: 파일 수정 시에 atime은 수정하지 않는다  
* a: 해당 파일에 추가만 가능하도록 설정한다 
* d: dump로 백업되지 않도록 설정한다 
* i: 해당 파일의 변경, 삭제, 이름 변경, 파일 추가, 등을 불가능하게 만든다 
* S: 파일이 변경될 경우에 디스크 동기화가 일어나도록 할때 사용한다
* e: 디스크 블록에 매핑하기 위해 확장된 파일임을 표시한다  

3. ACL  
파일이나 디렉터리에 접근 권한을 제어할 수 있도록 만든 시스템이다  

3-1. getfacl  
파일이나 디렉터에 설정된 접근 권한 리스트를 확인하는 명령  

사용법  
getfacl [option] [파일명]  

* -d: 접근 권한 리스트의 기본값을 출력한다 

3-2. setfacl  
파일이나 디렉터리에 접근 권한 리스트를 설정하는 명령으로 root 사용자만 이용할 수 있다.  

사용법  
setfacl [option] [파일명]  

* -m: 권한을 지정하거나 수정할 때 사용한다 
* -x: 권한을 삭제할 때 사용한다  
* -R: 하위 디렉터리와 파일까지 권한을 변경할 때 사용한다 
* -b: 권한 및 mask 등 지정한 권한을 전부 제거한다  

* 주요 보안 도구 소개  
1. nmap  
네트워크 탐지 도구 및 보안 스캐너로 시스템의 서비스 중인 포트를 스캔하여 관련 정보를 출력한다. 이 도구를 사용하면 운영 중인 서버에 불필요하게 작동하고 있는 서비스 포트를 확인할 수 있다  

사용법  
nmap -O -p 1-65535 localhost  
(로컬 호스트의 운영체제와 포트 1번 부터 65535번까지 점검해서 결과를 출력한다)  

2. tcpdump  
명령행에서 사용하는 네트워크 트래픽 모니터링 도구로서 특정한 조건식을 설정하여 네트워크 인터페이스를 거치는 패킷들의 헤더 정보를 출력할 수 있다  

3. nessus  
서버의 보안 취약점을 검사하는 도구로서 문제가 되는 서비스에 대한 정보를 알려주고, 관련 사이트나 대처 방안 등을 제시하는 프로그램이다  

4. GnuPG  
공개키와 비밀키를 생성하여 암호화하는 기법인 OpenPGP를 공개 버전으로 구현한 것으로 데이터의 암호화와 디지털 서명 생성 등에 사용된다  

5. Jhon The Ripper  
유닉스 계열 패스워드 크랙 도구이다. 아뫃로 사용될 만한 목록이 들어있는 텍스트 형태의 사전 파일을 이용해서 일일이 대입한 후에 암호화된 패스워드가 들어있는 /etc/shadow와 비교해서 사용자의 패스워드를 알아내는 방식이다  

* SELinux  
1. 개요  
특정 데몬의 버그를 통해 root 권한을 획득하더라도 해당 데몬에만 root 권한을 행사할 뿐 다른 데몬이나 시스템에는 접근이 불가능하도록 하여 시스템의 보안을 강화하는데 있다.  

2. SELinux의 적용

셸 상에서는 getenforce 및 setenforce 명령으로 확인 및 설정 가능하다 


### 3.3 시스템 백업  
##### 3.3.1 정책 수립  

1. 백업의 개요  
자료를 보호하는 하나의 방법으로 여러 개의 자료 복사본을 만들어 놓고 자료를 잃어버릴 경우에 대비하는 것이다  

2. 백업의 종류  
전체백업: 특정 디스크나 파티션등 전체를 백업하는 것 
부분 백업: 선택한 파일들만 백업하는 것인데, 증가된 내용만 백업하는 증분 백업과 바뀐 부분만을 백업하는 차등 백업으로 구분할 수 있다  

3. 매체 선택 여부  
어떠한 백업 매체를 사용할 것인지를 결정해야 한다. 매체는 경제성, 효율성, 신뢰성, 편리성, 처리속도, 가용성 등을 고려해서 선택해야 한다.  

##### 3.3.2 Backup 및 Restore  

* tar  
유닉스와 리눅스 시스템 모두 사용 가능한 유틸리티로 파일이나 디렉터리들을 하나의 파일로 묶어 주는 역할을 한다. 사용이 쉽고 소량의 파일 백업에 좋으며 원본 파일을 남겨두므로 안전하다  

1. 전체 백업  

[전체 백업 예]  
* tar cvfp home.tar /home  
/home 디렉터리를 펌션 등을 그대로 유지하면서 home.tar라는 파일로 묶는다  

[전체 백업 복원 예]  
* tar xvf home.tar  
home.tar 파일을 풀어서 현재 디렉터리에 복원한다  

2. 증분 백업  

[증분 백업 예]  
* tar -g list -cvfp home2.tar /home  
list 라는 파일의 내용과 비교하여 증가한 것만 home.tar로 백업한다  

[증분 백업 복원 예]  
* tar xvf home2.tar -C /  
처음에 백업한 파일부터 순차적으로 복원하면 된다  

3. 날짜를 이용한 부분 백업  
[부분 백업 예]  
* tar -c -v -N '13 May 2013' -f home.tar /home  
/home에서 2013-5-13 이후로 변경된 파일만을 home.tar로 백업한다  

4. 분할 및 압축 백업  

[압축 및 분할 백업 예]  
* tar zcvf -/home | split -m 10m -home.tar.gz  
/home 디렉터리를 압축하여 10MB 단위로 백업한다  

[복원 예]  
* cat home.tar.gza* | tar zxvf -  

* cpio  
tar와 비슷한 유틸리티로 많은 양의 데이터에 대해서는 tar 보다 빠르다  

사용법  
[ 셸명령어 | ] cpio option > 파일명  
cpio option < 파일명  

* -o: 표준 출력으로 보내어 사용한다 
* -i: 표준 입력으로 받을 때 사용한다. 백업한 자료를 불러올 떄 사용한다 
* -v: 과정을 상세히 출력한다 
* -c: 아카이브 포맷 형식을 'new SVR4 portable with no CRC'로 지정한다 
* -d: 필요할 경우 디렉터리를 생성한다 
* -t: 내용만 확인할 때 사용한다 
* -F: 표준 입출력 전환 기호 대신에 파일명을 지정할 때 사용한다
* -B: 입출력 블록 사이즈를 조절할 때 사용한다 
* -H 포맷: 아카이브 포맷 형식을 지정하는 옵션이다.  

* dump 및 restore  
파일들이 아닌 파일 시스템 전체를 백업할 떄 사용하는 유틸리티로 보통 파티션 단위로 백업할 때 많이 사용한다. 전체 백업과 증분 백업을 지원한다  

1. dump  
파티션 단위로 백업하는 명령이다  

사용법  
dump option 파일명 백업대상  

* -0~9: 레벨을 지정한다
* -f: 백업할 매체나 파일명을 적는다 
* -u: dump 작업 후에 /etc/dumpdates 라는 파일에 관련 정보를 기록한다  

2. restore  
dump로 백업한 내용을 복원할 때 사용하는 명령이다  

사용법  
restroe option 백업파일명  

* -i: 대화식으로 복구할 파일을 선택한 후에 복원할 때 사용한다
* -f: 백업할 매체나 파일명을 적는다 
* -r: 전체 복원시에 사용한다  

* dd  
파티션이나 디스크 단위로 백업할 때 사용하는 유틸리티로 사용하기는 쉬우나 많은 시간이 소모된다  

* rsync  
네트워크로 연결된 원격지의 파일들을 동기화하는 유틸리티로 예전에 사용하던 rcp에 비해 처리속도가 빠르다  

1. 링크된 파일이나 디바이스 파일도 복사 가능하고, 그룹을 포함한 소유권 및 허가권도 유지하여 복사할 수 있다  
2. GNU tar와 유사한 옵션을 사용하며, 특정한 부분만 백업할 수 있다.  
3. CVS의 exclude 모드처럼 같은 파일들은 제외할 수 있다 
4. ssh나 rsh을 이용하여 전송 가능하고 root 권한이 필요하지 않다 
5. 내부 파이프라인을 통하여 전송 시간을 줄인다 
6. 익명 사용자 및 인증된 사용자를 지원하여 미러링에 이상적이다 
7. 로컬 시스템의 백업 시에는 별다른 서버 설정 없이 사용할 수 있다  

사용법  
rsync [option] source destination  

* -r: 하위 디렉터리까지 실행한다 
* -l: 심볼릭 링크를 그대로 보존한다 
* -L: 심볼릭 링크가 참고하고 있는 파일들을 복사한다 
* -p: 퍼미션을 그대로 보존한다
* -t: 타임스탬프를 그대로 보존한다
* -g: 그룹 소유권을 그대로 보존한다 
* -o: 소유권을 그대로 보존한다
* -D: 디바이스 파일을 그대로 보존한다 
* -H: 하드링크를 그대로 보존한다
* -a: -rlptgoD를 한번에 실행할 때 사용하는 옵션이다 
* -v: 진행 상황을 자세히 출력한다
* -u: 업데이트된 내용만 전송한다
* -z: 전송할 때 압축한다
* -b: 백업할 때 동일한 파일이 존재하는 경우에 ~를 붙여서 백업 파일을 생성한다
* -e: 원격지에서 사용할 셸 명령을 기재한다
* --progress: 명령이 실행되는 동안의 전송 상황 정보를 출력한다
* --delete: 송신측에서 없는 파일이 수신측의 백업 디렉터리에 존재하면 삭제한다
