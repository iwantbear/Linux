### 2.1 장치의 설치 및 관리  
##### 2.1.1 장치의 설치 및 관리 

* 모듈  
1. 모듈의 개요  
프로그램이나 하드웨어 기능 단위, 교환할 수 있는 구성 부분, 우주선의 일부를 이루지만 독립적으로 행동할 수 있는 작은 소선을 뜻한다. 리눅스에서의 모듈은 우주선의 경우와 비슷한데, 우주선의 모체는 커널 이미지이고, 작은 소선이 커널 모듈이다. 보통 모듈이 로드되었다는 것은 커널 밖에 존재하던 해당 드라이버가 커널의 일부로 되었다는 뜻이고, 언로드 하면 다시 커널과 별도로 존재한다.  

2. 모듈 관리  
리눅스 시스템에서 시스템 장치를 제어하는 코드는 컴파일되어 커널에 포함될 수 있다.  
모놀리식은 커널이 시스템 장치를 제어하는 기능을 내장하고 하나의 덩어리 형태로 구성되는 방식을 말한다.  

3. 모듈 관련 명령어  
3-1. lsmod  
리눅스 커널에 적재된 모듈 정보를 출력하는 명령으로 모듈명, 크기, 다른 모듈에서 사용중인 수, 사용 중인 모듈 순으로 출력한다  

사용법  
lsmod  

3-2. insmod  
커널에 모듈을 적재하는 명령으로 해당 모듈을 자동으로 검색되고 삽입한다.  

사용법  
insmod 모듈파일명  

3-3. rmmod  
커널에서 모듈을 제거하는 명령으로 다른 모듈에 의해 사용 중인 모듈은 제거할 수 없다.  

사용법  
rmmod 모듈  

3-4. modprobe  
리눅스 커널에 모듈을 적재하거나 제거하는 명령이다. 단일 모듈, 의존성이 있는 여러 모듈, 특정 디렉터리의 모든 모듈들을 적재할 수 있다.  

사용법  
modprobe [option] 모듈 [기호=값]  

* -r: 모듈을 제거할 때 사용하는 옵션으로 의존성이 있는 모듈들을 찾아서 사용되지 않는다면 자동으로 제거한다 
* -c: 모듈 관련 환경 설정 파일의 내용을 전부 출력한다 

3-5. modinfo  
모듈 파일에 대한 정보를 출력하는 명령이다.  

사용법  
modinfo [option] 모듈파일명  

4. 모듈 관련 설정 파일  
리눅스 커널 2.4 -> /etc/modprobe.conf  
리눅스 커널 2.6 -> /etc/modprobe.d 디렉터리 안에 '.conf'로 끝나는 모든 파일 인식  
RHEL 7 -> /etc/modprobe.d 및 /lib/modprobe.d 디렉터리 내의 '*.conf' 파일 내용을 인식  

5. 모듈 의존성 파일 : modules.dep  
모듈 간의 의존성을 기록한 파일로 '/lib/modules/커널버전' 디렉터리 안에 위치한다  

6. 모듈 의존성 관련 명령어 : depmod  
커널 모듈 간의 의존성을 관리하는 명령어로 '/lib/modules/커널버전/modules.dep' 파일과 맵 파일을 생성  

사용법  
depmod [option] [파일명]  

* -a: modules.dep와 맵 파일을 새롭게 생성한다
* -A: 새롭게 추가된 모듈이 있는 경우에만 갱신한다
* -h: 명령어의 사용법을 화면에 출력한다  

* 커널  
1. 커널의 개요  
리눅스에서 커널은 시스템 자원을 소유하고 관리하는 역할을 담당한다  

2. 커널 컴파일  
커널 컴파일은 커널 소스를 다운로드하여 사용하는 시스템에 최적화된 커널을 만드는 과정이다. 이러한 과정을 통해 불필요한 항목들을 제거하고, 안정성과 성능 향상을 기대한다.  

[커널 컴파일 순서]  
* 커널 소스 파일 압축 해제 
* 커널 소스의 설정값 초기화 
* 커널 컴파일 옵션 설정 작업
* 커널 이미지 파일 생성 작업
* 커널 모듈 생성을 위한 컴파일 작업
* 커널 모듈 설치 작업
* 커널 모듈 파일 복사, grub 환경 설정 파일 수정 
* 재부팅  

[cleaning Targets]  
* make clean : 오브젝트 파일을 제거하고 다시 configure할 때 쓰이는 명령이다.  
* make mrproper : configure 작업을 통해 생성된 오브젝트 파일 뿐만아니라 config 파일, 다양한 백업 파일등도 제거한다
* make distclean : mrproper 작업과 더불어 편집된 백업 및 패치 파일도 모두 제거한다  

[커널 컴파일 주요 도구]  
* make config: 텍스트 기반의 설정 도구로 터미널 환경에서 y,m,n으로 설정한다  
* make menuconfig: 텍스트 기반의 컬러 메뉴를 제공하고, 커서를 이용해서 이동이 가능하다.  
* make nconfig: 텍스트 기반의 컬러 메뉴를 제공하고, 커서와 [F1] ~ [F9] 까지의 기능키를 제공하는 도구이다 
* make xconfig: X 윈도 환경에서 Qt 기반으로 설정하는 도구  
* make gconfig: X 윈도 환경에서 Gtk+ 기반으로 설정하는 도구  

### 2.2 주변 장치 관리  
##### 2.2.1 주변 장치 설정  

* 프린터  
1. 프린팅 시스템의 개요  
리눅스에서 프린터를 지원하는 인쇄 시스템으로 초기에는 LPRng를 기본으로 사용했으나, 최근 배포판에서는 CUPS 라는 시스템을 추가로 사용하고 있다.  

[CUPS 관련 파일]  
* /etc/cups/cupsd.conf: CUPS 프린터 데몬의 환경 설정 파일 
* /etc/cups/printers.conf: 프린터 큐 관련 환경 설정 파일로 lpadmin 명령 사용  
* /etc/cups/classes.conf: CUPS 프린터 데몬의 클래스 설정 파일

2. 프린터의 설정  
좌측 상단에 있는 [프로그램] -> [시스템 도구] -> [설정] -> [장치] -> [프린터]  

3. 프린터 관련 명령어  
3-1. lpr  
프린터 작업을 요청하는 명령  

사용법  
lpr [option] [파일명]  

* -# 값: 인쇄할 매수(1-100)를 지정하는 옵션이다 
* -m: 작업이 완료되면 관련 정보를 E-mail로 전송한다 
* -P 프린터명: 기본 설정된 프린터 이외에 다른 프린터를 지정할 때 사용한다
* -T: 타이틀 페이지에 들어갈 타이틀명을 설정한다
* -r: 출력한 뒤에 지정한 파일을 삭제한다
* -l: 필터링 없이 직접 보낸다  

3-2. lpq  
프린터 큐에 있는 작업의 목록을 출력하는 명령  

사용법  
lpq [option]  

* -a: 설정되어 있는 모든 프린터의 작업 정보를 출력한다
* -l: 출력 결과를 자세학 출력한다
* -P 프린터명: 특정 프린터를 지정할 때 사용한다

3-3. lprm  
프린터 큐에 대기 중인 작업을 삭제하는 명령으로 취소할 프린트 작업의 번호를 입력한다  

사용법  
lprm [option] [파일명]  

* --: 프린터 큐에 있는 모든 작업을 취소한다 
* -U 사용자명: 지정한 사용자의 인쇄 작업을 취소한다
* -P 프린터명: 특정 프린터를 지정할 때 사용한다
* -h 서버[:port]: 지정한 서버의 인쇄 작업을 취소한다  

3-4. lpc  
라인 프린터 컨트롤 프로그램으로 프린터나 프린터 큐를 제어할 때 사용한다  

* disable: 새로운 프린터 작업을 할 수 없도록 한다 
* enable: 프린트 작업을 가능하게 한다
* down: 지정한 프린터를 사용할 수 없게 한다
* up: 모든 환경을 활성화시키고, 관련 데몬을 새롭게 구동한다
* help, ?: 사용 가능한 명령을 출력한다
* quit, exit: lpc 명령을 종료한다

3-5. lp  
System V 계열에서 사용하는 인쇄 명령으로 BSD 계열의 lpr 명령과 유사하다  

사용법  
lp [option] [파일명]  

* -d: 다른 프린터를 지정한다
* -n: 인쇄할 매수를 지정한다 (1-100)  

3-6. lpstat  
프린터 큐의 상태를 출력하는 명령이다  

사용법  
lpstat [option]  

* -p: 프린터의 인쇄 가능 여부를 출력한다 
* -t: 프린터의 상태 정보를 출력한다
* -a: 받아들이는 요청들의 상태를 출력한다  

3-7. cancel  
프린트 작업을 취소하는 명령으로 lpstat를 이용하여 먼저 요청 ID를 확인해야 한다  

사용법  
cancel 요청-아이디  

* -a: 모든 인쇄 작업을 취소한다  

* 사운드 카드  
1. 고급 리눅스 사운드 아키텍처 (ALSA)  
사운드 카드용 장치 드라이버를 제공하기 위한 리눅스 커널의 요소이다.  

2. 오픈 사운드 시스템 (OSS)  
리눅스 및 유닉스 계열 운영체제에서 사운드를 만들고 캡쳐하는 인터페이스이다.  

3. 사운드 관련 명령어  

3-1. alsactl  
ALSA 사운드카드를 제어하는 명령이다  

사용법  
alsactl [option] [command]  

* -d: 디버그 모드를 사용한다
* -f: 환경 설정 파일을 선택한다. 기본 파일은 /etc/asound.state 이다.  
* store: 사운드카드에 대한 정보를 환경 설정 파일에 저장한다 
* restore: 환경 설정 파일로부터 선택된 사운드카드 정보를 다시 읽어들인다
* init: 사운드 장치를 초기화한다  

3-2. alsamixer  
커서 라이브러리 기반의 ALSA 사운드카드 오디오 믹서 프로그램이다.  

사용법  
alsamixer  

3-3. cdparanoia  
오디오 CD 에서 음악 파일을 추출할 때 사용하는 명령이다.  

사용법  
cdparanoia [option]  

* -w: wav 파일로 추출한다
* -a: Aplle AIFF-C 포맷으로 추출한다
* -B: 모든 트랙의 음악을 Cdda2wav 스타일로 추출한다 'track#.' 형태로 파일 이름이 생성된다  

* 스캐너  
1. SANE  
평판 스캐너, 핸드 스캐너, 비디오 캠 등 이미지 관련 하드웨어를 사용할 수 있도록 해주는 API 이다.  

2. XSANE  
SANE 스캐너 인터페이스를 이용하여 X-Window 기반으로 만든 프로그램이다.  

3. 스캐너 관련 명령어  

3-1. sane-find-scanner  
USB 및 SCSI 스캐너와 관련 장치 파일을 찾아주는 명령이다.  

사용법  
sane-find-scanner [option] [장치파일명]  

* -q: 스캐너 장치만 출력한다
* -v: 자세한 정보를 출력한다
* -p: 병렬 포트에 연결된 스캐너만 찾는다  

3-2. scanimage  
이미지를 스캔하는 명령이다.  

사용법  
scanimage [option]  

* -h: 도움말 옵션으로 사용 가능한 옵션 목록을 출력한다
* -d: SANE의 장치 파일명을 적는 옵션이다.  
* -L: 사용 가능한 스캐너 장치 목록을 출력하낟
* --format: 이미지 형식을 지정하는 옵션으로 pnm과 tiff를 지정할 수 있다.  

##### 2.2.2. LVM  

* LVM  
lvm2 라는 패키지 이름으로 설치가 되어 있고, 자동 파티션을 선택하면 LVM 기반으로 파티션이 할당된다.  
LVM의 구성은 관련 명령어를 이용해서 PV -> VG -> LV 순으로 순차적으로 구성한다  

* LVM 관련 명령어  
1. pvscan  
생성된 물리적 볼륨(PV)의 정보를 출력하는 명령이다.  

사용법  
pvscan [option]  

* -u: PV에 설정된 UUID 값도 함께 출력한다  

2. pvs  
생성된 물리적 볼륨(PV)의 정보를 출력하는 명령으로 설정된 속성값을 확인할 수 있다.  

사용법  
pvs [option]  

[주요속성]  
위치, pv_aatr bit, 설명  
1, d, duplicate  
1, a, allocatable  
1, u, used  
2, x, exported  
3, m, missing  

[주요옵션]  
* -a: PV가 아닌 장치 파일을 포함해서 모두 출력한다  

3. pvdisplay  
생성된 물리적 볼륨(PV)의 크기, PE 크기, PE 할당 여부 등 다양한 정보와 함께 출력한다  

사용법  
pvdisplay  

4. vgscan  
생성된 볼륨 그룹(VG)의 정보를 출력하는 명령이다.  

사용법  
vgscan  

5. pvcreate  
물리적 볼륨(PV)을 생성하는 명령어이다. fdisk를 이용하여 파티션을 분할하고 LVM 속성으로 지정한 뒤에 해당 파티션을 PV로 만들 때 사용하는 명령이다.  

사용법  
pvcreate [option] 디바이스명  

* -f: 강제로 생성할 때 사용하는 옵션이다. 응급상황에서는 -ff를 사용하기도 한다  

6. vgcreate  
볼륨 그룹을 생성하는 명령어로 지정한 PV를 묶어서 하나의 볼륨 그룹으로 생성한다  

사용법  
vgcreate [option] VG명 PV명 [PV명]  

* -s 사이즈: PE의 크기를 지정 

7. vgdisplay  
볼륨 그룹의 속성과 정보를 보여주는 명령어이다.  

사용법  
vgdisplay [option] [볼륨그룹명]  

* -v: 관련 정보를 자세히 출력하는 옵션으로 VG 이외에도 LV와 PV 정보도 같이 출력한다  

8. lvcreate  
볼륨 그룹 안에 LV를 생성하는 명령이다.  

사용법  
lvcreate 옵션 볼륨그룹명  

* -L: LV의 크기를 지정하는 옵션으로 단위를 붙여서 지정한다 
* -l: LV의 크기를 PE 개수로 지정하는 옵션이다.  
* -n: LV의 이름을 지정하는 옵션이다  
* -i: RAID-0과 같이 stripe 사용 시 개수를 지정하는 옵션이다.  
* -l stripe_size: 스트라이핑되는 크기를 지정하는 옵션이다. 기본 단위는 킬로바이트이고 크기는 반드시 2의 배수로 지정한다  
* -s: 유동적인 디렉터리 백업을 위해서 snapshot LV를 생성할 때 사용한다  

9. lvscan  
전체 디스크에 있는 LV를 찾아준다  

사용법  
lvscan [option]  

* -v: 관련 정보를 자세히 출력하는 옵션이다.  

10. lvdisplay  
LV의 속성 정보를 출력하는 명령이다.  

사용법  
lvdisplay [option] [LV명]  

* -v: 관련 정보를 자세히 출력하는 옵션이다.  

11. vgextend  
미리 생성되 VG에 PV를 추가하여 VG를 확장할 때 사용한다  

사용법  
vgextend VG명 PV명 [PV명]  

12. vgreduce  
VG에 있는 PV를 삭제하는 명령이다. 물론 해당 PV의 PE가 할당되어 있으면 제거되지 않는다  

사용법  
vgreduce [option] 볼륨그룹명 PV명  

* -a: 특별한 PV명을 지정하지 않는 경우에 비어있는 PV를 모두 제거한다  

13. lvextend  
LV의 용량을 확장하는 명령  

사용법  
lvextend 옵션 [볼륨그룹명] LV명  

* -L +사이즈: 확장할 LV의 크기를 지정하는 옵션으로 단위를 지정하지 않으면 기본 메가바이트 이다.  
* -l +사이즈: 확장할 LV의 크기를 PE 개수로 지정하는 옵션이다.  

14. lvreduce  
LV의 용량을 줄이는 명령이다. 데이터가 파괴된 경우에는 다시 포맷 해서 사용해야 한다  

사용법  
lvreduce 옵션 [볼륨그룹명] LV명  

* -L -사이즈: 지정한 크기만큼 LV의 옹량을 줄인다. 단위를 지정하지 않으면 기본 메가바이트 이다.  
* -l -사이즈: 지정한 PE 개수에 해당하는 만큼 용량을 줄인다.  

15. lvrename  
LV의 이름을 변경하는 명령이다.  

사용법  
lvrename old_LV_절대경로 new_LV_절대경로  
lvrename VG명 old_LV명 new_LV명  

16. lvremove  
LV를 제거하는 명령으로 사용 중인 경우에는 제거되지 않으므로 미리 언마운트를 해야한다  

사용법  
lvremove 논리적볼륨_절대경로  

17. vgchange  
VG의 속성을 변경하는 명령으로 VG 사용 가능 여부, 최대 LV 개수 등을 변경할 수 있다.  

사용법  
vgchange 옵션 볼륨그룹명  

* -a y|n: 지정한 VG의 사용 여부를 지정한다 
* -l 개수: 해당 VG에 생성할 수 있는 최대 LV수를 지정한다  

18. vgremove  
VG를 제거하는 명령이다. 해당 VG가 사용 중이면 제거가 되지 않으므로 언마운트 후에 속해있는 LV를 전부 제거해야 한다  

사용법  
vgremove VG명  

19. vgmove  
PE를 이동시키는 명령이다  

사용법  
pvmove [option] 장치명 [장치명]  

* -v: 관련 정보를 자세히 출력하는 옵션이다  

20. fsadm  
특정 장치의 파일 시스템의 크기를 조정하고 점검하는 명령이다.  

사용법  
fsadm [option] check 장치명  

* -v: 관련 정보를 자세히 출력하는 옵션이다  
* -f: 몇 가지 검사를 무시하고 넘어갈 때 사용한ㄷ ㅏ
* -y: 명령 실행 시에 나오는 질의에 'yes'라고 답한다  

* LVM 실습 준비사항  
1. 장착된 디스크의 확인 (fdisk -l)  
2. 파티션 생성 (fdisk /dev/sdb)  
3. 파티션 생성 내역 확인 (fdisk의 command 메뉴에서 p 명령을 입력해서 확인)  
4. fdisk 저장 및 종료 (fdisk의 command 메뉴에서 w 명령을 입력해서 저장후 종료)  
5. 시스템 적용 여부를 확인 (cat /proc/partitions 명령을 실행해서 확인)  

* LVM 설정  
문제: 2GB씩 할당된 /dev/sdb1과 /dev/sdb2를 이용해서 lvm0이라는 볼륨 그룹을 생성하고, 3GB는 /ldata1, 나머지 1GB는 /ldata2 로 구성한다  

1. 3GB의 /ldata1 생성하기  

* fdisk /dev/sdb를 실행해서 파티션을 생성하고, 파티션의 속성을 LVM 속성으로 변경  
* 파티션 설정 내역을 확인한 후에 저장 및 종료 
* PV를 생성 (pvcreate /dev/sdb1 /dev/sdb2)
* 생성된 PV를 특정한 볼륨 그룹 (VG)로 구성 (vgcreate lvm0 /dev/sdb1 /dev/sdb2)  
* 생성된 VG를 확인 (vgscan or vgdisplay -v)  
* LV를 생성 (lvcreate -L 3G -n ldata1 lvm0)  
* LV 생성 정보를 확인 (lvscan or lvdisplay) 
* VG의 사용량 정보 확인 (vgdisplay -v)  
* 파일 시스템을 생성 (mkfs.xfs /dev/lvm0/ldata1)  
* 마운트할 디렉터리를 생성 (mkdir /ldata1)  
* 마운트 실행 (mount -t xfs /dev/lvm0/ldata1 /ldata1)  
* mount 및 df 명령을 이용해서 확인  
* 재부팅 시에도 계속 사용하려면 /etc/fstab 파일에 등록 (vi /etc/fstab)  

2. 남은 공간인 1GB를 /ldata2 생성하기  

* 남은 공간 확인하기 (vgdisplay)
* LV 생성 (lvcreate -l 184 -n ldata2 lvm0)  
* LV 생성 정보 확인 (lvscan)  
* 파일 시스템 생성 (mkfs.xfs /dev/lvm0/ldata2)  
* 마운트할 디렉터리 생성 (mkdir /ldata2)
* 마운트 실행 (mount -t xfs /dev/lvm0/ldata2 /ldata2)  
* mount 및 df 명령으로 확인 (mount -> df -h)  
* 재부팅 시에도 계속 사용하려면 /etc/fstab 파일에 등록 (vi /etc/fstab)  

* LVM 확장  
1. 개요 
LV의 확장하는 절차를 간단히 정리하면 다음과 같다.  

* 새롭게 추가된 하드디스크를 PV로 생성한다 
* 생성된 PV를 VG에 포함해서 VG를 확장한다
* LV를 확장한다
* 파일 시스템의 크기를 늘린다
* 증설된 용량을 확인한다

2. 준비 사항  
2-1. fdisk 명령을 이용해서 /dev/sdb에 파티션을 생성하고 속성을 변경  
2-2. 파티션 속성을 확인 후에 저장하고 종료  

3. /ldata1 영역 1GB 증설  
3-1. 새롭게 추가된 하드 디스크를 PV로 생성 (pvcreate /dev/sdb3)  
3-2. VG 확장 (vgextend lvm0 /dev/sdb3)  
3-3. 확인 (vgdisplay -v)  
3-4. 파일 시스템 크기 늘리기 (fsadm resize /dev/lvm0/ldata1)  
3-5. 증설된 용량 확인 (mount -t xfs /dev/lvm0/ldata1 /ldata1, df -h)  

4. /ldata2 영역 2GB 증설  
4-1. VG에서 남아 있는 용량 확인 (vgdisplay -v)  
4-2. LV 확장  

4-2-1. mount가 되어 있다면 언마운트를 실행 (umount /dev/lvm0/ldata2)  
4-2-2. lvextend 명령을 이용해서 확장 (lvextend -l +441 /dev/lvm0/ldata2)  
4-2-3. lvscan 및 lvdisplay 명령을 이용해서 증가된 용량을 확인한다  

4-3. 파일 시스템 크기 늘리기 (fsadm resize /dev/lvm0/ldata2)  
4-4. 증설된 용량 확인 (mount -t xfs /dev/lvm0/ldata2 /ldata2, df -h)

##### 2.2.3 RAID  

* Rocky Linux 8과 RAID 구성  
mdadm 이라는 패키지를 이용해서 Software RAID를 구성한다. 현재 mdadm으로 구성할 수 있는 RAID 레벨에는 1,4,5,6,10이 있다. 장치 파일명은 /dev/mdn 형태로 생성하는데 n은 정수 형태로 0부터 사용한다  

* mdadm  
리눅스 소프트웨어 RAID인 MD를 구성하는 명령  

사용법  
mdadm options \<raiddevice> \<component-devices>  

* -C: RAID 장치를 생성하는 옵션이다 
* -l: -C 옵션과 함께 사용되며 RAID 레벨 값을 지정한다
* -n: -C 옵션과 함께 사용되며 구성할 디바이스의 개수를 지정한다
* -D: 지정한 RAID의 자세한 정보를 보여준다 
* -S: 구성된 RAID의 자세한 정보를 보여준다
* -Q: 지정한 레이드 장치의 간단한 정보를 보여준다
* -f: 특정 장치에 오류를 발생시킬 때 사용한다
* -a: 작동중인 RAID에 디바이스를 추가할 때 사용한다
* -r: RAID에서 디바이스를 제거한다  

* RAID 실습
RAID -0 (461p ~ 463p)  
RAID -1 (464p ~ 466p)  
RAID -5 (466p ~ 467p)  
RAID -6 (467p ~ 468p)  

* RAID 상태 확인  
1. /proc/mdstat  
2. /proc/partitions  

* RAID 결함 허용 테스트 (469p ~ 472)  

* RAID 오류 디스크 교체  
1. 오류 발생 디스크 제거 (mdadm /dev/md2 -r /dev/sdb8)  
2. 새로운 디스크를 레이드 장치에 추가 (mdadm /dev/md2 -a /dev/sdc1) (sdc1은 가정)  
3. 레이드 장치 상태 확인 (mdadm -D /dev/md2)
