### 2.1 네트워크 침해 유형 및 특징  
##### 2.1.1 네트워크 침해 유형 및 특징  

* DOS 공격  
시스테밍나 네트워크의 구조적인 취약점을 공격하여 정상적인 서비스를 할 수 없도록 방해하는 것이다  

[DOS 공격의 분류]  
* 파괴 공격: 디스크, 데이터, 시스템 파괴 
* 시스템 자원 고갈 공격: CPU, 메모리, 디스크의 부하 가중으로 인한 고갈 
* 네트워크 자원 고갈 공격: 불필요한 패킷 유발을 통한 네트워크 대역폭 고갈  

1. Ping of Death  
Ping을 이용하여 ICMP 패킷을 정상적인 크기보다 아주 크게 만들어 보내는 공격방법이다  

2. UDP Flooding  
UDP 패킷을 대량 발생시켜 특정 시스템의 서비스를 방해하는 공격이다.  

3. TCP SYN Flooding  
네트워크 서비스들이 동시에 접속하는 사용자 수의 제한이 있다는 점을 악용한 공격법으로 특히 TCP의 Three-way Handshaking과 밀접한 관계가 있다.  

4. Teardrop Attack  
데이터를 전송하기 위해서는 패킷을 분할하고 시퀀스 번호를 생성하는데, 이러한 시퀀스 번호를 조작하거나 중첩시켜서 패킷화된 데이터를 재조합할 때 혼란을 일으켜 내부에 과부하를 발생시키는 공격이다  

5. Land Attack  
공격자가 임의로 자신의 IP 주소 및 포트를 대상 서버의 IP 주소 및 포트와 동일하게 하여 서버를 공격한다  

6. Smurf Attack  
공격자는 IP 주소를 공격 서버의 IP 주소로 위장하고, ICMP Request 패킷을 브로드캐스트로 보내어 다수의 시스템에 전송한다  

7. Mail Bomb  
다량의 메일을 발송하여 메일 서버의 디스크 공간을 가득 채우는 일종의 스팸 메일 형태의 공격이다  

8. 시스템 자원 고갈 공격  
디스크,메모리,프로세스에 대한 자원 고갈 공격에 대한 예를 만들어 실행한다  

1. vi disk.c  
2. gcc -o disk disk.c  
3. ./disk  

* 용량의 변화 확인 : ls -l/root/disk 또는 df -h  
* 계속 모니터링: watch df -h  
* 메모리 사용량의 변화: free -m  
* 계속 모니터링: watch free -m

* DDos 공격  
여러 대의 공격자를 분산 배치하여 동시에 Dos 공격을 진행함으로써 공격 대상이 되는 시스템이 정상적인 서비스를 할 수 없도록 방해하는 공격이다  

[DDos 도구]  
* Trinoo: 다수의 호스트로부터 통합된 UDP Flooding 공격을 시행  
* TFN: Trinoo와 거의 유사한 공격 도구이지만 공격자 시스템과 마스터 시스템 간 연결이 평문으로 되어 있어 공격자가 노출될 가능성이 높다  
* TFN 2K: 특정 포트를 사용하지 않고, 암호화를 사용한다 
* Stacheldraht: 독일어로 철조망이라는 뜻으로 마스터와 에이전트들이 자동으로 갱신되며 암호화 기능이 추가되며 패스워드 입력을 요구한다  

##### 2.1.2 기타 공격  

* SSH 무작위 대입 공격  
SSH를 기반으로 로그인 시에 무차별적으로 아이디와 패스워드 대입으로 접근을 시도하는 공격이다. 

### 2.2 대비 및 대처 방안  
##### 2.2.1 기본적인 대응책  

* DDos 및 DDos 공격에 대한 대응책  
1. 방화벽을 이용하여 패킷 및 포트 필터링을 통한 진입 차단
2. 침입 탐지 시스템을 이용한 공격 탐지
3. 침입 방지 시스템을 이용한 대처
4. 취약점 및 버그를 이용한 악성 코드 및 침입 방지를 위한 시스템 패치
5. 안정적인 네트워크 설계
6. 다양한 보안 도구를 활용하여 DoS 공격 탐색과 제거 
7. 로드 밸런싱을 통한 대용량 트래픽 분산 및 네트워크 성능 강화
8. 서비스별 대역폭 제한
9. 불필요한 서비스 제거  

* 방화벽  
1. 방화벽의 개요 및 기능 
내부 네트워크를 보호하기 위해 외부에서의 불법적인 트래픽 유입을 막고, 허가되고 인증된 트래픽만을 허용하는 기능  

[방화벽의 기능]  
* 접근 제어: 접근을 허가할 대상과 거부할 대상을 정하는 컷으로 패킷 필터링 방식과 프락시 방식으로 나뉜다  
* 로깅과 감사추적: 접근 허가 및 거부된 내용을 기록하고 추적한다
* 인증: 메시지인증, 사용자인증, 클라이언트인증을 통해 허가된 접근을 인증한다
* 데이터 암호화: 가상 사설망을 이용하여 방화벽에서 다른 방화벽까지 전송되는 데이터를 암호화한다  

2. 방화벽의 종류  

* 배스천 호스트
방화벽 시스템에서 접근 제어, 응용 시스템의 게이터로서 가상 서버의 설치, 인증, 로그 등을 담당한다  

* 스크린 라우터
방화벽 역할을 수행하는 라우터로서 외부 네트워크와 내부 네트워크의 경계에 위치하며 패킷의 헤더 내용을 보고 필터링 할 수 있는 기능을 제공한다  

* 단일 홈 게이트웨이  
스크린 라우터 보다 좀 더 발전된 형태의 방화벽으로 방화벽의 가장 기본적인 기능을 수행한다  

* 이중 홈 게이트웨이  
최소 2개 이상의 네트워크 카드를 이용해서 구성하는 방화벽으로 하나의 네트워크로부터 전송된 패킷이 게이트웨이로 정의된 장치와 통신하게 된다  

* 스크린 호스트 게이트웨이  
스크린 라우터와 단일 홈 게이트웨이를 조합한 형태로 스크린 라우터에서 패킷 필터링을 1차로 수행, 배스천 호스트 역할을 수행하는 단일 홈 게이트웨이에서 2차 방어를 한다  

* 스크린 서브넷 게이트웨이  
외부 네트워크와 내부 네트워크 사이에 완충지대를 두는 방식으로 완충지대의 네트워크를 서브넷이라고 부른다  

* 침입 탐지 시스템(IDS)  
서버나 네트워크에 대한 공격을 실시간으로 탐지하는 시스템으로 데이터 수집, 데이터 필터링과 축약, 침입 탐지, 책임 추적성과 대응 기능을 수행한다  

* 침입 방지 시스템 (IPS)  
네트워크에서 공격 서명을 찾아내어 자동으로 모종의 조치를 함으로써 비정상적인 트래픽을 중단시키는 보안 시스템이다  

* 가상 사설망(VPN)  
인터넷과 같은 공중망을 사설망처럼 사용할 수 있도록 해주는 서비스로 암호화기법을 사용해서 기업 본사와 지사 또는 지사 간에 전용망을 설치할 것 같은 효과를 거둘 수 있다  

* 리눅스 기반 IDS 및 IPS 시스템  
1. Snort  
공개형 IDS 및 IPS 프로그램으로 탐지 룰을 기반으로 네트워크 트래픽을 감시하고 분석하여 침입 여부를 탐지한다  

[Snort의 기능 분류]  
* Packet Sniffer: 네트워크의 패킷을 읽어 보여주는 기능 
* Packet Logger: 모니터링한 패킷을 저장, 로그 기록 및 트래픽 디버깅에 유용  
* Network IDS: 네트워크 트래픽 분석, 공격 탐지 기능 
* Snort Inline: 패킷 분석 및 공격 차단 기능  

2. Suricata  
Snort의 대안으로 개발한 공개형 IDS 및 IPS 프로그램이다. 

##### 2.2.2 iptables  

* iptables의 개요  
패킷 필터링을 직접적으로 수행하지 않고, 커널에 있는 넷필터라는 모듈이 필터링을 수행한다 

[iptables의 테이블과 사슬]  
* INPUT: filter, nat, mangle, security  
* FORWARD: filter, mangle, security  
* OUTPUT: filter, nat, mangle, raw, security  
* PREROUTING: nat, mangle, raw  
* POSTROUTING: nat, mangle  

INPUT: raw 만 없음
FORWARD: 얘만 nat이 없음 
OUTPUT: 모두 있음
PREOUTING: 세개만 있음
POSTROUTING: 가장 적음  

[Filter 테이블의 사슬 및 기능]  
* INPUT: 패킷 필터링 및 방화벽 관련 정책들을 설정하는 사슬이다  
* OUTPUT: 다른 시스템으로 나가는 접근을 차단할 때 사용하는 사슬이다 
* FORWARD: 리눅스 시스템을 통과하는 패킷을 관리하는 사슬이다 


* iptables 의 사용하기  
가장 중요한 것은 실질적 룰에 해당하는 매치와 타깃이다.  

타깃은 iptables에서 패킷이 규칙과 일치할 때 취하는 동작이다  
매치는 iptables가 규칙 타깃에 의해 명시되는 동작에 따라 패킷을 처리하기 위해서 만족해야 하는 조건들이다  

[사용법]  
iptables [-t table] action chain match [-j target]  

[주요 액션]  
* -N: 새로운 사용자 정의 사슬을 만든다
* -X: 비어있는 사슬을 제거한다
* -P: 사슬의 기본 정책을 설정한다
* -L: 현재 사슬의 규칙을 나열한다
* -S: 현재 사슬의 규칙을 적용한 명령 형태로 출력한다
* -F: 사슬로부터 규칙을 제거한다
* -Z: 사슬내의 모든 규칙들의 패킷과 바이트의 카운트를 0으로 만든다
* -A: 사슬에 새로운 규칙을 추가한다
* -I: 사슬에 규칙을 맨 첫 부분에 삽입한다
* -R: 사슬의 규칙을 교환한다
* -D: 사슬의 규칙을 제거한다 

[주요 매치]  
* -s: 출발지 IP 주소나 네트워크와 매칭
* -d: 목적지 IP 주소나 네트워크와 매칭
* -p: 특정 프로토콜과 매칭
* -i: 입력 인터페이스와 매칭
* -o: 출력 인터페이스와 매칭
* !: 특정 매칭을 제외할때 사용
* -m: 좀더 세밀하게 제어할 때 사용하는 옵션
* --state: 연결 상태와 매칭 
* --string: 특정한 패턴과 매칭  

[기타 옵션]  
* -n: 다른 옵션과 같이 사용되며 출력 정보를 숫자값으로 표현
* -v: 다른 옵션과 같이 사용되며 패킷, 바이트 수 등을 추가하여 정보를 자세히 출력  
* --line-number: 정책 앞에 번호를 붙여 출력하며 삭제나 삽입할 때 유용  

[주요 타깃]  
* ACCEPT: 패킷을 허가하는 것으로 본래 라우팅 대로 진행 
* DROP: 패킷을 거부하는 것으로 더 이상 어떤 처리도 수행되지 않고 버림 
* LOG: 패킷을 syslog에 전달하여 기록
* REJECT: 패킷을 버리고 동시에 적당한 응답 패킷을 전달
* RETURN: 호출 사슬 내에서 패킷 처리를 계속 진행 

* iptables의 확장 사용하기  

1. TCP의 확장 
먼저 -p tcp로 프로토콜을 지정하고 추가로 옵션을 사용한다

* --sport: 발신지에서의 하나의 포트 또는 포트 범위를 지정 
* --dport: 도착지의 포트를 지정하는 것
* --tcp-flags: TCP flag를 지정하는 옵션 
* --syn: --tcp-flags SYN, RST, ACK, FIN SYN과 같음  

2. UDP의 확장  
먼저 -p udp로 프로토콜을 지정하고 추가로 옵션을 사용한다  

3. ICMP의 확장  
먼저 -p icmp로 프로토콜을 지정하고 --icmp-type을 추가 옵션으로 지정  
(--icmp-type: 숫자값 형태의 ICMP 타입 값 또는 코드로 지정)  


* iptables를 이용한 설정 예  

1. 들어오는 모든 패킷을 거부하고, 192.168.12.22로부터 들어오는 패킷에 대해서만 허가  
* iptables -P INPUT DROP  
* iptables -A INPUT -s 192.168.12.22 -j ACCEPT  

2. 들어오는 모든 패킷을 허가하고 192.168.1.18로부터 들어오는 모든 패킷에 대해서 거부하기  
* iptables -P INPUT ACCEPT  
* iptables -A INPUT -s 192.168.1.18 -j DROP  

3. 192.168.4.40 에서 들어오는 패킷 중에 tcp 프로토콜 관련 패킷만 거부하기  
* iptables -A INPUT -s 192.168.1.18 -j DROP 

4. 포트 번호 20번 부터 30번까지를 목적지로 들어오는 패킷들은 거부하고, ssh 포트 22번 포트만 허용하기  
* iptables -A INPUT -p tcp --dport 22 -j ACCEPT  
* iptables -A INPUT -p tcp --dport 20:30 -j DROP 

* iptables의 설정 규칙 저장  

1. 관련 명령어 사용  

1-1. iptables-save  
iptables로 설정된 정책을 표준 출력에 출력하는 명령으로 > 기호와 조합해서 현재 설정된 정책을 파일로 저장할 수 있다  

사용법  
iptables-save [option]  

* -t: 특정 테이블에 설정된 정책만 출력한다
* -c: 출력되는 정책 앞부분에 패킷수와 바이트를 표시한다

1-2. iptables-restore  
iptables의 정책 설정을 반영하는 명령으로 < 기호를 이용해서 iptables-save 로 저장된 파일을 불러올 때 사용한다  

사용법  
iptables-restore [option]  

* -c: 기록된 패킷수와 바이트를 그대로 저장한다
* -n: 현재 설정된 정책을 초기화하지 않고 추가하는 형식으로 저장한다

* iptables를 이용한 NAT 구현 

1. NAT의 개요  
네트워크 주소를 변환하여 주는 역할을 한다. 공인 IP 주소를 공유하여 여러 대의 컴퓨터가 인터넷 접속이 가능하게 하는 설정이 대표적이다  

2. NAT의 사용예  

* 하나의 공인 IP 주소를 공유하여 다수의 PC에서 인터넷을 사용하는 경우  
* 하나의 IP 주소를 이용하여 다수의 서버를 연결하여 사용하는 겨웅  
* 투명 프락시  

3. iptabls에서 NAT의 분류 및 설정  

* SNAT  
패킷의 소스 주소인 발신지를 변경한다는 의미이며 공인 IP 주소 하나로 다수의 컴퓨터가 인터넷 접속이 가능하게 하는 경우이다  

* DNAT
패킷의 도착지 주소를 변경한다는 의미이며 하나의 공인 IP 주소로 다수의 서버를 운영하는 경우가 해당한다 

* iptables를 이용한 공격 대비 설정 예  

1. ICMP Flooding 대비  

* iptables -N ICMP 
ICMP 라는 이름의 사슬을 생성한다  

* iptables -A INPUT -p icmp -j ICMP  
ICMP 프로토콜 관련 패킷은 ICMP 사슬로 보낸다  

* iptables -A ICMP -p icmp --icmp-type echo-request -j DROP  
icmp echo requset 요청은 DROP 시킨다  

2. UDP Flooding 대비  

* iptables -N UDP  
UDP 라는 이름의 사슬을 생성  

* iptables -A INPUT -p udp -j UDP  
UDP 프로토콜 관련 패킷은 UDP 사슬로 보낸다  

* iptables -A UDP -p udp --dport 80 -m recent --update --seconds 1 --hitcount 10 -j DROP  
같은 IP 주소에서 1초에 10개 이상 UDP 패킷이 들어오면 DROP 시키는 정책을 UDP 사슬에 추가한다  

* iptables -A UDP -j LOG --log-prefix "UDP FLOOD"  
UDP 사슬 로그에 "UDP FLOOD"라고 기록한다  

3. TCP Flooding 대비  

* iptables -A INPUT -p tcp --dport 80 --syn -m limit --limit 100/s -j ACCEPT  
초 당 허용되는 접속 개수 제한  

* iptables -A INPUT -p tcp --dport 80 --syn -m connlimit --limit -above 30 -j DROP  
IP 주소당 허용되는 접속 개수 제한  

* iptables -A INPUT -p tcp --dport 80 -m recent --update --seconds 1 --hitcount 10 -j DROP  
특정 IP 주소에서 오는 초 당 요구 수 제한  

4. SSH 무작위 대입 공격 대비  

* iptables -N SSH  
SSH 라는 이름의 사슬을 생성  

* iptables -A INPUT -p tcp --dport 22 -m state --state NEW -j DROP  
22번 포트로 접속하는 패킷은 SSH 사슬로 보낸다  

* iptables -A SSH -p udp --dport 22 -m recent --update --seconds 60 --hitcount 15 -j DROP  
같은 IP 주소에서 60초동안 15번 이상 SSH 접속을 시도하면 DROP 시키는 정책을 SSH 사슬에 추가  

* iptables -A SSH -j LOG --log-prefix "SSH Brute"  
SSH 사슬 로그에 "SSH Brute"라고 기록한다  

[기타 도구]  
* fail2ban: ssh 뿐 아니라 다양한 서비스에 대해 공격을 차단할 수 있는 도구 
* sshguard: 방화벽 프로그램을 이용하여 SSH 무작위 대입 공격을 차단하는 ㄷ구 
* DenyHosts: SSH 관련 로그 파일인 /var/log/secure 를 검사하여 /etc/hosts.deny에 추가해서 차단하도록 하는 파이썬 프로그램  

##### 2.2.3 firewalld  

* firewalld의 개요  
D-Bus 인터페이스가 있는 호스트 기반 방화벽을 제공하는 서비스 데몬이다 

* iptables 기반 도구와 firewalld의 비교  

1. 정적 방화벽과 동적 방화벽
정적 방화벽: 정책을 변경한 경우에 다시 시작해야함 
동적 방화벽: 방화벽 정책을 실시간 환경에서 즉시 변경할 수 있다  

2. 서비스, 응용 프로그램, 사용자의 쉬운 설정 

3. 다양한 방화벽 구성 도구 기본 제공  

4. 런타임 및 영구 설정 분할 지원  

##### 2.2.4 nftables  

* nftables의 개요  
네트워크 대역별로 사용중인 가상 머신에 대해 새로운 커널 내 패킷 분류 프레임워크를 제공한다  

* nftables의 주요 개선 사항  
선형 처리 대신에 내장된 조회 테이블 제공, IPv4, IPv6 프로토콜을 하나의 단일 프레임워크로 지원한다  

* nftables 사용하기  

1. nft list ruleset  
설정된 방화벽 규칙 정보를 출력한다  

2. nft flush ruleset  
설정된 방화벽 규칙 정보를 추기화시켜서 모두 제거한다  

* nftables 스크립트 작성  
/etc/nftables/ 디렉터리에 *.nft 파일이 존재하는데 이 스크립트 파일의 내용을 확인해보면 기본적인 형식을 확인할 수 있다  

* nftables 스크립트의 실행  

1. nft 명령어를 사용한 실행  
* nft -f /etc/nftables/firewall.nft  

2. 직접 실행  
* chmod a+x /etc/nftables/firewall.nft  
* /etc/nftables/firewall.nft
