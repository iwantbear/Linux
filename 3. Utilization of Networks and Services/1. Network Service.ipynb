### 1.1 웹 관련 서비스  
##### 1.1.1 웹 관련 서비스의 이해  

* 웹 작동 원리  
웹 브라우저를 사용하는 클라이언트와 웹 서버 사이에 HTTP 프로토콜을 사용하기 때문에 웹 브라우저를 HTTP 클라이언트, 웹 서버를 HTTP 서버라 부른다  

1. 주요 HTTP 요청 메소드  

* GET: 서버에서 자료를 가져오는 요청  
* POST: 서버에 정보를 저장하는 요청 
* HEAD: GET과 유사하나 HTTP 헤더 정보만 가져옴 
* PUT: 서버에 자료를 올릴 때 사용 
* DELETE: 서버의 자료를 삭제할 떄 사용  

telnet 명령을 이용하여 운영중인 서버의 80번 포트로 접속한 뒤에 'GET / HTTP/1.1' 이라고 입력하면 해당 웹 서버의 정보를 알 수 있다  

2. HTTP 응답  
HTTP 서버는 클라이언트의 요청에 대한 응답으로 HTTP 프로토콜 버전, 상태 코드, 데이터 등을 전송한다.  

* REST  

1. 개요  
분산 하이퍼미디어 시스템을 위한 소프트웨어 아키텍처의 한 형식이다  

2. 주요 특징  

2-1. 클라이언트-서버 구조  
2-2. 무상태성 연결  
2-3. 캐시 기능  
2-4. 균일한 인터페이스  
2-5. 계층화된 시스템  
2-6. 주문형 코드

2. extra 디렉터리 안의 파일  
2-1. 개요  
/usr/local/apache/conf/extra 디렉터리 안에 존재하는 파일은 기본 적으로 사용되지 않도록 설정 되어 있는데 사용하기 위해서는 두가지를 확인해야 한다  

* httpd.conf 파일의 하단에 위치한 Include 항목의 주석을 제거 
* 각 파일에서 요구하는 아파치 모듈을 찾아서 LoadModule 항목의 주석을 제거  

2-2. extra 디렉터리 안의 파일과 역할  

* httpd-mpm.conf: MPM관련 설정을 하는 파일  
* httpd-multilang-errordoc.conf: 에러 처리 문서에 대한 설정을 하는 파일 
* httpd-autoindex.conf: 디렉터리 목록을 표시할 때 여러 가지 설정을 하는 파일  
* httpd-userdir.conf: 개인 사용자의 홈 디렉터리 설정 시 사용하는 파일  
* httpd-info.conf: 웹 브라우저를 이용하여 웹 서버의 상태 정보를 파악할 때 사용하는 파일 
* httpd-vhosts.conf: 버추얼 호스트 사용할 때 설정하는 파일 
* httpd-dav.conf: WebDAV 관련 설정 파일 
* httpd-default.conf: Timeout 등 서버 운영의 기본적인 설정하는 파일 
* httpd-languages.conf: 문서의 언어 표시 관련 설정을 하는 파일 
* httpd-manual.conf: 운영 중인 서버 상에서 아파치의 매뉴얼 문서를 볼 수 있게 설정하는 파일  
* httpd-ssl.conf: SSL 관련 설정을 하는 파일 
* proxy-html.conf: HTML4 및 XHTML 1 을 이해하기 위해 mod_proxy_html 을 설정하는 파일  

2-3. httpd-userdir.conf 파일 활성화 예  

* httpd.conf 파일에서 LoadModule, Include 항목 활성화 (주석 없애기)  
* httpd-userdir.conf 파일 설정 (vi /usr/local/apache/conf/extra/httpd-userdir.conf)  
* 아파치 웹 서버 대몬 재시작 (/usr/local/apache/vin/apachectl restart)  


[httpd.conf 옵션 항목]  
* NONE: 어떤 옵션도 이용할 수 없으므로 모든 접근을 거부 
* ALL: MultiViews 옵션을 제외한 모든 옵션을 부여 
* Includes: 서버 측의 추가적인 정보를 제공 
* Indexes: URL이 지정된 디렉터리에 index.html과 같은 지정된 파일이 없을 경우 해당 디렉터리의 파일목록을 보여주는 옵션  
* FollowSymLinks: 디렉터리의 심볼릭 링크 사용을 허용 
* ExecCGI: CGI 스크립트를 실행할 수 있도록 허용 
* IncludesNOEXEC: SSI는 허용하지만 #exec 명령과 CGI 스크립트 #include 불허  
* MultiViews: ALL 옵션이 설정되었을 때만 지정된 목록의 multiviews를 허용  

[AllowOverride 항목]  
* NONE: AllowOverride를 사용하지 않음. 즉, 사용자 인증 파일을 사용하지 않음 
* ALL: httpd.conf 파일의 AccessFileName 지시자로 설정한 파일을 사용하고 사용할 수 있음 
* Authconfig: AccessFileName 지시자에 명시한 파일에 대해서 사용자 인증 지시자 사용을 허락  
* FileInfo: AccessFileName 지시자로 설정한 파일에 대해서 문서 유형을 제어하는 지시자 사용을 허락  
* Indexes: AccessFilename 지시자로 설정한 파일에 대해서 디렉터리 Indexing을 제어하는 지시자 사용을 허락  
* Limit: AccessFileName 지시자로 설정한 파일에 대해서 호스트 접근을 제어하는 지시자 사용을 허락 
* Options: AccessFileName 지시자로 명시한 파이에 대해서 Options, XBiHack 등과 같은 지시자 사용을 허락

1. httpd 명령을 이용한 실행  
아파치 웹 데몬을 실행하는 명령이다  

사용법  
httpd [option] [filename]  

* -t: 환경 설정 파일인 httpd.conf의 문법적 오류를 검사한다 
* -f: 새로 설정한 conf 파일이 있을 경우에 그 설정 파일로 아파치 데몬을 구동할 때 사용한다 
* -S: 설정된 가상호스트 및 주요 설정 정보를 출력한다
* -I: 정적 컴파일된 모듈에 대한 목록을 출력한다
* -k: httpd 데몬에 시그널을 보내는 옵션이다  
* -M: 정적 컴파일 모듈과 공유 모듈을 전부 출력한다 
* -V: httpd에 빌드된 파라미터와 버전 정보를 출력한다
* -D: 환경설정과 관련된 파라미터를 지정해서 해당 정보만을 확인할 수 있다.  

2. 스크립트 파일을 이용한 실행  
아파치 웹 서버를 소스 설치하면 apachectl 이라는 데몬 실행 제어 스크립트를 이용하여 실행 및 제어를 할 수 있다.  

사용법  
apachectl command  

* help: 간단한 도움말을 출력한다
* start: 아파치 데몬을 시작한다
* stop: 아파치 데몬을 중지시킨다
* restart: 아파치 데몬을 재시작한다
* reload: 아파치 데몬을 재시작하지 않고, 환경 설정 파일만 다시 읽어들인다 
* configtest: httpd.conf 파일의 문법적 오류를 검사한다
* graceful: 웹 서버에 연결된 접속을 차단하지 않고, httpd.conf의 설정을 다시 읽어들인다  
* graceful-stop: 웹 서버에 연결된 접속은 차단하지 않고, httpd 데몬을 중지시킨다
* status, fullstatus: 웹 서버의 상태를 간단히 또는 전체적으로 출력한다

3. 부팅 시 자동 실행되게 설정하기

5. 웹 브라우저로 확인하기  

* 웹 서버 접근 통제  
아파치 웹 서버의 기본적인 설정은 모든 클라이언트의 접근이 허락되어 있다. 그러나 접근 통제가 필요한 경우에는 클라이언트의 IP 주소 또는 도메인에 대해 Order, Allow, Deny 지시자를 이용해서 제어할 수 있다.  

사용법  
1. Order Deny,Allow  
Deny from 주소  
Allow from 주소  
Allow 지시자가 지정한 주소만 접속을 허가할 때 사용한다  

2. Order Allow, Deny  
Allow from 주소  
Deny from 주소  
Deny 지시자에 지정한 사용자만 접속을 거부할 때 사용한다  

3. Order Mutual-failure  
Allow from 주소  
Deny from 주소  
Allow 지시자에 있고, Deny 지시자에는 없는 호스트만 접근을 허용한다  

[허용되는 주소 표기]  
* All, all: 모든 클라이언트를 의미한다 
* 도메인명: 특정 호스트명이 포함된 도메인명 또는 해당 도메인 전체를 지정할 수 있다.  
* IP 주소: 특정 호스트의 IP 주소를 사용할 수 있다.  
* 네트워크 대역: 특정 네트워크 대역에 속한 전체 호스트를 지정할 수 있다.

1. smb.conf 파일 내부의 유형  

* #, ;으로 시작하는 행: 주석으로 간주하여 무시된다.  
* []을 사용하는 행: 섹션을 정의하는 역할을 하고, 하나의 섹션이 정의된 후 그 다음 섹션이 정의될 때까지의 행들은 해당 섹션에 속한다.  
* name = value 행: 사용하는 옵션과 해당 값을 설정하는 행이다.  

2. smb.conf 의 주요 섹션  

* \[global]: 삼바 서버의 전체적인 환경 설정을 담당하는 섹션  
* \[homes]: 각 사용자들이 자신의 홈 디렉터리로 접근할 때의 권한을 설정하는 섹션  
* \[printers]: 프린터 관련 권한을 설정하는 섹션
* \[print$]: 프린터 드라이버 관련 섹션

3. [global] 섹션의 주요 항목  

* workgroup = SAMBA  
윈도우의 작업 그룹에 해당하는 항목으로 공유 그룹명을 지정한다  

* server string = Samba Server  
서버에 대한 설명을 설정한다  

* netbios name = MYSERVER  
윈도우에서 이름으로 접속할 때 관련 이름을 지정한다  

* interfaces = lo eth0 192.168.12.2/24 192.168.13.2/24  
네트워크 인터페이스별로 접근 제어할 때 설정한다  

* hosts allow = 127. 12.168.12. 192.168.13  
삼바 서버에 접속을 허용할 호스트를 지정할 때 사용한다  

* log file = /var/log/samba/log.%m  
삼바 서버에 접속하는 호스트의 로그를 기록하는 파일을 지정한다  

* max log size = 50  
로그 파일의 최대 크기를 설정하는 부분이다  

* security = user  
삼바 서버에 접속할 때 인증 모드를 설정하는 항목이다  

user: 삼바 서버에 접속할 때 사용자명과 패스워드를 입력해야 한다  
domain 또는 ads: 윈도우의 도메인 컨트롤러에 전달하여 유효한 사용자 여부를 확인할 때 사용  

* passdb backend = tdbsam  
security가 user일 때 사용자의 패스워드를 설정하는 방식  

4. [share] 섹션의 주요 항목  

* comment: 간단한 설명을 설정  
* path: 공유 디렉터리의 경로를 설정  
* read only: 공유 디렉터리를 읽기 전용으로 설정  
* writable, write ok: 공유 디렉터리를 쓰기 가능하게 설정  
* valid users: 공유 디렉터리를 이용할 수 있는 사용자를 지정 
* write list: 공유 디렉터리에 접근 및 쓰기 권한을 행사할 수 있는 사용자를 지정
* public, guest ok: 다른 사용자들이 이용할 수 있도록 설정할 때 사용한다 
* browseable, browsable: 공유 디렉터리의 리스트를 보여줄 때 설정한다
* printable: 공유 디렉터리에 스풀 파일을 지정할 때 설정한다  
* create mask, create mode: 파일을 생성할 때 허가권을 지정할 때 설정한다

##### 1.3.2 NFS 서버 관리  

* NFS 개요  
TCP/IP 네트워크 상에서 다른 컴퓨터의 파일 시스템을 마운트하고 공유하여 상대방의 파일 시스템 일부를 마치 자기 자신의 디렉터리 인것 처럼 사용할 수 있게 해준다 

* NFS 관련 주요 RPM 패키지  

1. rpcbind: RPC 기반 연결을 위해 필요한 패키지로 rpcbind,rpcinfo 등을 포함하고 있다  
2. nfs-utils: NFS 서버 관련 패키지로 관련 데몬 및 명령어를 포함하고 있다 

* NFS 서버 설정의 개요  

1. 접근 관련 설정 
* vi /etc/exports  

2. NFS 관련 데몬 시작  
* systemctl start rpcbind  
* systemctl start nfs-server  

3. 서버 동작과 관련 메시지 확인 
* systemcl -l status nfs-server  

4. 관련 서버 동작의 확인  
* ps aux |egrep 'rpcbind | nfs-server  

5. NFS 서버의 재시작  
* systemctl restart nfs-server

6. 부팅 시 nfs 관련 서버 활성화

* 환경 설정 파일: /etc/exports  

[설정법]  
공유할_디렉터리 허가할_호스트(option) [허가할_호스트(option)]  

* ro: 읽기 전용으로 지정하는 옵션
* rw: 읽기 및 쓰기가 가능하도록 지정
* root_squash: NFS 클라이언트에서 접근하는 root 사용자를 무시하고 서버상의 nobody(또는 nfsnobody) 로 매핑시키는 옵션  
* no_root_squash: NFS 클라이언트에서 접근하는 root 사용자를 root 로 인정한다
* no_subtree_check: 하위 디렉터리를 검사하지 못하도록 할 때 설정한다
* all_squash: 일반 사용자의 권한을 nobody(또는 nfsnobody)로 매핑시킨다
* secure: 포트 번호가 1024 이하의 요청에만 허가할 때 사용한다
* sync: 변경 사항이 안정적으로 저장된 경우에만 관련 요청에 응답하도록 한다. 
* async: 쓰기가 설정된 디스크 스토리지에 사용하면 유용하며 데이터 변경에 대비해 비동기적 처리를 할 때 사용한다
* anonuid: 접근하는 사용자 권한을 지정하는 uid로 매핑시킬 때 사용한다
* anongid: 접근하는 그룹 권한을 지정하는 gid로 매핑시킬 때 사용한다

* NFS 클라이언트에서 사용법  

1. mount 및 mount.nfs 명령 이용하기

* timeo: RPC 타임아웃이 발생된 후에 첫 번째 재전송 요구를 보낼 떄의 시간으로 단위는 1/10초이다 
* retrans: 타임아웃이 발생된 후에 재전송 요구의 횟수를 지정한다
* soft: NFS 서버 요청에 실패하면 retrans에 설정한 횟수반큼 재요청을 시도한다
* hard: NFS 서버 요청에 실패하면 무한정 재시도한다. 
* rsize: NFS 서버로부터 읽어 들이는 바이트 값을 지정한다

##### 1.3.3 FTP 서버 관리  

* FTP의 개요  
TCP/IP 프로토콜 기반으로 서버와 클라이언트 사이에 파일을 주고받기 위한 프로토콜이다  

* vsftpd 서버 설정  

1. vsftpd 관련 주요 파일  

* /etc/vsftpd/vsftpd.conf: vsftpd의 환경 설정 파일  
* /etc/pam.d/vsftpd: vsftpd의 PAM 관련 설정 파일 
* /etc/vsftpd/ffpusers: PAM 관련 설정 파일인 /etc/pam.d/vsftpd에 사용되는 파일로 기본 설정이 접근 거부될 사용자 목록으로 이용된다  
* /etc/vsftpd/user_list: vsftpd에서 이용하는 사용자 목록 파일로 허가 또는 거부 목록 파일로 사용할 수 있으나, 기본 설정이 userlist_deny=YES이어서 거부 목록 파일로 사용된다  
* /etc/logrotate.d/vsftpd: vsftpd의 로그 로테이트 관련 파일이다  

2. vsftpd의 파일 분석  

* anonymous_enablde=NO  
익명 계정의 허가 유무를 지정  

* local_enable=YES

### 1.5 DNS 관리  
##### 1.5.1 DNS의 이해  

* DNS 서버의 종류  

1. Primary Name Server: 사용하는 도메인을 관리하기 위해 필수적으로 구성하는 서버이다   

2. Secondary Name Server: Primary Name Server의 zone 파일을 백업하는 역할을 수행한다  
3. Caching Name Server: 관리하는 도메인 없이 리졸빙만을 제공하기 위해 구성하는 서버이다  

* DNS 서버 프로그램: BIND  
가장 널리 쓰이는 DNS 서버 프로그램으로 DHCP 서버 프로그램과 함께 공개 소프트웨어로 배포되고 있다  

[bind 관련 파일 및 디렉터리]  
* /etc/named.conf: DNS 서버의 전반적인 환경설정을 담당 파일로 서버에서 사용하는 zone 파일을 지정한다  

* /var/named: 루트 도메인 서버에 대한 정보를 담고 있는 named.ca를 비롯하여 사용자가 선언하는 zone 파일 등이 위치하는 디렉터리이다 

##### 1.5.2 DNS 관리 및 활용  

* /etc/named.conf 파일  

1. 파일의 구성과 특징  

* 파일의 구성은 크게 주석문과 구문으로 구성되어 있다 
* 주석은 C에서 사용하는 /* */, C++에서 사용하는 //, 유닉스에서 사용하는 # 등 모두 사용가능하다  
* 구문에는 options, logging, controls, zone, acl, serve, key 등이 존재한다  
* 각 구문은 중괄호로 둘러싸고 끝날 때는 세미콜론(;)을 사용한다
* include 지시자를 선언하여 별도의 파일에 추가 정의할 수 있다  

2. 주요 구문  
[options 구문]  
DNS 서버의 동작 및 제어와 관련된 여러 가지 설정을 하는 영역으로 존 파일이 위치하는 디렉터리명은 반드시 명기해야 한다 

* listen-on port 53 { 127.0.0.1; };  
DNS 서버의 포트 번호인 53번 포트에 접근할 수 있는 호스트를 지정하는 항목  

* liten-on-v6 port 53 { ;;1; };  
DNS 서버 포트 번호인 53번 포트에 접근할 수 있는 호스트를 지정하는 항목으로 IPv6 주소 기반 설정시에 사용한다  

* directory "/var/named";  
존 파일이 위치하는 디렉터리명을 명기한다  

* dump-file "/var/named/data/cache_dump.db";  
named는 정보가 갱신될 떄 dump 파일로 저장하는데, 그 덤프 파일이 생성될 위치와 파일명을 지정한다  

* statistics-file "/var/named/data/named_stats.txt"  
네임 서버의 통계 파일을 생성할 위치와 파일명을 지정한다  

* memstatistics-file "/var/named/data/named_mem_stats.xtx";  
네임 서버의 메모리 통계 파일을 생성할 위치와 파일명을 지정한다  

* forward (only|first);  
forwarders 구문과 함계 사용되며, only나 first 두 값중 하나를 갖는다.  

only: 자신에게 들어온 도메인 질의를 지정한 다른 서버로 넘기도록 하는 것으로 다른 서버가 그에 대한 응답이 없을 경우 그 자신도 그 질의에 대해 응답하지 않을 경우에 설정한다  

first: 타 서버에서 응답이 없을 떄 자신이 응답하도록 할 때 설정한다  

* forwards {네임서버주소1; 네임서버주소2;...}  
도메인에 대한 질의를 다른 서버로 넘길 때 사용하는 구문이다  

* allow-query { 192.168.0/24 }; 
네임 서버에 질의할 수 있는 호스트를 지정한다  

* allow-transfer { 192.168.0/24 };  
zone 파일의 내용을 복사할 대상에 제한을 걸 때 지정한다  

* datasize 512M;  
DNS 관련 정보를 캐싱할 때 사용하는 메모리를 제한하는 설정이다  

* recursion yes;  
질의 중 인터넷을 통해 다른 DNS 서버에 질의를 보내고, 질의를 다시 받을 수 있다.

* any: 모든 호스트를 나타낸다 
* none: 어떠한 호스트도 거부한다

* localnet: 시스템에 부여된 IPv4 및 IPv6 네트워크의 모든 호스트를 매칭시킨다

* zone 파일  
1. zone 파일의 개요  

zone 파일은 /etc/named.conf 파일에 지정된 디렉터리 안에 지정한 파일명으로 생성해야 한다  
기본 설정 디렉터리는 /var/named  
최근 zone 파일명은 /etc/named.frc1912.zones 이다  

2. zone 파일의 구조  
크게 SOA 레코드와 자원 레코드로 나눌 수 있다.  

3. zone 파일 항목 설명  

* $TTL  
zone 파일의 첫 줄에 설정해야 한다. 역할은 다른 서버에서 도메인 관련 정보를 조회한 경우 조회한 서버의 캐시에 해당 정보가 얼마나 머물지를 결정한다  

* SOA record  
존 파일을 시작하는 레코드로 시작은 도메인명 또는 도메인명을 나타내는 기호인 '@' 를 사용한다 항목에 대한 설명이 필요한 경우 주석 기호인 세미콜론을 사용하면 된다  

[주요구성]  
* nameserver: 네임 서버의 호스트명과 도메인 명을 적는다 
* contact_email_address: 관리자의 이메일 주소를 적는다 
* serial_number: 일종의 일련 번호로 zone 파일이 갱신되면 숫자가 더 크도록 설정한다
* refresh_number: 보조 네임 서버가 자신의 정보를 업데이트하기 위해 주 네임 서버에 얼마나 자주 체크할 것인지를 설정한다  
* retry_number: 보조 네임 서버가 주 네임 서버 접근에 실패했을 경우에 재시도할 주기를 설정한다  
* expire_number: 보조 네임 서버가 주 네임 서버에 접근을 실패하여 retry_number에 설정한 주기로 재시도할 때의 만기 시간을 지정한다 
* minimum_number: 일종의 SOA 영역의 TTL로써 가져간 정보의 최소 보관 시간을 설정한다  

* 도메인 
도메인명, 호스트명, 공백, @, * 등이 올 수있다. @는 현재 도메인, *는 모든 호스트명을 뜻한다  

* TTL 
해당 레코드에 대한 TTL 값을 설정하는 항목이다  

* class  
레코드의 클래스를 지저하는 부분으로 IN, CH, HS 등을 사용할 수 있다.  

* type  
해당 레코드의 타입을 지정하는 영역이다 

[주요 타입]  
* A: 도메인 주소를 IPv4 형식의 주소로 매핑할 때 사용한다 
* AAAA: 도메인 주소를 IPv6 형식의 주소로 매핑할 때 사용한다
* NS: 도메인의 네임 서버를 지정할 떄 사용한다
* MX: 메일 서버를 지정할 때 사용한다. 0또는 양의 정수값을 기입, 우선순위를 정한다
* CNAME: 별칭을 지정할 때 사용한다
* PTR: 리버스 존에서만 사용하는 레코드로 IP 주소를 도메인으로 변환하기 위해 지정한다
* TXT: 도메인에 대한 텍스트 정보를 기재할 때 사용한다  

* Rdata  
IP주소, 도메인명, 호스트명을 지정하는 영역이다

##### 1.7.2 프락시 관리  

* Proxy 개요  
1. Proxy 정의  
프락시 서버란 보통 네트워크 속도가 느린 환경에서 보다 빠른 인터넷을 이용하기 위해 사용하며 자주 방문하는 사이트의 정보를 저장하는 일종의 캐시 서버로서 처리 속도를 높인다  

2. Proxy의 서버 구성과 이용  
squid를 주로 사용한다  

* Squid  
1. Squid의 개요  
Proxy 서버 프로그램인 squid를 설치하면 환경 설정 파일은 /etc/squid/squid.conf 이다  

2. squid.conf 파일의 주요 설정  
* http_port 3128  
squid 프락시 서버의 포트 번호를 지정하는 항목이다  

* cache_dir ufs /var/spool/squid 100 16 256  
캐시 정보가 저장될 경로를 지정하는 항목으로 ufs는 squid의 저장 포맷이고, /var/spool/squid가 관련 디렉터리이다. 100은 캐시 정보의 크기, 16은 캐시가 저장되는 첫 번째 하위 디렉터리의 갯수이며, 256은 두 번째 하위 디렉터리의 개수이다  

3. 프락시 서버의 접근 제어  
squid.conf 파일에서 acl 항목을 이용해서 접근 제어를 할 수 있다. 

사용법  

1. acl 설정  
* acl 별칭 src IP주소/넷마스크값
* acl 별칭 dst IP주소/넷마스크값
* acl 별칭 srcdomain .foo.com
* acl 별칭 dstdomain .foo.com  

2. acl에 대한 접근 제어  
* http_access allow 별칭  
* http_access deny 별칭  

##### 1.7.3 DHCP 관리  

* DHCP 서버의 개요  
클라이언트에게 자동으로 IP 주소, Gateway 주소, NameServer 주소 등을 할당하는 서버를 말한다  

[관련 프로토콜]  
* BOOTP: 초기 적재 통신 규약, 하드 디스크가 없는 장치에 IP 주소 등을 할당하고 관리하기 위해 개발된 프로토콜  
* DHCP: 동적 호스트 설정 통신 규약  
--------------------------------------------------------------------  
* DHCP 서버의 설치 및 운영  

1. DHCP 관련 주요 패키지  
* dhcp-server: DHCP 서버 관련 패키지로 데몬 파일인 dhcpd, 환경 설정 파일인 dhcp.conf 등을포함  

* dhcp-common: DHCP 서버와 클라이언트 모두에게 필요한 파일을 포함하는 패키지  
* dhcp-libs: DHCP 서버 및 클라이언트에 의해 사용되는 라이브러리 모음
* dhcp-client: DHCP 클라이언트 관련 환경 설정 파일 및 명령어를 포함하고 있는 패키지  

2. DHCP 서버 패키지 설치  
* dnf install dhcp-server  

3. DHCP 서버 패키지 주요 파일  
* /etc/dhcp/dhcpd.conf
환경 설정 파일로 'rpm -ql dhcp-server' 명령으로 dhcpd.conf.example을 찾아서 복사하거나 이름을 바꾸어서 생성  

* /usr/share/doc/dhcp-server/dhcpd.conf.example  
dhcpd.conf 파일의 실질적인 설정 내용이 들어있는 일종의 파일 샘플  

4. DHCP 서버의 시작  
* systemctl start dhcpd  


* /etc/dhcp/dhcpd.conf 의 주요 설정  

1. max-lease-time 7200;  
DHCP 서버를 사용하는 모든 네트워크 설정되는 도메인명과 도메인 네임 서버, 임대 요청 관련 시간을 설정한다  

2. log-facility local7;  
syslog에 전달한 로그의 facility를 지정한다  

[주요 설정]  
* range: 클라이언트에게 할당할 IP 주소를 시작 주소와 마지막 주소를 기재  
* range dynamic-bootp: DHCP 클라이언트, BOOTP 클라이언트 함께 할당할 경우에 사용  
* option domain-name: 도메인명을 지정할 때 사용 
* option domain-name-servers: 네임 서버를 지정할 때 사용하는 항목으로 도메인 또는 IP 주소를 기재  
* option routers: 게이트웨이 주소를 지정할 때 사용 
* option broadcast-address: 브로드캐스트 주소를 지정할 때 사용
* default-lease-time: 임대를 요청하는 클라이언트가 특정한 만료 시간을 요청하지 않았을 경우에 할당되는 시간으로 단위는 초이다 
* max-lease-time: 임대를 요청한 클라이언트가 IP 주소를 가지고 있을 수 있는 최대 시간이다  

##### 1.7.4 VNC 관리  

* VNC 의 정의 및 개요  
RFB 프로토콜을 이용하여 원격의 다른 컴퓨터에서 그래픽 환경 기반으로 테스크톱을 공유할 수 있는 시스템이다  

[VNC 구성 요소 및 패키지]  

* tigervnc-server: TigerVNC 서버 관련 패키지로 환경 설정 파일, 명령어 등을 포함하고 있다  
* tigervnc-selinux: TigerVNC 서버 관련된 SELinux 정책 모듈을 포함하고 있는 패키지이다  
* tigervnc: TigerVNC 클라이언트 패키지로 원격지의 VNC 서버에 접근할 때 사용하는 vncviewer 명령어를 포함하고 있다.  
--------------------------------------------------------------------  

* VNC 서버 설치 및 설정  

1. VNC 서버 설치  
* dnf install tigervnc-server  

2. VNC 서버 접근 설정 
VNC 서버를 사용하기 위해서는 SELinux 설정은 disable로, X 서버도 Wayland 대신에 Xorg를 사용해야 한다  

2-1. SELinux 설정 변경  
* vi /etc/sysconfig/selinux  

2-2. X 서버를 Xorg로 변경  
* vi /etc/gdm/custom.conf 명령을 실행해서 WaylandEnable=false로 변경  

3. VNC 서버 설정  
3-1. vnc 서버 패스워드 설정  

사용법  
vncpasswd  

3-2. vncserver 기본 설정 확인 및 병경  
* /etc/tigervnc/vncserver-config-defaults 파일에서 데스크톱 및 해상도 설정  

3-3. 사용자들 특정 포트에 매핑하도록 설정  
* /etc/tigervnc/vncserver.users 파일에 ':디스플레이_포트=사용자' 형식으로 지정  

3-4. VNC 동작을 위한 서비스 파일 생성  
* /lib/systemd/vncserver@.service 파일을 /etc/systemd/system/vncserver@:1.service 파일로 복사한다  

3-5. VNC 서버 관련 데몬 등록  
* systemctl daemon-reload  

3-6. VNC 서버 실행  
* systemctl start vncserver@:1  

3-7. VNC 서버 동작 및 메시지 확인  
* systemctl status vncserver@:1  

3-8. 부팅 시 VNC 서버 구동  
* systemctl dnable vncserver@:1  

3-9. 서비스 포트 확인  
* nerstat -tlnp


##### 1.7.5 NTP 관리  

* NTP의 개요  
컴퓨터간의 시간을 동기화하는데 사용되는 프로토콜로 여러 계층으로 구성된다  

* NTP 서버 프로그램  
chrony 패키지 설치를 통한 chronyd를 사용하고 환경 설정 파일은 /etc/chrony.conf 이다  

* /etc/ntp.conf의 주요 설정  

1. pool 2.pool.ntp.org iburst  
2.pool.ntp.org 사이트에 등록된 일종의 공적 타임 서버를 이용하는 설정이다  

2. driftfile /var/lib/chrony/drift  
지역 시스템 의 시간을 정확하게 유지하는 파일을 지정하는 항목으로 시간 오차의 평균값을 저장한다  

3. allow 192.168.0.0/16  
NTP 서버에 접근을 허용할 클라이언트를 지정한다  

4. server time.bora.net iburst  
기준이 되는 NTP 서버를 지정하는 항목이다  

* 관련 명령어  
1. chronyc  
명령행에서 사용되는 chrony 데몬의 정보 확인 및 변경하는 명령이다  

사용법  
chronys [option] [command]  

* Z-v: chronyc 버전 정보를 출력한다  
* sources: chronyd 데몬이 접근하는 서버들에 대한 정보를 출력한다
* tracking: 시간을 받아오는 서버에 대한 다양한 파라미터 정보를 출력한다
