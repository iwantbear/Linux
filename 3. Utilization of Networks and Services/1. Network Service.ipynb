### 1.1 웹 관련 서비스  
##### 1.1.1 웹 관련 서비스의 이해  

* 웹 작동 원리  
웹 브라우저를 사용하는 클라이언트와 웹 서버 사이에 HTTP 프로토콜을 사용하기 때문에 웹 브라우저를 HTTP 클라이언트, 웹 서버를 HTTP 서버라 부른다  

1. 주요 HTTP 요청 메소드  

* GET: 서버에서 자료를 가져오는 요청  
* POST: 서버에 정보를 저장하는 요청 
* HEAD: GET과 유사하나 HTTP 헤더 정보만 가져옴 
* PUT: 서버에 자료를 올릴 때 사용 
* DELETE: 서버의 자료를 삭제할 떄 사용  

telnel 명령을 이용하여 운영중인 서버의 80번 포트로 접속한 뒤에 'GET / HTTP/1.1' 이라고 입력하면 해당 웹 서버의 정보를 알 수 있다  

2. HTTP 응답  
HTTP 서버는 클라이언트의 요청에 대한 응답으로 HTTP 프로토콜 버전, 상태 코드, 데이터 등을 전송한다.  

* REST  

1. 개요  
분산 하이퍼미디어 시스템을 위한 소프트웨어 아키텍처의 한 형식이다  

2. 주요 특징  

2-1. 클라이언트-서버 구조  
2-2. 무상태성 연결  
2-3. 캐시 기능  
2-4. 균일한 인터페이스  
2-5. 계층화된 시스템  
2-6. 주문형 코드  

##### 1.1.2 웹 관련 서비스의 운영  
* Apache HTTP 서버와 PHP  
웹 서버 프로그램인 아파치는 HTML 문서를 보여주는 역할을 하고, PHP는 C언어 같은 특별한 언어적 구조를 갖는다.  

아파치 서버에 PHP 문서가 해석될 수 있도록 장착하는 것을 모듈이라부른다  

[주요 MPM 모듈]  
* prefork: 하나의 프로세스가 하나의 요청을 처리한다  
* worker: 하나의 프로세스가 여러 개의 요청을 처리한다 
* event: worker와 유사한 방식으로 아파치 2.4 버전 이후 소스 컴파일로 설치하면 기본 MPM으로 선택된다

2. extra 디렉터리 안의 파일  
2-1. 개요  
/usr/local/apache/conf/extra 디렉터리 안에 존재하는 파일은 기본 적으로 사용되지 않도록 설정 되어 있는데 사용하기 위해서는 두가지를 확인해야 한다  

* httpd.conf 파일의 하단에 위치한 Include 항목의 주석을 제거 
* 각 파일에서 요구하는 아파치 모듈을 찾아서 LoadModule 항목의 주석을 제거  

2-2. extra 디렉터리 안의 파일과 역할  

* httpd-mpm.conf: MPM관련 설정을 하는 파일  
* httpd-multilang-errordoc.conf: 에러 처리 문서에 대한 설정을 하는 파일 
* httpd-autoindex.conf: 디렉터리 목록을 표시할 때 여러 가지 설정을 하는 파일  
* httpd-userdir.conf: 개인 사용자의 홈 디렉터리 설정 시 사용하는 파일  
* httpd-info.conf: 웹 브라우저를 이용하여 웹 서버의 상태 정보를 파악할 때 사용하는 파일 
* httpd-vhosts.conf: 버추얼 호스트 사용할 때 설정하는 파일 
* httpd-dav.conf: WebDAV 관련 설정 파일 
* httpd-default.conf: Timeout 등 서버 운영의 기본적인 설정하는 파일 
* httpd-languages.conf: 문서의 언어 표시 관련 설정을 하는 파일 
* httpd-manual.conf: 운영 중인 서버 상에서 아파치의 매뉴얼 문서를 볼 수 있게 설정하는 파일  
* httpd-ssl.conf: SSL 관련 설정을 하는 파일 
* proxy-html.conf: HTML4 및 XHTML 1 을 이해하기 위해 mod_proxy_html 을 설정하는 파일  

2-3. httpd-userdir.conf 파일 활성화 예  

* httpd.conf 파일에서 LoadModule, Include 항목 활성화 (주석 없애기)  
* httpd-userdir.conf 파일 설정 (vi /usr/local/apache/conf/extra/httpd-userdir.conf)  
* 아파치 웹 서버 대몬 재시작 (/usr/local/apache/vin/apachectl restart)  


[httpd.conf 옵션 항목]  
* NONE: 어떤 옵션도 이용할 수 없으므로 모든 접근을 거부 
* ALL: MultiViews 옵션을 제외한 모든 옵션을 부여 
* Includes: 서버 측의 추가적인 정보를 제공 
* Indexes: URL이 지정된 디렉터리에 index.html과 같은 지정된 파일이 없을 경우 해당 디렉터리의 파일목록을 보여주는 옵션  
* FollowSymLinks: 디렉터리의 심볼릭 링크 사용을 허용 
* ExecCGI: CGI 스크립트를 실행할 수 있도록 허용 
* IncludesNOEXEC: SSI는 허용하지만 #exec 명령과 CGI 스크립트 #include 불허  
* MultiViews: ALL 옵션이 설정되었을 때만 지정된 목록의 multiviews를 허용  

[AllowOverride 항목]  
* NONE: AllowOverride를 사용하지 않음. 즉, 사용자 인증 파일을 사용하지 않음 
* ALL: httpd.conf 파일의 AccessFileName 지시자로 설정한 파일을 사용하고 사용할 수 있음 
* Authconfig: AccessFileName 지시자에 명시한 파일에 대해서 사용자 인증 지시자 사용을 허락  
* FileInfo: AccessFileName 지시자로 설정한 파일에 대해서 문서 유형을 제어하는 지시자 사용을 허락  
* Indexes: AccessFilename 지시자로 설정한 파일에 대해서 디렉터리 Indexing을 제어하는 지시자 사용을 허락  
* Limit: AccessFileName 지시자로 설정한 파일에 대해서 호스트 접근을 제어하는 지시자 사용을 허락 
* Options: AccessFileName 지시자로 명시한 파이에 대해서 Options, XBiHack 등과 같은 지시자 사용을 허락

1. httpd 명령을 이용한 실행  
아파치 웹 데몬을 실행하는 명령이다  

사용법  
httpd [option] [filename]  

* -t: 환경 설정 파일인 httpd.conf의 문법적 오류를 검사한다 
* -f: 새로 설정한 conf 파일이 있을 경우에 그 설정 파일로 아파치 데몬을 구동할 때 사용한다 
* -S: 설정된 가상호스트 및 주요 설정 정보를 출력한다
* -I: 정적 컴파일된 모듈에 대한 목록을 출력한다
* -k: httpd 데몬에 시그널을 보내는 옵션이다  
* -M: 정적 컴파일 모듈과 공유 모듈을 전부 출력한다 
* -V: httpd에 빌드된 파라미터와 버전 정보를 출력하낟
* -D: 환경설정과 관련된 파라미터를 지정해서 해당 정보만을 확인할 수 있다.  

2. 스크립트 파일을 이용한 실행  
아파치 웹 서버를 소스 설치하면 apachectl 이라는 데몬 실행 제어 스크립트를 이용하여 실행 및 제어를 할 수 있다.  

사용법  
apachectl command  

* help: 간단한 도움말을 출력한다
* start: 아파치 데몬을 시작한다
* stop: 아파치 데몬을 중지시킨다
* restart: 아파치 데몬을 재시작한다
* reload: 아파치 데몬을 재시작하지 않고, 환경 설정 파일만 다시 읽어들인다 
* configtest: httpd.conf 파일의 문법적 오류를 검사한다
* graceful: 웹 서버에 연결된 접속을 차단하지 않고, httpd.conf의 설정을 다시 읽어 드린다  
* graceful-stop: 웹 서버에 연결된 접속은 차단하지 않고, httpd 데몬을 중지시킨다
* status, fullstatus: 웹 서버의 상태를 간단히 또는 전체적으로 출력한다

3. 부팅 시 자동 실행되게 설정하기  
소스 컴파일로 설치한 경우에는 관련 실행 데몬 스크립트인 apachectl을 이용해서 데몬을 실행하거나 제어한다 

##### 1.1.4 웹 서버 운영

4. 개인 사용자의 설정

5. 웹 브라우저로 확인하기  

* 웹 서버 접근 통제  
아파치 웹 서버의 기본적인 설정은 모든 클라이언트의 접근이 허락되어 있다. 그러나 접근 통제가 필요한 경우에는 클라이언트의 IP 주소 또는 도메인에 대해 Order, Allow, Deny 지시자를 이용해서 제어할 수 있다.  

사용법  
1. Order Deny,Allow  
Deny from 주소  
Allow from 주소  
Allow 지시자가 지정한 주소만 접속을 허가할 때 사용한다  

2. Order Allow, Deny  
Allow from 주소  
Deny from 주소  
Deny 지시자에 지정한 사용자만 접속을 거부할 때 사용한다  

3. Order Mutual-failure  
Allow from 주소  
Deny from 주소  
Allow 지시자에 있고, Deny 지시자에는 없는 호스트만 접근을 허용한다  

[허용되는 주소 표기]  
* All, all: 모든 클라이언트를 의미한다 
* 도메인명: 특정 호스트명이 포함된 도메인명 또는 해당 도메인 전체를 지정할 수 있다.  
* IP 주소: 특정 호스트의 IP 주소를 사용할 수 있다.  
* 네트워크 대역: 특정 네트워크 대역에 속한 전체 호스트를 지정할 수 있다. 

* Apache 사용자 인증  
1. 설정 절차  
아파치 웹 서버에서는 특정 웹 페이지에 대해 다음과 같은 절차를 통해 사용자 인증을 할 수 있다.  

1-1. 사용자 생성 및 사용자 계정 파일 생성  
* htpasswd -c /etc/password posein  

1-2. httpd.conf 파일에 인증할 디렉터리 관련 설정  

1-3. 인증할 디렉터리에 관련 정보를 담고 있는 .htaccess 파일 생성  
* vi /usr/local/apache/htdocs/admin/.htaccess  


2. .htaccess 지시자  
* AuthType: 인증 형태를 지정한다. Basic 또는 Digest를 선택할 수 있다.  
* AuthName: 웹 브라우저 인증 창에 표시되는 메시지를 설정한다  
* AuthUserFile: 사용자 계정 파일명을 설정한다 
* Require: 접속을 허가할 사용자를 설정한다.  

3. 관련 명령어: htpasswd  
아파치 사용자 인증 파일을 생성하고 관리하는 명령이다.  

사용법  
htpasswd [option] 계정파일 사용자명  

* -c: 사용자 계정 파일을 생성할 때 사용하는 옵션으로 계정 생성할 때 한 번은 반드시 사용해야 한다 
* -B: Blowfish 기반의 bcrypt를 이용해서 패스워드를 암호화한다  
* -b: 암호를 묻는 명령줄 대신에 직접 지정한다 
* -n: 파일 업데이트와 관련된 내용을 표준 출력에 표시하지 않는다  

4. 생성 예  

4-1. 웹 사용자 계정 생성  
htpasswd 명령어를 사용해서 웹 사용자를 생성한다. 웹 사용자 계정은 리눅스의 계정과 무관하다

/usr/local/apache/conf/httpd.conf 파일 안에 인증할 디렉터리를 \<Directory> 태그를 이용해서 선언한다  

4-3. 해당 디렉터리에 인증 파일인 .htaccess 파일 생성  
httpd.conf 파일에서 선언한 디렉터리 내에 .htaccess 파일을 생성한다

### 1.2 인증 관련 서비스  
##### 1.2.1 인증 관련 서비스의 이해  

* 인증의 개요

* LDAP  
DAP 기반으로 만들어진 디렉터리 데이터베이스에 접속하기 위한 통신 규약이다. IP 프로토콜 기반으로 사용자, 시스템, 네트워크, 서비스 정보 등의 디렉터리 정보를 공유할 수 있다.

* st: 주 이름
* l: 보통 도시를 나타ㅐㄴ지만, 특정 지역 단위에서도 사용가능 
* street: 도로명 주소 
* o: 조직 이름
* ou: 조직의 부서 이름
* cn: 일반적으로 이름과 성의 조합
* sn: 성을 나타냄
* givenName: 이름을 나타냄
* dc: 도메인 네임의 요소를 나타냄
* mail: E-mail 주소 
* telephoneNumber: 전화번호를 나타냄

##### 1.3.1 삼바 서버 관리

* samba: 삼바 서버 패키지로 삼바 데몬, 서버 관련 라이브러리, 실행 데몬 스크립트 등이 들어있다.

1. 관련 데몬  
* smbd: 파일과 프린터 공유, 사용자의 권한 부여 및 확인 등 사용자 인증을 담당한다 
* nmbd: WINS를 담당하는 데몬이다  

2. 삼바 서버의 설정 절차  

2-1. 접근 관련 설정

* 환경 설정 파일: smb.conf  

1. smb.conf 파일 내부의 유형  

* #, ;으로 시작하는 행: 주석으로 간주하여 무시된다.  
* []을 사용하는 행: 섹션을 정의하는 역할을 하고, 하나의 섹션이 정의된 후 그 다음 섹션이 정의될 때까지의 행들은 해당 섹션에 속한다.  
* name = value 행: 사용하는 옵션과 해당 값을 설정하는 행이다.  

2. smb.conf 의 주요 섹션  

* \[global]: 삼바 서버의 전체적인 환경 설정을 담당하는 섹션  
* \[homes]: 각 사용자들이 자신의 홈 디렉터리로 접근할 때의 권한을 설정하는 섹션  
* \[printers]: 프린터 관련 권한을 설정하는 섹션
* \[print$]: 프린터 드라이버 관련 섹션

3. [global] 섹션의 주요 항목  

* workgroup = SAMBA  
윈도우의 작업 그룹에 해당하는 항목으로 공유 그룹명을 지정한다  

* server string = Samba Server  
서버에 대한 설명을 설정한다  

* netbios name = MYSERVER  
윈도우에서 이름으로 접속할 때 관련 이름을 지정한다  

* interfaces = lo eth0 192.168.12.2/24 192.168.13.2/24  
네트워크 인터페이스별로 접근 제어할 때 설정한다  

* hosts allow = 127. 12.168.12. 192.168.13  
삼바 서버에 접속을 허용할 호스트를 지정할 때 사용한다  

* log file = /var/log/samba/log.%m  
삼바 서버에 접속하는 호스트의 로그를 기록하는 파일을 지정한다  

* max log size = 50  
로그 파일의 최대 크기를 설정하는 부분이다  

* security = user  
삼바 서버에 접속할 때 인증 모드를 설정하는 항목이다  

user: 삼바 서버에 접속할 때 사용자명과 패스워드를 입력해야 한다  
domain 또는 ads: 윈도우의 도메인 컨트롤러에 전달하여 유효한 사용자 여부를 확인할 때 사용  

* passdb backend = tdbsam  
security가 user일 때 사용자의 패스워드를 설정하는 방식  

4. [share] 섹션의 주요 항목  

* comment: 간단한 설명을 설정  
* path: 공유 디렉터리의 경로를 설정  
* read only: 공유 디렉터리를 읽기 전용으로 설정  
* writable, write ok: 공유 디렉터리를 쓰기 가능하게 설정  
* valid users: 공유 디렉터리를 이용할 수 있는 사용자를 지정 
* write list: 공유 디렉터리에 접근 및 쓰기 권한을 행사할 수 있는 사용자를 지정
* public, guest ok: 다른 사용자들이 이용할 수 있도록 설정할 때 사용한다 
* browseable, browsable: 공유 디렉터리의 리스트를 보여줄 때 설정한다
* printable: 공유 디렉터리에 스풀 파일을 지정할 때 설정한다  
* create mask, create mode: 파일을 생성할 때 허가권을 지정할 때 설정한다  

* 관련 명령어  

1. smbclient  
리눅스 및 유닉스에서 사용하는 삼바 클라이언트 명령으로 윈도우 서버로 접근할 때 사용된다  

사용법  
smbclient [option] [호스트명]  

* -L: 접속하려는 삼바 서버의 공유 디렉터리 정보를 출력한다
* -U: 삼바 서버에 접속할 때 사용자명을 입력한다
* -M: 지정한 호스트명에 메시지를 전송할 때 사용된다.  

2. smbstatus  
삼바 서버에 사용하는 명령어로 클라이언트와 연결된 상태를 출력한다  

사용법  
smbstatus  

3. testparm  
삼바의 환경 설정 파일인 smb.conf의 설정 여부를 확인한다  

사용법  
testparm [파일경로] [호스트명 IP주소]  

4. nmblookup  
WINS 서버에 질의할 때 사용하는 명령으로 DNS 서버에 질의하는 nslookup 명령과 유사하다  

사용법  
nmblookup [option]  

* -U: 조회할 서버 이름을 지정한다
* -R: 조회할 이름을 지정한다  

5. mount.cifs  
CIFS 기반의 마운트 명령으로 보통 윈도우에 공유된 디렉터리를 마운트할 때 사용한다  

사용법  
mount.cifs // 공유서버/디렉터리 /마운트할_디렉터리 [-o 옵션]  

* username=값: 접근할 때 사용할 사용자명을 지정 
* password=값: 접근할 때 지정하는 사용자의 패스워드를 지정할 때 사용  

6. smbpasswd  
삼바 사용자를 생성 및 삭제, 패스워드 변경, 활성 및 비활성화등 관련 정보를 변경하는 명령  

사용법  
smbpasswd [option] [사용자명]  

* -a: 삼바 사용자를 추가할 때 사용한다 
* -x: 삼바 사용자를 제거할 때 사용한다 
* -d: 삼바 사용자를 일시적으로 비활성화 할 때 사용한다
* -e: 삼바 사용자를 활성화 할 때 사용한다 
* -n: 패스워드 없이 로그인이 가능하도록 할 때 사용한다  

7. pdbedit  
삼바 사용자의 데이터베이스 파일인 SAM database를 관리하는 명령이다.  

사용법  
pdbedit [option] [사용자명]  

* -a: 삼바 사용자를 추가할 때 사용한다 
* -L: 등록된 삼바 사용자의 간단한 목록을 출력한다
* -v: 등록된 삼바 사용자의 정보를 자세히 출력한다
