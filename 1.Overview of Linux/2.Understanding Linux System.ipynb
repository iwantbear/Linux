### 2.1 리눅스와 하드웨어 
##### 2.1.1 하드웨어의 개요  

리눅스 시스템에서 CPU 정보확인  

* cat /proc/cpuinfo: 가상 파일 시스템으로 운영되는 /proc 디렉터리에서 CPU 정보는 cpuinfo 파일에서 저장되어 있다.  

* lscpu [options]: CPU 정보를 출력  
-h: 사용법에 대한 도움말 정보를 출력  
-e: 출력되는 형식을 지정해서 CPU 정보를 목록으로 간단히 확인할 수 있다.  
-J: 관련 정보를 json 문서 형식으로 출력한다.  

* lshw [options]: 하드웨어 목록 정보를 확인하는 명령이다.  
-C: 특정 항목에 대한 정보만 출력한다. cpu,memory 등을 지정할 수 있다.   
-h: 사용법에 대한 도움말 정보를 출력한다.  
-html: 하드웨어 정보를 html 문서 형식으로 출력한다.  
-json: 관련 정보를 json 문서 형식으로 출력한다.  

메모리 정보 확인  

* cat /proc/meminfo: 가상 파일 시스템으로 운영되는 /proc 디렉터리에서 메모리 관련 정보는 meminfo 파일에서 저장되어 있다.  

* lsmem [options]: 사용 중인 메모리의 상태 정보를 확인하는 명령이다.  
-h: 사용법에 대한 도움말 정보를 출력  
-J: 관련 정보를 json 문서 형식으로 출력  

* free [options]: 현재 사용 중인 메모리의 상태를 출력하는 명령으로 swap 메모리 정보도 확인할 수 있다.  
-m: 메모리의 상태를 MB 단위로 출력  
-k: 메모리의 상태를 KB 단위로 출력  
-h: 사람들이 보기 편하게 바이트, 기가바이트등 단위를 붙여 출력  

하드 디스크 드라이브  
리눅스는 현존하는 대부분의 하드 디스크 사용이 가능하다  
* IDE 디스크 설명: /dev/hda, /dev/hdb  
* SCSI, S-ATA, USB 메모리, SSD: /dev/sda, /dev/sdb등  

모니터와 비디오 어댑터  
리눅스는 현재 존재하는 대부분의 그래픽 카드와 모니터를 지원한다.  

네트워크 인터페이스  
대부분의 네트워크 인터페이스를 지원한다. 특히, 이더넷 카드는 제조사에 상관없이 대부분 지원  

키보드 및 마우스  
리눅스는 현재 존재하는 키보드와 마우스를 기본적으로 지원  

CD-ROM, DVD-ROM  
하드 디스크 인터페이스와 같이 사용되어 대부분 자동으로 인식되며, 하드 디스크와 마찬가지로 파일로 관리된다. 최근 리눅스 배포판에서는 사용자의 편의를 위해 /dev/cdrom, /dev/dvd 라는 이름으로 심볼릭 링크 파일을 생성해서 제공  


##### 2.1.2 하드웨어의 선택  

RAID: 여러 개의 하드 디스크가 있을 때 동일한 데이터를 다른 위치에 중복해서 저장하는 방식이다. 백업가능, 안정적인 데이터의 보존과 유지기능, 속도 향상등에 사용한다.  
단순히 여러 디스크를 하나로 묶어주는 기술로는 Linear가 있다.  

RAID 에서 사용하는 기술 
* 스트라이핑: 연속된 데이터를 여러 개의 디스크에 라운드로빈 방식으로 기록하는 기술  
* 미러링: 디스크에 에러가 발생 시 데이터의 손실을 막기 위해, 추가적으로 하나 이상의 장치에 중복 저장하는 기술  

RAID의 종류  
* RAID-0: 스트라이핑 기술을 사용하여 빠른 입출력 속도 제공
* RAID-1: 미러링 기술을 사용하여 두 개의 디스크에 데이터를 동일하게 기록 
* RAID-2: 디스크들의 에러를 감지하고 수정하기 위해 ECC 정보를 사용  
* RAID-3: 스트라이핑 기술을 사용하며 디스크를 구성하고, 패리티 정보를 저장하기 위해 별도로 하나의 디스크를 사용, 보통 대형 레코드가 많은 시스템에서 사용
* RAID-4: 블록 형태의 스트라이핑 기술을 사용하여 디스크를 구성하는데, 배열 형태로 구성된 하나의 디스크에 패리티 정보를 저장한다. 
* RAID-5: 패리티 정보를 이용하여 하나의 디스크가 고장이 발생한 경우에도 사용이 가능한 구성 방식으로 최소 3개의 디스크로 구성해야 한다.  
* RAID-6: 디스크에 2차 패리티 구성을 포함하여 매우 높은 고장대비 능력을 발휘 
* RAID-7: 하드웨어 컨트롤러에 내장되어 있는 실시간 운영체제를 사용하여 구성하는 방식으로 속도가 빠른 버스를 이용하고, 여러 가지 특성들을 제공
* RAID 0+1: 디스크 2개를 RAID-0의 스트라이핑 기술로 구성하고, 다시 RAID-1의 미러링으로 구성하는 방식으로 최소 4개의 디스크가 필요하다. 만약, 6개의 디스크라면 보통 3개를 RAID-0으로 구성후에 다시 RAID-1로 구성된다. 
* RAID-10: RAID 0+1의 반대 개념으로 디스크 2개를 먼저 미러링으로 구성하고 다시 스트라이핑하는 방식. 최소 4개의 디스크가 필요 
* RAID-53: RAID-3 방식에 별도로 스트라이프 어레이를 구성하는 방식  

LVM: 쉽게 찰흙으로 생각하면 된다. 여러개의 하드 디스크를 하나로 뭉쳐서 하나의 디스크인 것처럼 만들 수도 있고, 2개의 하드 디스크를 3개의 하드 디스크인것 처럼 만들 수도 있다.  
* 물리적 볼륨(PV): 실제 디스크에 물리적으로 분할한 파티션
* 볼륨 그룹(VG): 물리적 볼륨이 모여서 생성하는 덩어리를 VG라 한다
* 논리적 볼륨(LV): VG에서 사용자가 필요한만큼 할당하여 만들어지는 공간으로 물리적 디스크에서 분할하여 사용하는 파티션 
* 물리적 확장(PE): PV에서 나누어 사용하는 일종의 블록같은 영역

### 2.2 리눅스의 구조 
##### 2.2.1 부트 매니저 

부팅을 도와주는 역할을 하는 프로그램으로 특히 한 컴퓨터에 여러 운영체제가 설치되어 있을 경우에 선택하여 부팅할 수 있도록 해준다. 부트로더라고도 부르며, 하드 디스크의 맨 앞쪽 영역인 MBR에 설치된다. MBR은 부트 섹터로 디스크의 첫 번쨰 섹터(0번 섹터)에 해당하고 이 영역에는 부트 매니저 프로그램과 파티션 정보가 기록된다. 대표적으로 LILO,GRUB가 있다. 

GRUB  
GNU 프로젝트에서 만든 부트로더로 LILO에 비교하여 다양한 파일 시스템을 지원하고, 부팅 시에 커널 인자를 조정하여 동적인 부팅을 지원한다. 또한, 메뉴 인터페이스 방식을 기본으로 사용하지만, Bash와 같은 명령행 인터페이스를 추가로 제공하고 있다. 그래픽 메뉴와 배경 그림 삽입도 가능  

GRUB의 부팅 모드  
* 편집 모드: GRUB 목록에서 [e]키를 누르면 설정된 항목들이 나타나면서 사용자가 직접 편집할 수 있도록 해준다. 

* 명령 프롬프트 모드: GRUB 부트 화면에서 [c]키를 누르면 진입하게 되며 상호대화식으로 직접 입력할 수 있는 모드다. 

* GRUB 환경 설정 파일: /boot/grub2/grub.cfg 혹은 심볼릭 링크 파일인 /etc/grub2.cfg
1. GRUB_TIMEOUT=5: GRUB 부트 화면에서 대기하는 시간을 초 단위로 지정하는 부분  
2. GRUB_DISTRIBUTOR="\$(sed 's, release .*$,,g' /etc/system-release)": GRUB 부트 화면에서 각 엔트리 앞에 보여질 리눅스 배포판 이름을 추출할 때 사용  
3. GRUB_DEFAULT=saved: 부트 화면에 제시된 목록 중에 기본적으로 부팅할 모드를 선택하는 항목으로 일반적으로 0부터 N에 해당하는 정수값을 입력  
4. GRUB_DISABLE_SUBMENU=true: grub2-mkconfig 에서 버전 번호가 가장 높은 커널에 대해 최상위 메뉴 항목으로 생성하고 다른 모드 커널 또는 복구 모드에 대한 대체 항목을 하위 메뉴로 생성한다. true로 설정하면 하위 메뉴를 생성하지 않도록 설정  
5. GRUB_TERMINAL_OUTPUT="console": GRUB이 출력되는 터미널 장치를 설정하는 항목  
6. GRUB_CMDLINE_LINUX="---생략---rhgb quiet": 커널 인자값을 지정하는 항목  
7. GRUB_DISABLE_RECOVERY="true": 부트 메뉴 엔트리에 복구 모드를 표시할 것인지를 지정하는 항목. 'ture'로 설정하면 복구 모드는 엔트리에 나타나지 않는다.  
8. GRUB_ENABLE_BLSCFG="true": BLSCFG는 다양한 부트 로더 구현과 관리를 지원  

GRUB 관련 명령어  

* grub2-mkconfig[-o 파일명]: GRUB 환경 설정 파일을 생성하는 명령으로 환경 설정 파일을 변경한 후에는 반드시 실행해야 한다.  
-o: GRUB의 환경 설정 파일을 생성할 때 지정한다.  

* grubby [options]: 명령행 기반으로 grub2 관련 설정 및 확인하는 명령  
--default-kernel: 부팅시 적용되는 커널을 출력  
--default-index: 부팅시 적용되는 인덱스 순서 번호를 출력  
--set-default: 부팅시 적용되는 커널을 변경  
--default-title: 부팅시 적용되는 커널의 타이틀 정보를 출력  
--info=커널경로: 지정한 커널의 기본 정보를 비롯하여 인덱스 번호, 타이틀, 파라미터 등의 정보를 출력  

GRUB의 활용  

* root 패스워드를 잊어버렸을 경우  
1. 시스템을 재부팅하여 GRUB 부트 메뉴가 나타나도록 한다.  
2. 기본 부트 엔트리에서 [e]키를 눌러서 커널 인자값을 수정할 수 있는 편집 상태로 진입한다.  
3. 커서를 아래로 이동시켜 커널 인자값이 있는 줄의 맨 마지막 부분으로 이동한다.  
4. 커널 인자값인 'rhgb quiet'를 지우고 'rw init=/bin/sh'라고 입력한다.(\는 무시)  
5. [Ctrl] + [x]를 눌러서 부팅을 진행  
6. 사용자 확인 후에 passwd 명령을 사용해서 패스워드를 변경  
7. SELinux를 사용하는 경우에는 라벨 변경 명령을 실행해야 한다.(touch / .autorelabel)  
8. 시스템을 재부팅한다.(exec /sbin/init)  

* GRUB 패스워드 설정  
1. vi나 gedit를 이용해서 /etc/grub.d/00_header 파일을 열고 GRUB 사용자 및 패스워드를 설정한다. 만약 사용자 이름을 abc, 패스워드를 1234로 설정하는 경우에는 다음과 같은 순서로 4줄을 작성한다.  
1-1. cat <<EOF  
1-2. set superusers="abc"  
1-3. password abc 1234  
1-4. EOF  
2. grub2-mkconfig -o /boot/grub2/grub.cfg 명령을 실행한다.  
3. 재부팅 한 후에 GRUB 부트 메뉴에서 [e]키를 눌러서 편집 모드로 진입한다.  

* 콘솔 모드의 터미널 해상도 변경  
1. vi /etc/default/grub  
2. GRUB_CMDLINE_LINUX="--생략--rhgb quiet 뒤에 vga=값"을 추가한다.  
3. 변경된 내용을 적용하기 위해 grub2-mkconfig -o /boot/grub2/grub.cfg 명령실행  
4. 재부팅(reboot)  

* 네트워크 어뎁터 이름 변경  
1. vi /etc/default/grub  
2. GRUB_CMDLINE_LINUX="--생략-- rhgb quiet 뒤에 net.ifnames=0을 추가  
3. 변경된 내용을 적용하기 위해 grub2-mkconfig -o /boot/grub2/grub.cfg  
4. 재부팅(reboot)  



##### 2.2.2 리눅스 디렉터리 구조 및 역할  
디렉터리란 파일을 보관하는 곳이다. 리눅스의 디렉터리는 최상위에 해당하는 루트(/)를 중심으로 하위 디렉토리에 다수의 디렉터리가 존재하는 형태의 트리 구조  

디렉터리의 종류와 특징 
* /: 최상위에 있는 디렉터리로 루트 디렉터리라 한다. 
* /bin: 우리가 흔히 말하는 명령어라는 것들이 들어있는 디렉터리로 보통 PATH가 들어 있다. 대표적인 명령어인 cp, mkdir, rmdir, mv, rm, rat 등  
* /boot: 부팅 이미지 파일이나 커널 등 시스템 부팅 시 필요한 파일이 들어있는 디렉터리  
* /dev: 하드디스크, CD-ROM, 터미널등 실제로 존재하는 물리적인 장치를 파일화 하여 관리하는 디렉터리  
* /etc: 시스템 환경 설정 파일 및 부팅과 관련된 여러가지 스크립트 파일들이 들어있다
* /home: 개인 사용자들이 파일 및 디렉터리를 생성할 수 있는 공간에 해당하는 홈 디렉터리가 위치 예를 들어 abc 라는 계정을 만들면 /home/abc 라는 디렉터리 생성  
* /lib: 각종 라이브러리가 저장된 디렉터리로 커널 모듈도 여기있다.  
* /lib64: RHEL 7 버전부터 x86_64 기반으로 라이브러리의 대부분이 여기에 존재 
* /mnt: CD-ROM, 플로피 디스크, 하드디스크, 네트워크 파일 시스템 등을 마운트할 때 포인터가 되는 디렉터리  
* /opt: 응용 프로그램들의 설치를 위해 사용되는 디렉터리 
* /proc: 가상 파일 시스템으로 시스템에서 운영되고 있는 다양한 프로세스의 상태 정보, 하드웨어 정보, 기타 시스템 정보 등을 담고 있다. 
* /root: 시스템 관리자인 root 사용자의 홈 디렉터리다. 
* /sbin: 시스템 관리에 대한 명령어들이 들어 있는 디렉터리로 보통 root가 사용 
* /tmp: 임시 저장 디렉터리로서 각종 프로그램이나 소켓 파일, 프로세스 작업을 할 때 임시로 생성되는 파일을 저장하는 공간 
* /usr: 시스템 운영에 필요한 명령, 응용 프로그램들이 위치하는 디렉터리이다. 커널 소스, C언어 헤더 파일, C-컴파일러, mysql, php등이 이 디렉터리의 하위 디렉터리에 설치됨 
* /var: 시스템 운영 로그 파일과 스풀링과 같은 가변적인 데이터를 보관하는 디렉터리 
* /media: 이동식 장치인 cdrom, floppy 등을 마운트하기 위해 제공되는 디렉터리 
* /srv: 사이트에서 생성되는 데이터를 저장하는 디렉터리
* /sys: usb 처럼 hot plug 하드웨어 정보를 가지고 있는 디렉터리이며 계층적인 구조로 정보를 제공 
* /run: 부팅 이후에 동작중인 프로세스의 런타임 데이터를 저장하고 있는 디렉터리 

디렉터리 비교 
/tmp: 재부팅하면 안에 있는 파일들이 삭제되고, 재부팅 하지 않아도 10일마다 정기적으로 삭제  
/var/tmp: 재부팅을 해도 파일들이 삭제되지 않고, 30일마다 정기적으로 삭제

##### 2.2.3.1 부팅

부팅의 개요  
부팅은 컴퓨터의 전원이 켜진 후 운영체제가 가동되어 사용자가 컴퓨터를 사용할 수 있도록 만드는 과정이라고 할 수 있다.  

부팅 과정  
1. 컴퓨터 전원을 켜면 바이오스는 컴퓨터에 장착된 하드웨어를 점검  
2. 바이오스는 하드웨어가 검사가 끝나면 CMOS에 설정된 첫 번째 부팅 하드 디스크를 확인한다  
3. 첫 번째 하드 디스크의 MBR 영역에 있는 부트 매니저 프로그램을 실행한다.  
4. 부트 매니저 프로그램은 관련 환경 설정 파일을 참고하여 운영체제 부팅을 시작한다.  

1. 로그인  
리눅스는 다수의 사용자가 사용할 수 있는 운영체제이기 때문에 로그인 절차가 필요  

* /etc/issue: 사용자가 로컬의 터미널로 로그인 할 떄 'login:' 이라는 메시지 전에 출력되는 내용을 적는 파일  
* /etc/issue.net: 사용자가 텔넷을 통한 네트워크 접속할때 출력되는 메시지를 기록
* /etc/motd: 성공적으로 로그인되었을 때 접속된 사용자에게 출력되는 메시지를 기록하는 파일  

2. 로그아웃  
리눅스에 로그인만 되어 있고 작업을 하지 않으면 자원이 낭비될 뿐만 아니라 보안상의 위험을 야기할 수 있다 따라서 로그아웃은 필수적으로 수행해야 한다.  
강제로그아웃(root 권한에서): /etc/profile에 TMOUT=초 를 지정  

3. systemd  
시스템의 부팅 및 서비스 관리해주는 프로그램이다. PID가 1번이 할당  

system 관련 디렉터리  
* /etc/systemd/system: 유닛의 환경 설정 파일이 위치하는 디렉터리 .wants 라는 디렉터리가 존재하는데,특정 유닛의 구동에 필요한 유닛들의 설정 파일을 심볼릭 링크로 담고 있다. 
* /run/systemd/system: 런타임 유닛 파일이 위치하는 디렉터리 
* /usr/lib/systemd/system: 시스템에서 사용되는 service 및 target 유닛 파일이 위치하는 디렉터리로 /lib/systemd/system 디렉터리로 접근해도 된다.  

systemd의 구조  
3-1. 유닛: 유닛이라 부르는 대상으로 나뉘고 다양한 유형이 존재한다. 
* service: 관련 파일명이 .service로 끝나는데 서버에서 제공하는 서비스가 해당되며 이 서비스를 데몬이라고 부른다 
* target: 관련 파일명이 .target으로 끝나는데, 부팅 레벨, 특정 동기화 지점과 같이 유닛을 그룹화 할 때 사용한다. 
* socket: 관련 파일명이 .socket으로 끝나는데, 프로세스 간 통신에 사용되는 IPC 소켓을 의미 
* path: 관련 파일명이 .path으로 끝나는데, 특정 파일 시스템이 변경될 때까지 서비스의 활성화를 지연시킬 때 사용된다. 프린팅 시스템의 스풀 디렉터리를 사용하는 서비스에 사용  

3-2. 서비스: 시스템에서 제공하는 서비스, 즉 웹 서버나 메일 서버 같은 데몬을 뜻한다. systemctl 이라는 하나의 명령으로 제어된다. 종료할때는 [q]키를 입력  
  
서비스 상태에 대한 주요값  
* loaded: 프로세스 로드되는 유닛의 환경 설정 파일을 나타낸다
* enabled: 부팅 시에 활성화됨을 나타낸다
* disabled: 부팅 시에 비활성화됨을 나타낸다
* active(running): 프로세스가 하나 또는 그 이상의 프로세스에 의해 동작중임을 나타낸다
* active(waiting): 동작 중인 상태이나 특정 이벤트에 의해 대기 중인 상태를 나타낸다
* inactive(dead): 프로세스가 종료된 상태를 나타낸다. 
* static: 활성화가 되지는 않지만, 활성화되는 다른 유닛에 의해 활성화가 가능한 상태를 나타낸다  

3-3. 타깃: 타깃은 부팅 레벨, 특정 동기화 지점과 같이 유닛을 그룹화할 때 사용한다. 
* poweroff.target: 시스템을 종료시키는 타깃이다. 
* rescue.target: 응급복구 모드(init 체제의 runlevel 1)로 전환하는 타깃이다. 
* emergency.target: rescue.target과 유사하고 /를 읽기전용으로 마운트한다. 
* multi-user.target: 콘솔 모드(init 체제의 runlevel 3)로 전환하는 타깃으로 텍스트 기반 로그인만 지원 
* graphical.target: X Window(init 체제의 runlevel 5)모드로 전환하는 타깃으로 텍스트 기반 로그인 이외의 X 기반의 로그인을 지원한다. 
* reboot.target: 재부팅시키는 타겟   

이전 버전의 init에서는 부팅과 관련된 런레벨을 0~6까지 숫자값으로 정의했는데, systemd 체제에서는 사용자의 편의를 도모하기 위해 runlevel.target 이라는 이름으로 심볼릭 링크를 제공하고 있다. 관련 파일은 /lib/systemd/system 디렉터리에서 확인 가능  

부팅과 관련된 타깃의 주요 역할은 특정 레벨에서 특정 서비스들이 부팅 시에 자동으로 활성 또는 비활성 여부를 지정하는 것이다.  

부팅 시에 특정 서비스 자동 활성화 여부는 타깃 관련 디렉터리에 실행 파일에 대한 심볼릭 링크 파일을 생성하고, 비활성화는 심볼릭 링크 파일을 삭제하는 방법으로 지정된다. 

3-4. 소켓: 소킷 유닛은 systemd에 의해 제어되면서 소켓 기반으로 동작하는 파일 시스템이 해당된다. systemd 에서는 이러한 서비스를 telnet.socket 이라는 유닛으로 제공한다.  

4. systemd 관련 명령어  
systemctl : systemd 기반의 시스템 및 서비스 관리를 제어하는 명령이다  
systemctl [option] 명령 [서비스명]  

* -l, --full: 유닛 관련 정보를 출력할 때 긴 이름도 약어가 아닌 전체 출력한다. 
* -t, --type: 유닛의 유형을 지정하는 옵션이다. 주요 유형에는 service, target, socket등이 있으며 사용 가능한 유형의 확인은 -t help 로 확인할 수 있다 
* -a, --all: 유닛 정보를 출력할 때 모든 유닛을 지정하는 옵션이다
* --failed: 실패한 유닛 정보만 출력한다
* --now: enable 명령과 함께 사용되는데, 부팅 시 활성화 시키는 것과 동시에 관련 데몬을 즉시 실행한다. disable, mask 명령과도 함께 사용되어서 즉시 동작을 중지시킬 수 있다. 

런레벨 관련 주요 명령  
* get-default: 현재 시스템에 설정된 런레벨 target 정보를 출력한다 
* set-default 타깃명: 시스템의 런레벨을 지정한 target으로 변경한다
* isolate 타깃명: 지정한 타깃의 런레벨로 즉시 변경한다
* rescue: 응급 복구 모드로 즉시 전환한다. 'emergency'와 유사하다. 
* poweroff: 시스템을 즉시 종료한다. 'halt'와 동일하다 
* reboot: 시스템을 즉시 재부팅한다. 'kexec'와 유사하다.  

상태 정보 관련 주요 명령 
* list-units: 유닛 관련 정보를 출력한다. 아무 옵션 없이 systemctl 명령 사용 시 기본 적용된다. 긴 이름의 유닛의 경우에는 약어로 표기된다
* list-unit-files: 설치된 유닛 파일의 목록 및 상태 정보를 출력한다
* list-sockets: 소켓 관련 유닛의 정보를 출력한다
* list-dependencies: 명시된 유닛의 의존성 관련 있는 유닛 정보를 출력한다. 특별히 명시된 유닛이 없으면 기본 타깃의 의존성 정보를 출력한다  

서비스 제어 관련 주요 명령  
* enable 서비스명: 부팅 시에 특정 서비스를 구동시킨다 
* disable 서비스명: 부팅 시에 특정 서비스를 구동시키지 않는다 
* is-enabled 서비스명: 부팅 시에 특정 서비스가 구동 여부를 확인한다. 구동되면 'enable', 구동되지 않으면 'disable'로 표시 
* start 서비스명: 특정 서비스를 즉시 시작
* stop 서비스명: 특정 서비스를 종료 
* restart 서비스명: 특정 서비스를 재시작
* reload 서비스명: 특정 서비스의 환경 설정만 다시 읽어 들인다
* mask 서비스명: 특정 서비스의 유닛 파일을 /dev/null로 링크시켜서 시작되는 것을 막을 때 사용한다. 
* unmask 서비스명: mask 설정된 서비스를 해체한다 
* is-active 데몬명: 특정 데몬이 활성화되어 있는 지 여부를 검사한다. 활성화 되어 있으면 'active'로 표시, 비활성화되어 있으면 'inactive'로 표시된다 
* status 데몬명: 특정 데몬에 대한 상태 정보를 출력한다
* kill 데몬명: 특정 데몬의 프로세스를 종료
* daemon-reload: systemd 매니저 관련 환경 설정을 다시 읽어 들인다. 

systemd-analyze: 시스템 부팅과 관련된 성능을 분석하는 명령으로 부팅에 걸린 시간을 알 수 있다.  
systemd-analyze [argument]  

* time: 부팅시에 소요된 시간 정보를 출력한다. 크게 커널,램 디스크, 서비스 영역 3가지 정보를 출력하고 아무 옵션 없이 명령 실행 시에 이 값이 출력된다
* blame: 서비스 별로 부팅 시에 소요된 시간을 출력한다
* critical-chain: 각 유닛의 시간을 트리 형태로 출력한다 @ 문자 다음에 표시되는 시간은 유닛이 시작 또는 활성화되기까지 소요된 시간이고, + 문자 다음에 표시되는 시간은 유닛을 시작하는데 사용되는 시간이다 
* plot: 관련 정보를 SVG 이미지 파일로 생성한다. 

5. systemd 로그 관리  
systemd 관련 로그는 systemd-journald가 생성하고 관리한다. 로그인 관련 정보를 커널로부터 받고, 사용자 프로세스 관련 정보는 syslog로 부터 받는다. 관련 정보는 메타 데이터로 /run/log/journal 디렉터리에 저장되는데, 재부팅 하면 관련 정보는 사라진다  

journalctl [option] [항목]  

* -l, --full: 출력 가능한 모든 필드의 정보를 출력한다
* -r, --reverse: 역순으로 출력해주는 옵션으로 가장 최근 정보부터 출력한다
* -p, --priority=: syslog에 사용하는 로그 레벨을 지정하는 옵션이다. 로그 레벨명이나 숫자값을 적으면 되고, 지정한 레벨 이상의 정보를 출력한다 높은 순으로 정리하면 emerg(0), alert(1), crit(2), err(3), warning(4), notice(5), info(6), debug(7)이다
* --since=: 특정 날짜 이후의 정보만 출력한다 "2023-03-13 00:13:00" 형식으로 지정
* --until==: 특정 날짜까지의 정보만 출력한다 "2023-03-13 00:13:00" 형식으로 지정

/run/log/journal 디렉터리에 저장되는 정보는 휘발성 정보이므로 지속적인 로그 관리를 하려면 /var/log 디렉터리 내에 journal 디렉터리를 생성하고 관련 명령을 실행해야 한다  

6. 시스템 설정 주요 명령어 
timedatectl [command]: 시스템의 시간 및 날짜를 확인하거나 설정하는 명령이다.  

* status: 시스템의 시간 및 RTC의 시간 정보를 출력한다 
* set-time: 시스템의 시간을 설정하는 명령으로 "2023-12-22 21:05:13" 형식으로 설정
* set-timezone: 타임존을 설정하는 명령. 우리나라인 경우에는 'Asia/Seoul'로 지정
* set-ntp 값: NTP 사용 여부를 지정하는 명령으로 값을 1로 설정하면 사용이고, 0으로 설정하면 미사용으로 설정된다. 

tzselect: 타임존 설정값을 확인하는 명령어로 대화형으로 진행된다. 실행하면 먼저 대륙이나 대양을 선택하고, 국가를 선택하면 타임존 설정값을 출력한다  

systemctl start chronyd(RHEL 8): 표준 시간 설정  

hostnamectl [command] [설정값]: 시스템에 설정된 호스트명 정보를 출력하거나 설정하는 명령어  

* status: 시스템에 설정된 호스트명과 관련된 정보를 출력한다. command 입력이 없을 경우 기본값으로 설정되어 있다. 
* set-hostname: 호스트명을 설정할 때 사용한다  

localectl [command] [설정값]: 일종의 표시 언어 및 키보드 레이아웃에 해당하는 시스템 로케일 정보를 확인하거나 설정하는 명령이다.  

* status: 시스템에 설정된 로케일 정보를 출력한다. 별도의 command 입력이 없으면 기본값으로 설정되어 있다. 
* list-localed: 설정 가능한 로케일 목록 정보를 출력한다
* set-locale: 로케일을 설정할 때 사용한다.

* -c: 예약된 셧다운 명령을 취소한다

halt [option]  

* -p: 시스템 종료하고 전원까지 끌 때 사용하는 옵션

poweroff  

5. init, telinit: 모든 프로세스의 조상인 init 프로세스에 직접 요청하여 실행레벨을 변경할 때 사용하는 명령으로 빠르게 실행되지만, 프로세스를 무조건적으로 종료한다

* 파일 공유를 위하여 판독 접근, 기록 접근, 수행 접근 등의 다양한 접근 제어 방법 제공

* 사용자가 파일이나 디렉터리에 접근하기 쉬운 인터페이스 및 명령어 제공
* 정보가 안전하게 보호되고 비밀이 보장될 수 있도록 정보의 암호화 및 복호화 기능 제공
* 사용자와 장치 간의 독립성을 유지하기 위해, 사용자가 물리적인 장치 이름 대신에 적절한 이름 제공

### 2.3.1 X 윈도  
##### 특징 
X 윈도는 네트워크 프로토콜에 기반한 그래픽 사용자 인터페이스 환경이다. 디스플레이 장치에 의존적이지 않고, 서로 다른 이 기종을 함께 사용할 수 있다.  

##### 구조
X 윈도는 클라이언트/서버 구조로 되어 있고, 서로간의 통신을 위해 X Protocol을 사용한다. X 서버는 그래픽하게 보여주는 하나의 시스템을 말하고, X 클라이언트는 X 서버 위에 동작하는 일종의 응용 프로그램을 말하며 TCP 포트 6000번을 통해 전달된다. 

##### X 윈도의 실행 
1. 부팅 모드 확인
systemctl get-default  

2. 부팅 모드의 변경  
systemctl set-default [TARGET_NAME]  

* multi-user.target : 콘솔 모드(init 체제의 runlevel 3)로 전환하는 타깃으로 텍스트 기반 로그인만 지원한다. runlecel3.target 으로도 지정 가능하고 .target은 생략 가능하다  
* graphical.target: X Window(init 체제의 runlevel 5)로 전환하는 타깃으로 텍스트 기반 로그인 이외에 X 기반의 로그인을 지원한다. runlevel 5로도 지정 가능하고 .target은 생략가능하다. 

##### startx  
startx [option]: X 윈도를 실행시키는 스크립트로 시스템 환경을 초기화시키고 시작하는 데 필요한 여러 프로그램을 호출하고, 최종적으로 xinit을 호출하는 명령이다.  

* --: 명령행의 인자값을 xinit에 넘겨준다.  

##### X 윈도의 설정 변경  
xrandr [options]: 명령행에서 화면 크기 및 방향 등을 확인 및 변경하는 명령이다.  

* -s: 화면 크기, 즉 해상도를 변경할 때 사용한다.

데스크톱 환경이란 GUI 환경을 이용하기 위해 사용자에게 제공되는 인터페이스 스타일을 말한다.

##### 2.3.3 윈도 매니저  
X 윈도 환경에서 윈도의 배치와 표현을 담당하는 시스템 소프트웨어이다. 대부분의 윈도 매니저 프로그램은 데스크톱 환경 구성에 도움을 주기 위해 설계되었다. 기본적인 기능으로 다양한 유틸리티를 제공한다.  

##### 2.3.4 웨이랜드  
웨이랜드는 클라이언트가 해당 프로토콜의 C 라이브러리 구현뿐만 아니라 클라이언트와 대화할 수 있는 compositor를 위한 프로토콜이다.  

처리과정  
1. 리눅스 커널에서 입력 장치로부터 커널 이벤트 수신 후 웨이랜드 compositor 에게 이벤트 전달
2. compositor가 이벤트를 처리할 웨이랜드 클라이언트에게 처리할 window 정보를 알고 이벤트를 전달
3. 웨이랜드 클라이언트가 자신의 window에 대해서 직접 렌더링을 수행하고 업데이트된 영역을 나타내기 위해 관련 요소가 필요한 경우에만 compositor에게 요청  
4. compositor는 각 cliens의 damage event 요청을 수집하여 화면을 재구성한다 또한 KMS와 함께 pageflip 스케줄을 위해 ioctl를 직접 발생한다.

* export DISPLAY="203.247.51.100:0 : X 클라이언트 프로그램을 203.247.51.100으로 전송

### 2.4 셸
##### 2.4.1 셸의 이해 
1. 셸 이란?
셸은 커널과 사용자 간의 다리 역할을 하는 것으로 사용자로부터 명령을 받아 그것을 해석하고 프로그램을 실행하는 역할을 한다. 셸은 커널에서 분리된 별도의 프로그램이어서 다양한 종류의 셸이 존재한다.  

2. 주요 셸의 특징 
* Bourne Shell: 유닉스 버전7의 기본 셸
* C 셸: C언어를 기반으로 만들어졌으며 강력한 프로그램 작성 기능을 가지고 있지만 명령행 편집 기능은 지원하지 않는다. 
* tcsh: 테넥스라는 운영체제에 명령행 완성, 편집 기능을 지원한다. 
* ksh: 콘 셸은 본 셸을 확장해서 만들어졌다. 
* bash: GNU 프로젝트를 위해 개발한 셸로 본 셸을 기반으로 만들어졌으며 현재 리눅스의 표준 셸
* zsh: ksh와 tcsh의 기능을 결합하여 만들었으며 macOS와 Kali Linux의 기본 셸
* dash: POSIX와 호환되는 /bin/sh를 가능한 한 작게 구현한 셸이지만 history 명령은 지원하지 않는다.  

3. 셸의 확인 
시스템에 로그인한 후에 사용 중인 셸을 확인하려면 echo $SHELL, 변경 가능한 셸의 확인은 chsh -l 혹은 cat /etc/shells  

4. 셸의 변경  
chsh 명령으로 셸을 변경할 수 있는데 사용자 암호를 물어본다. 입력한 후 셸을 절대경로로 입력하면 변경할 수 있다. 변경한 셸의 적용은 다음 로그인부터 유효하다. 참고로 명령행에서 /bin/csh로 셸을 변경할 수 있으며 이 경우에 변경된 셸의 확인은 ps로 , exit를 사용하여 기본 셸인 bash로 돌아갈 수 있다.  

5. 사용자의 로그인 셸 정보 확인  
* grep [사용자 이름] /etc/passwd : 사용자의 로그인 셸 정보는 /etc/passwd의 7번째 필드에 기록된다.  

6. chsh [option] [사용자명] : 사용자의 로그인을 변경하는 명령어, 일반 사용자인 경우 사용자 패스워드를 입력해야 변경이 가능하며 실제 적용은 다음 로그인부터 유효하다.  

* -s: 변경하려는 셸을 명시할 때 사용한다 
* -l: 사용 가능한 셸의 목록 정보를 출력한다. 
* -u: chsh 명령어의 사용법을 화면에 출력한다
* -v: 설치한 패키지의 버전 정보를 출력한다 

* 셸 변수: 특정한 셀 에서만 적용되는 변수를 말한다. 변수명=값 형태로 지정하여 사용할 수 있고, 변수값을 출력할 때는 변수명 앞에 $을 붙이고 echo 명령어로 확인할 수 있다. 선언된 셸 변수를 전부 확인하려면 set 명령을 사용하면 된다.  

예시  
city=seoul  
echo city  

* 환경 변수: 프롬프트 변경, PATH 변경 등과 같이 셸의 환경을 정의하는 중요한 역할을 수행하는 변수를 말한다. 미리 예약된 변수명을 사용하고 현재 설정된 전체 환경 변수의 값은 env 명령으로 확인 가능하고, 특정 환경 변수의 값 확인과 설정은 일반 셸 변수 설정과 같다.  

* 주요 환경 변수  

* HOME: 사용자의 홈 디렉터리
* PATH: 실행 파일을 찾는 디렉터리 경로 
* LANG: 셸 사용 시 기본으로 지원되는 언어 
* TERM: 로그인한 터미널 종류 
* PWD: 사용자의 현재 작업 디렉터리 
* SHELL: 사용자의 로그인 셸
* USER: 사용자의 이름 
* DISPLAY: X 윈도에서 프로그램 실행 시 출력되는 창 
* PS1: 프롬프트 변수 
* PS2: 2차 프롬프트 변수
* HISTFILE: 히스토리 파일의 절대 경로 
* HISTSIZE: 히스토리 파일에 저장되는 명령어의 개수 
* HISTFLIESIZE: 히스토리 파일의 파일 크기
* HOSTNAME: 시스템의 호스트명 
* MAIL: 도착한 메일이 저장되는 경로 
* TMOUT: 사용자가 로그인한 후 일정 시간 동안 작업을 하지 않을 경우에 로그아웃 시키는 시간으로 단위는 초 
* UID: 사용자의 UID  

* 환경 변수의 사용 및 변경 
환경 변수를 사용하여 각 사용자 고유의 셸 환경을 구축할 수 있고, 명령어와 결합하여 이용할 수 있다.  

사용 예시  
1. mkdir $HOME/data : 해당 사용자 홈 디렉터리 안에 data라는 디렉터리를 만든다 
2. PATH=$PATH:$HOME/data : PATH 경로에 $HOME/data라는 경로를 추가한다
3. echo $LANG : 현재 설정된 언어를 확인
4. LANG=C : 영어로 변경
5. echo $PS1: 설정된 프롬프트를 확인 후 변경한다  

주요 프롬프트 형식  
* /d: '요일 월 일' 형태로 날짜를 표시 
* /h: 호스트 이름을 표시 
* /s: 사용중인 셸의 이름을 표시 
* /t: 24시 형태의 현재 시간을 표시한다 
* /T: 12시 형태의 현재 시간을 표시한다
* /@: 12시 형태의 현재 시간에 AM/PM을 추가로 표시한다
* /u: 현재 사용자의 이름을 표시한다 
* /w: 현재 작업 디렉터리를 절대 경로로 표시한다 
* /W: 현재 작업 디렉터리 중 마지막 디렉터리만 표시한다
* /!: 현재 명령의 히스토리 넘버를 보여준다
* //: \를 표시한다

##### bash의 주요 기능  

1. 명령행 History 기능 
bash 에서는 입력 후 실행했던 모든 명령들은 히스토리 버퍼에 스택으로 저장된다. history라는 명령을 입력하면 히스토리 리스트에 있는 명령어들이 출력된다. 또한 사용자들이 실행한 명령들은 각 사용자의 홈 디렉터리 안에 .bash_history라는 파일에 추가로 기록되며, 로그아웃 할 때 메모리에 기억된 명령의 목록을 파일에 저장한다. 

1-1 history [숫자값]: 사용자가 입력한 명령어를 확인하는 명령
* 자연수 : 최근에 입력한 마지막 자연수 개의 명령어 목록을 출력 
* !! : 마지막에 사용한 명령을 실행
* !n : n번째 사용한 명령을 실행한다
* !-n : 사용한 명령 목록을 역으로 세어서 n 번째 명령을 실행한다
* !문자열 : 가장 최근에 사용한 명령중에 '문자열'로 시작하는 명령을 찾아서 실행한다 
* !?문자열? : 가장 최근에 사용한 명령중에 '문자열'을 포함하고 있는 명령을 찾아서 실행
* ^문자열1^문자열2 : 마지막에 사용한 명령문의 '문자열1'을 '문자열2'로 대체한 후에 실행

1-2 히스토리 관련 환경 변수 
* HISTSIZE: 히스토리 스택의 크기가 지정되어 있는 변수로 단위는 명령의 개수이다. 이 변수에 지정한 값만큼만 history 명령 실행 시에 출력되고, 방향키로 검색했을 경우에도 이 값 내에서만 가능하다 
* HISTFILESIZE: 실질적인 히스토리 파일의 크기이다
* HISTFILE: 히스토리 파일의 위치를 보여준다
* HISTCONTROL: 연속으로 중복된 명령에 대한 기록 유무를 지정하는 변수
* HISTTIMEFORMAT: history 명령 실행 시 출력되는 시간 형식을 지정할 때 사용

1-3 관련 파일: .bash_history  
이 파일은 사용자의 홈 디렉터리 내에 존재하면서 사용자들이 입력했던 명령들이 기록되는 파일이다. 파일명이 '.'으로 시작되어서 숨겨진 파일 형태로 존재하고, 현재 로그인해서 내린 명령들은 로그아웃할 때 기록되므로 다음 번 로그인 시에 확인할 수 있다.  

2. alias 기능  
별명이란 뜻으로 어떠한 명령을 지정해 놓으면 사용자가 그 명령을 실행했을 때 alias로 지정해 놓은 명령이 대신 실행

사용법  
2-1. alias 설정 : $ alias 나만의_명령어='기존 명령어의 조합'
2-2. alias 확인 : $ alias 지정한_명령어
2-3. alias 해제 : unalias 지정한_명렁어  

3. 명령행 편집 기능  

* [Ctrl]+[b] : 커서를 왼쪽으로 한 칸 이동 
* [Ctrl]+[f] : 커서를 오른쪽으로 한 칸 이동 
* [ESC] 후 [b] : 커서를 왼쪽으로 한 단어 이동 
* [ESC] 후 [f] : 커서를 오른쪽으로 한 단어 이동 
* [Ctrl]+[a] : 맨 왼쪽으로 이동 
* [Ctrl]+[e] : 맨 오른쪽으로 이동 
* [Ctrl]+[k] : 커서 왼쪽 행 전체 삭제 
* [Ctrl]+[u] : 행 전체 삭제 
* [Ctrl]+[y] : 삭제 취소 
* [Ctrl]+[d] : 커서 오른쪽 한 글자 삭제 
* [ESC] 후 [Backspace] : 커서 왼쪽 한 단어 삭제 
* [ESC] 후 [d] : 커서 오른쪽 한 단어 삭제 

4. 명령 대체(명령 치환) 기능  
특정 명령의 결과를 다른 명령어의 인자값으로 사용하는 것을 말한다. 예를 들면 passwd라는 명령의 허가권을 알기 위해서는 먼저 'which passwd'라는 명령으로 passwd 명령어의 위치를 찾아내고, 다시 'ls-l' 명령을 수행해야 한다.

5. 그룹 명령 실행 
하나의 명령행에 여러개의 명령어를 동시에 사용할 수 있는데, ';'를 사용하여 명령어를 구분하면 순차적 처리가 가능하다. 

* ; : 단순히 한 줄에 여러 명령을 나열하기 위해 사용하는데, 입력한 순서대로 순차처리 한다
* || : 논리적 OR이라 부르며 앞의 명령이 성공이면 결과를 출력하고, 그렇지 않으면 뒤의 명령을 실행하여 결과를 출력한다
* && : 논리적 AND라 부르며 앞의 명령이 성공적으로 수행되어야만 다음 명령을 수행한다
* (): 특정 명령들을 그룹으로 묶어서 실행할 수 있다. 

6. 표준 입출력 제어 기능 
* 표준 입력: 입력을 담당하는 키보드, stdin 이라고 하며 셸에서는 숫자값 0으로 표기 
* 표준 출력: 명령의 결과값이 보이는 화면, stdout 이라고 하며 셸에서는 숫자값 1로 표기
* 표준 에러: 에러값이 보이는 화면, std err이라고 하며, 셸에서는 숫자값 2로 표기 

존재하지 않는 명령을 입력하면 오류 메시지를 확인할 수 있지만, 숫자 2와 리다이렉션 기호를 사용하면 오류 메시지를 없앨 수 있다 (명령어 2>/dev/null)  

7. 리다이렉션 
리다이렉션은 어떤 프로세스의 입/출력을 표준 입출력이 아닌 다른 입출력으로 변경할 때 사용한다. 출력 결과를 파일로 저장하거나 파일의 내용을 프로세스의 입력으로 사용하는 기법이다.  

* \> : 프로세스의 출력을 표준 출력에서 다른 출력으로 변경한다
* \>> : 프로세스의 출력을 지정한 출력에 추가한다
* < : 프로세스의 입력을 표준 입력에서 다른 입력으로 변경한다
* <<: 기호 뒤에 지정한 문자열을 입력으로 받아 해당 문자열이 나오기 전까지를 표준 입력을 삼아 해당 내용을 출력한다 

8. 파이프  
프로세스의 통신을 위해 도입한 것으로 어떤 프로세스의 표준 출력이 다른 프로세스의 표준 입력으로 쓰이게 하는 것  

* 명령어1 | 명령어2 : 명령어1의 결과값이 명령어2의 입력으로 사용된다  

예: ls -alF | more  
ls 명령의 출력이 more라는 filter의 입력으로 사용되어 결과가 한 화면씩 출력된다  

관련 명령어 8-1  
tee : 파이프 연결 출력을 두 갈래로 나눌 때 사용하는 명령  

tee [option] [파일]  

* -i : 인터럽트를 무시  
* -a : 지정된 파일로 출력을 덮어쓰지 않고, 파일 내용 뒤에 추가  

예: ls -l | tee list.txt |more  
파일의 목록을 list.txt라는 파일에 기록함과 동시에 more를 사용하여 화면에도 출력 내용을 보여준다  

관련 명령어 8-2  
xargs : 표준 입력으로부터 값을 받아 처리하는 명령으로 보통 다른 명령어와 파이프 기호 조합으로 사용  

명령어 | xargs [option]  

* -n : 한번에 처리하는 최대 인자값을 지정하는 옵션  

예: echo a.txt | xargs ls -l  
a.txt의 ls -l 결과를 출력  

9. 산술 연산 기능  
expr 명령을 이용하여 산술 연산이 가능하다.  

expr 값 연산자 값 : expr값, 연산자 사이는 반드시 한 칸씩 띄어야 한다. 또한 값 대신에 미리 지정한 변수를 사용해도 된다.  

* \+ : 더하기
* \- : 빼기
* \* : 곱하기
* \/ : 나누기
* %  : 나머지 값 구하기  

예: expr 3 + 2  
3과 2를 더하는 연산으로 5라는 값 출력 반드시 명령어, 연산자, 피연산자 사이를 띄워야 함 

##### 확장된 내부 명령어  
bash 자체적으로 해석하는 set, export 등의 내부 명령어  

1. bash 관련 파일 및 디렉터리  
사용자가 명령행에서 설정한 환경 변수나 alias는 일시적으로 사용 가능하고, 다음 로그인 시에는 적요되지 않는다. 따라서 해당 설정들을 지속적으로 이용하려면 관련 파일에 설정해야 한다.  

* /etc/profile : 모든 사용자에 적용되는 환경 변수와 시작 관련 프로그램을 설정
* /etc/bashrc : 모든 사용자에 적용되는 alias와 함수를 설정  
* ~/.bash_profile : 개인 사용자의 환경 설정, 시작 프로그램 설정과 관련이 있는 파일로 로그인 시 읽어 들인다. 경로, 환경 변수 등의 설정이 들어있고, 사용자가 PATH와 같은 환경 변수 수정 시 사용하면 된다. 
* ~/.bashrc : 개인 사용자가 정의한 alias와 함수들이 있는 파일이다. alias 를 지속적으로 사용하려면 이 파일에 설정한다 
* ~/.bash_logout : 개인 사용자가 로그아웃 할 때 수행하는 설정을 지정하는 파일 
* /etc/profile.d : 몇몇 응용 프로그램들이 시작할 때 위한 필요한 스크립트가 위치하는 디렉터리로 보통 /etc/profile 에서 호출된다. 일반 사용자의 alias 설정등과 관련된 스크립트가 존재한다  

2. 셸에서 사용되는 특수 문자 1
셸에서 특별한 의미를 가지고 있는 특수문자들. 메타 문자라고도 한다  

* ~(tilde) : 홈 디렉터리를 나타냄 
예1 : ~posein : posein이라는 사용자의 홈 디렉터리  
예2 : ~+ : 현재 디렉터리를 나타내고, '.'과 같다  
* . : 현재 디렉터리를 나타내거나 명령행 맨 앞에서 sourece라는 의미  
* .. : 한단계 위 디렉터리, 즉 부모 디렉터리 
* '' : 모든 문자나 특수 문자들을 일반 문자로 취급한다 
* "" : $,`,\,! 를 제외한 모든 문자들을 일반 문자로 취급한다.
* `` : 명령 대체 기능을 수행하는데, 명령의 결과를 대체해서 사용한다 
* \# : 주석
* $ : 쉘 변수 기호로 뒤에 오는 문자열을 변수로 취급 
* & : 특정 명령을 후면 작업으로 실행할 때 사용 
* \* : 아무것도 없는 경우를 포함한 모든 문자를 뜻함 
* ? : 보통 한 문자를 대체할 때 사용, 추가로 특정 명령에 대한 결과를 갖는 Return 변수로도 쓰이는데, 셸 상에서 내린 명령의 에러 유무를 확인할 수 있다. 
* () : 부속 셸을 뜻하는 기호로 하나의 셸 단위로 묶어준다 
* ₩ : 탈출 문자로, 바로 다음에 오는 특수 문자의 기능을 없앤다. 또한, alias가 설정된 명령어 앞에 사용하는 경우에는 alias를 없애준다. 추가로 셸에서 긴 명령행 입력 시에 행을 연장할 때도 사용한다 
* ; : 셸 명령 분리자. 명령어를 순차적으로 실행할 때 사용
* / : 디렉터리 경로 분리자 
* {} : {} 안에 제시된 문자열 중 하나로 대치시킨다 
예: b{ed, ot}s : bed, bots  
* [] : 브라켓이라고 부르며 '['과']' 사이에 선택할 수 있는 문자를 나열하여 ~중의 하나 라고 표현한다. 범위를 지정할 때는 -를 사용한다 
예1 : [abc] : a,b 또는 c  

3. 셸에서 사용되는 특수 문자 2  
문자나 단어의 특별한 의미, 즉 메타 문자가 가지고 있는 특수한 기능을 제거할 때 사용하는 부호를 인용 부호라 한다. 위에서 언급되었지만 많이 사용되므로 다시 정리한다.  

* \$ : \$는 셸에서 변수를 나타낼 때 사용한다. 변수에 특정한 값으로 부여할 때는 = 를 사용  
예시  
1. name=posein
2. echo name  
echo 명령은 텍스트 문자열을 그대로 출력하는 명령어 이므로 name이 그대로 출력됨  

3. echo $name  
name을 변수 취급해서 변수에 저장된 값을 출력 (posein)  

* '' : 모든 특수 문자의 의미를 제거한다  
예시  
1. echo '$PATH'  
'' 안의 $PATH를 단순한 문자열로 표기한다 

* "" : 따옴표 안에 있는 $,`,\ 를 제외한 모든 문자(*,?등)의 특별한 의미를 없앤다 (!도 예외로 인정)  

* ₩(\) : ₩(\)기호 다음에 나오는 문자의 특별한 의미를 제거하건, 명령행이 긴 경우 다음 라인까지 명령행을 연장할 때 사용한다. 아울러, 특정 명령어에 옵션 기본 장착과 같은 alias가 설정되어 있을 경우 앞부분에 ₩를 덧붙여 실행하면 옵션없이 기본 명령어로 실행된다  

* \`\` : 명령어 대체 역할을 하는 문자로서 \`\`로 묶여진 문자열은 명령어로 인식된다 
예시  
echo "Current directory \`pwd`"  
\`` 안의 pwd를 명령어로 인식하여 현재 위치를 출력한다  


##### 2.4.2 Shell 프로그래밍  

1. 개요  
셸 프로그래밍이란 셸에서 사용되는 여러 명령어들을 모아 하나의 파일로 만드는 과정을 말하며, 이러한 과정을 통해 만들어진 파일 또는 프로그램을 셸 스크립트라고 부른다. 셸은 for문 및 case문 같은 구문, 변수, 함수등과 같은 보통 언어가 가지고 있는 기본적인 특성을 가지고 있다.  

2. 셸 스크립트 작성하는 방법  

2-1. vi 편집기 등을 사용해서 파일을 생성한다.  
* vi who2.sh  

2-2. 첫번째 줄은 사용할 셸을 명시한다. 셔뱅이라고 부르는 해시와 느낌표 기호의 조합인 #! 기호로 시작하면서 스크립트의 맨 첫 줄에 위치한다. bash를 사용하는 경우에는 다음과 같이 명시한다  
* #!/bin/bash  

2-3. 두번째 줄 부터는 원하는 명령 및 구문 등을 입력한다  
* echo "Login List"  
* who  

2-4. 스크립트를 실행 가능한 파일로 만든다  
* chmod 755 who2.sh  

2-5. 실행시킨다  
* ./who2.sh  

3. 셸 스크립트 실행하기  

3-1. 경로(PATH)등록  
셸 스크립트를 생성한 뒤에 실행하려면 파일명 앞에 반드시 ./를 붙여야 한다. 이 의미는 현재 디렉터리 안에 있는 해당 파일을 실행시키라는 뜻이다.  

계속해서 사용할 셸 스크립트라면 등록된 디렉터리 중에 위치시켜야 하는데 개인 사용자의 홈 디렉터리만 사용할 수 있다. 따라서 'mkdir ~/bin' 또는 'mkdir ~/.local/bin' 명령을 실행한 후에 이 디렉터리 안에 넣어두면 현재 디렉터의 위치와 상관없이 실행시킬 수 있다.  

3-2. 허가권 없이 셸 스크립트 실행하기  
셸 스크립트에 실행 권한을 부여하지 않고도 실행하려면 다음과 같은 3가지 방법을 사용할 수 있다.  

3-2-1. 앞부분에 sh 명령 덧붙이기  
* sh who.sh  

3-2-2. 앞부분에 source 명령 덧붙이기  
* source who2.sh  

3-2-3. 앞부분에 . 덧붙이기  
* . who2.sh  


##### 셸 프로그래밍 문법  

1. 변수  
변수는 모든 프로그래밍 언어의 가장 기본이 되는 요소로 어떠한 데이터를 저장하는 공간이다. 셸에서의 변수형은 문자열 만을 가지며 C언어처럼 변수형 선언이 필요 없다.  

사용법  
var=value  
var 변수에 값 value를 저장한다. 셸에서 변수를 사용할 때는 변수명 앞에 $를 붙인다. 또한 변수에 데이터를 저장할 때는 대입 연산자는 '='를 사용하여 대입 연산자, 연산자/피연산자 사이에는 공백이 존재해서는 안되며 언더스코어를 제외한 특수 문자와 숫자로 시작해서는 안된다  

기타 변수 대응법  

* $[name] : name이라는 변수에 들어있는 값으로 치환한다.
* $[name:=value] : name이 Null이면 value로 할당하여 저장하고, name에 값이 있으면 그 값을 사용한다.  
* $[name:+value] : name이 null이 아니라면 value를 사용하지만 name에 저장하지는 않는다 (1회 사용)  
* $[name:-value] : 이미 name에 값이 있다면 그 값을 사용하고, 값이 없으면 value 값을 대입한다. name에 value값을 저장하지는 않는다.  
* $[name:?value] : 이미 name 값이 있다면 기본값으로 하고, 값이 없으면 error를 내면서 value 값을 보여준다.  
* $[#name] : name의 문자열 길이를 반환한다 
* $[name:offset] : name 값에서 offset 만큼 삭제한 후에 값을 반환한다  
* $[name:offset:length] : name 값에서 offset 만큼 삭제한 후에 length 만큼 센 뒤 값을 반환한다  

관련 환경 변수  

* $IFS : 입력필드 구분자로서 셸 상에서 입력을 읽어 들일 때 글자를 구분하기 위한 목적으로 사용되는 문자 목록을 뜻한다. 보통은 스페이스, 탭, 개행 문자를 사용한다.  

argment 변수  
특별한 내장 변수로 '위치 매개 변수' 라고도 한다. 이 변수는 매개 변수를 불러올 때 스크립트의 명령행 인자를 담당한다. 위치 매개 변수는 그 이름이 1, 2, 3 등으로 되어 있으며, 그 값을 각각 \$1, \$2, \$3 으로 표시하고 0은 스크립트 파일명을 뜻한다  
* $# : 스크립트에 넘겨진 argment 개수  
* \$$ : 셸 스크립트의 PID  
* \$* : 스크립트에 전달된 인자 전체를 하나의 변수에 저장하면 IFS 변수의 첫 번째 문자로 구분  
* \$@ : \$* 와 동일 (IFS를 사용하지 않음)  
* $? : 실행한 뒤의 반환 값, 참이면 0, 거짓이면 1  
* $- : 현재 Shell이 호출될 때 사용한 옵션들  

Shell 변수 관련 명령어  
* set : 일반적으로는 셀 변수를 출력하는 명령
* env : 환경 변수를 출력하는 명령
* unset : 선언된 변수를 제거 
* export : 특정 변수의 범위를 환경 데이터 공간으로 전송하여 자식 프로세스에서도 특정 변수를 사용 가능하게 한다  

2. echo 문과 escape 문자   
echo 문은 라인을 제어하는 명령해석기이며 -e 옵션과 같이 ₩으로 시작하는 escape 특수 문자를 사용할 수 있다.  

escape 특수 문자의 종류  
* ₩f : 폼피드. 앞 문자열만큼 열을 밀어서 이동시킨다. 
* ₩n : 새로운 줄로 바꾼다 
* ₩r : 캐리지 리턴. 앞 문자열의 앞부분 부터 뒷 문자열 만큼 대체하고 반환한다
* ₩t : 탭만큼 띄운다 
* ₩₩ : ₩를 표기한다  

예시  
echo -e "yuloje\ralin"  
alinje  
앞 문자열의 yuloje를 받아서 앞부분 부터 alin으로 대체한 후에 반환한다  

3. 간단한 조건식  
명령행에서는 test 명령어를 이용하여 간단히 비교할 수 있고 셸 스크립트에서는 보통 if 문과 같이 사용된다. 셸 스크립트에서 사용할 경우에 test라는 명령어를 생략하고 []를 이용한다.  

사용법  
test 표현식  

* [ string ] : string이 빈 문자열이 아니라면 참 
* [ string1 = string2 ] : 두 문자열이 같아면 참  
* [ string1 != string2 ] : 두 문자열이 다르면 참  
* [ -n string ] : 문자열이 Null이 아니라면 참  
* [ -z string ] : 두 문자열이 null이면 참  

산술 비교 표현식  
* [ expr1 -eq expr2 ] : 두 표현식 값이 같다면 참 
* [ expr1 -ne expr2 ] : 두 표현식 값이 같지 않다면 참 
* [ expr1 -gt expr2 ] : expr1 > expr2 이면 참  
* [ expr1 -ge expr2 ] : expr1 >= expr2 이면 참 
* [ expr1 -lt expr2 ] : expr1 < expr2 이면 참 
* [ expr1 -le expr2 ] : expr1 <= expr2 이면 참 
* [ ! expr ] : expr이 참이면 거짓, 거짓이면 참 
* [ expr1 -a expr2 ] : expr1 AND expr2 의 결과, 즉 둘 다 참이면 참 
* [ expr1 -o expr2 ] : expr1 OR expr2 의 결과, 즉 중 하나만 참이면 참  

파일 조건 표현식  
* [ -b FLIE ] : FILE이 블록 디바이스면 참  
* [ -c FLIE ] : FILE이 문자 디바이스면 참 
* [ -d FLIE ] : FILE이 디렉터리이면 참 
* [ -e FLIE ] : FILE이 존재하면 참 
* [ -f FLIE ] : FILE이 존재하고 정규파일이면 참 
* [ -g FLIE ] : FILE에 SGID가 있으면 참  
* [ -k FLIE ] : FILE에 Sticky bit가 있으면 참   
* [ -L FLIE ] : FILE이 심볼릭 링크면 참 
* [ -p FLIE ] : FILE이 Named pipe이면 참  
* [ -r FLIE ] : 현재 사용자가 읽을 수 있는 파일이면 참 
* [ -s FLIE ] : 파일이 비어있지 않으면 참  
* [ -t FD ] : FD(File Descriptor)가 열려진 터미널이면 참  
* [ -u FLIE ] : FILE에 SUID가 있으면 참 
* [ -w FLIE ] : 현재 사용자가 쓸 수 있는 파일이면 참 
* [ -x FLIE ] : 현재 사용자가 실행할 수 있는 파일이면 참 
* [ -O FLIE ] : FILE의 소유자가 현재 사용자면 참  
* [ -G FLIE ] : FILE의 그룹이 현재 사용자의 그룹과 같으면 참  
* [ FLIE1 -nt FILE2 ] : FILE1이 FILE2보다 새로운 파일이면 참 
* [ FLIE1 -ot FILE2 ] : FILE1이 FILE2보다 오래된 파일이면 참 
* [ FLIE1 -ef FILE2 ] : FILE1이 FILE2의 하드링크, 즉 I-node 값이 같으면 참  

4. 조건문  

4-1. if 문  
참인지 거짓인지를 판단할 때 사용된다. 참이라면 then 부분을 실행하고 그렇지 않으면 else 부분을 실행한다. if로 시작하면 반드시 fi로 끝나야 한다  

4-2. case 문  
문자열과 일치하는 정규식 부분을 찾아 해당하는 정규식 다음에 있는 명령어를 실행시킨다. 보통 맨 마지막에는 *)로 표기하여 해당하지 않는 모든 경우를 처리한다.  

4-3. select 문  
다른 셸에는 없고 Korn Shell 및 Bash 에만 존재하는 구문으로 간단히 나열할 때 사용한다.  

5. 반복문  
지정된 숫자나 몇 가지 조건을 만날 때까지 일련의 명령들을 반복적으로 수행할 수 있게 한다.  
5-1. for 문  
for 문은 지정된 범위 안에서 루프를 수행한다. 범위는 어떤 집합도 가능하다. in 뒤의 값1, 값2,...가 순서대로 지정된 변수에 배정되면서 do 이하의 문장을 수행한다. 즉 매 루프를 돌 때마다 변수의 값은 in 이후의 값으로 대체된다  

5-2. while 문  
while 뒤의 조건문 동안 do 이하의 문장을 실행한다. for 문과 같이 실행 횟수를 지정하지 않아도 된다.  

5-3. until 문  
until은 whlie 문 과 동일한 효과를 내지만 조건이 반대이다. 즉, while 문은 조건이 참일 동안 루프를 수행해야 한다. until 문은 거짓일 동안 루프를 수행한다.  

6. 함수  
함수란 일종의 '스크립트 안의 스크립트' 라고 할 수 있는데, 명령어 그룹을 메모리 내에 정의하여 수행 속도를 향상시킨다.  

7. 패턴과 패턴비교  
일종의 문자열을 연산하는 것으로 특정한 패턴을 놓고 변수의 문자열 값이 일부분이라도 이 패턴과 일치하는지 검사할 때 쓰인다. 이 패턴에는 와일드카드 문자 (*, ?, []를 포함한 문자 세트)를 포함해도 된다.  

* ${variable#pattern} : 처음부터 패턴과 맞는 variable의 부분을 찾아 이 중 가장 작은 부분을 제거하고 나머지를 반환  
* ${variable##pattern} : 처음부터 패턴과 맞는 variable 부분을 찾아 가장 큰 부분을 제거하고 나머지 부분을 반환 
* ${varibale%pattern} : 끝에서부터 패턴과 일치하는 variable의 최소 부분을 제거하고 나머지를 반환 
* ${variable%%pattern} : 끝에서부터 패턴과 일치하는 variable의 최대 부분을 제거하고 나머지를 반환  

##### 셸 프로그래밍 관련 명령어  
1. read  
임의의 값을 키보드로부터 입력받는 명령어로서 연속으로 여러 값을 받을 수도 있다. 인자값은 차례대로 처리된다.  

사용법  
read 변수1  [변수2,...]  

예시  
read flower  
rose  
echo $flower  
rose  

2. echo  
해당 문자열을 표준 출력에 표시하고 행 바꿈을 한다. 만일 문자열이 없으면 행 바꿈만 실행  

사용법  
echo [option] [문자열]  

* -n : 출력 결과에서 행 바꿈을 하지 않는다. 
* -e : escape 문자를 사용할 수 있게 해준다  

3. break  
for, while, until 등의 순환문에서 빠져 나온다.  

사용법  
break [n] : 순환문의 n 번째 루프에서 빠져나온다.  

4. continue  
for, while, until 등의 순환문을 계속해서 실행한다.  

사용법  
continue [n]  

5. exit  
상태 코드 n과 함께 스크립트를 종료한다. n은 0(성공)이나 0이 아닌 숫자(실패)가 될 수 있다. 만일 n이 주어지지 않으면 가장 최근에 사용된 명령의 상태 코드를 사용한다.  

6. function  
함수를 정의한다. 위치 매개 변수 ($1, $2, -)가 명령어 부분에서 사용 가능하다.  

사용법  
function 함수명  

7. getopts  
명령행 인자를 처리해 올바른지 점검한다. getopts 명령은 주로 셀 스크립트 순환문에서 사용되며 명령행 옵션이 표준 형식에 맞는지를 확인한다. 문자열에는 옵션을 나타내는 문자들이 포함되며 스크립트가 실행될 때 getopts에 의해 처리한다. 적합한 옵션들은 하나씩 처리되어 지정된 변수명에 저장된다.  

사용법  
getopts 문자열 변수명  

8. return  
함수 정의에서 사용되며 함수가 종료될 때 상태 코드 n을 반환한다. 만일 n이 생략되면 직전에 실행된 명령 상태 코드를 반환한다. 

사용법  
return [n]  

9. set 
특별한 옵션 없이 set이라고 입력하면 셸 변수와 값을 출력한다 set은 셸에서 보통 벼수와 값을 선언할 때 주로 사용하는데, c-shell 에서는 반드시 사용해야 하지만 bash에서는 생략해서 사용한다. 추가 기능으로는 셸에서 매개인자를 설정하고 다양한 셸 환경 설정할 때도 사용한다  

사용법  
set [option] [문자열]  

* -f : 셀에서 *,?와 같은 와일드카드 문자를 이용한 확장을 없앤다. -o noglob 와 같다 
* -C : 리다이렉션 기호인 >를 이용한 파일의 덮어쓰기를 제한한다. -o noclobber 와 같다 
* -o : bash 환경 설정 관련 값을 확인하고 설정하는 옵션이다. noglob, noclobber 등등  

10. shift  
위치 매개 변수를 이동할 때 사용하는 명령이다  

사용법  
shift [n]  
위치 매개 변수를 n 만큼 이동하는데, 만약 n이 생략되면 기본적으로 1이 적용된다. 즉 \$2는 \$1으로 이동한다. 

11. source  
지정한 텍스트 파일의 내용을 읽고 실행한다. 점(.)과 source는 같은 기능을 한다  

사용법  
source 파일 [인수]  
. 파일 [인수]  

12. eval  
인자의 값을 구하는데 사용하고 만약 값이 실행 명령어와 같은 문자열인 경우에는 해당 명령을 실행시킨다  

13. expr  
인자로 사용되는 표현식의 값을 구할 때 사용하는데 산술 연산이 가능하다. 셸 상태에서 명령어로도 사용가능하다. 다만 셸 상태에서는 연산자 사용 시 * 등을 사용할 경우 앞에 ₩를 붙여서 사용해야 하고, 값과 연산자 사이는 반드시 한 칸 띄워야 한다  

사용법  
expr 값 연산자 값

최근에 등장한 Shell 에서만 사용 가능한 명령어로 형식화된 출력을 할 때 사용한다.  

* %d : 십진수 출력 
* %c : 한문자 출력 
* %s : 문장 출력

16. type

사용법  
type [option] 인자

### 2.5 프로세스 
##### 2.5.1 프로세스의 개념 및 종류  

프로세스의 정의  
실행중인 프로그램을 프로세스라 하고, 실행시에 번호(PID)가 할당되어 관리된다  

* 백그라운드 : 사용자의 입력에 관계없이 실행  

1. 실행시키려면 기존 포어그라운드 프로세스 실행 명령뒤에 &를 붙임
2. 그림에서 [1]은 작업번호, 5636은 PID

* 포어그라운드 : 명령 입력 후 수행종료까지 기다려야 하는 프로세스  

프로세스의 생성  
* fork  
새로운 프로세스를 위해 메모리를 할당받아 복사본 형태의 프로세스를 실행하는 형태로 기존의 프로세스는 그대로 실행되어 있다. 새롭게 생성된 프로세스는 원래의 프로세스랑 똑같은 코드를 기반으로 실행된다.  
* exec  
원래의 프로세스를 새로운 프로세스로 대체하는 형태로 호출한 프로세스의 메모리에 새로운 프로세스의 코드를 덮어씌운다. 

1. 리눅스가 부팅을 시작하면 커널이 systemd라는 최초의 프로세스를 발생시킨다 (PID 1)  
2. 이후 시스템 운영에 필요한 데몬을 비롯한 다른 프로세스들은 fork 방식으로 systemd 프로세스의 자식 프로세스로 생성 
3. pstree -h 명령으로 확인

* ps -l : fork 기반의 프로세스 생성 확인 (PPID가 bash의 PID)  


멀티태스킹과 작업 전환  
* 멀티태스킹 : 동시에 하나 이상의 프로그램을 실행하는 것  
* 사용자가 실행한 프로세스는 중간에 중지시킬 수 있다.
* 포어그라운드에서 백그라운드로, 백그라운드에서 포어그라운드로 전환할 수 있다.  

1. 포어그라운드 프로세스 -> 백그라운드 프로세스  

1-1. [Ctrl] + [z] 로 작업중인 프로세스를 일시 중지.(서스팬드) 작업의 상태는 'jobs'라는 명령, 'Stopped'로 표시  

1-2. 일시적으로 중단된 상태에서 'bg'라는 명령을 입력  

2. 백그라운드 프로세스 -> 포어그라운드 프로세스  

2-1. 'fg' 명령어  
다수의 작업이 백그라운로 수행중이라면 'jobs'로 내용 확인 후 'fg 작업번호' 또는 'fg % 작업번호' 명령으로 원하는 작업을 전환할 수 있음  

jobs 명령을 실행하면 +,-가 표시됨  

* \+ : 주로 처리되는 프로세스, 보통 가장 늦게 실행한 프로세스. 이땐 작업번호 없이 fg만 수행  
* \- : \+ 작업의 다음 우선순위의 작업

##### 3.5.2 프로세스 관리의 이해  

1. IPC  
프로세스들 사이에 서로 데이터를 주고받는 행위 또는 그에 대한 방법이나 경로를 통칭하는 말. IPC 방식에는 파일, 신호, 소켓, 공유 메모리, 메시지 큐, 세마포, 파이프, D-Bus 등이 존재  
D-Bus : 같은 머신에서 동시에 실행 중인 여러 컴퓨터 프로그램 간의 통신이 가능하도록 지원하는 소프트웨어 버스  

2. 시그널  
리눅스에서는 프로세스끼리 서로 통신할 때 사용한다. 즉, 특정 프로세스가 다른 프로세스에게 메시지를 보낼 때 시그널을 이용한다  

* 번호=1, SIGHUP(HUP) : 로그아웃과 같이 터미널에서 접속이 끊겼을 때 보내지는 시그널  
* 번호=2, SIGINT(INT) : 키보드로부터 오는 인터럽트 시그널로 실행을 중지시킨다  
* 번호=3, SIGQUIT(QUIT) : 키보드로부터 오는 중지 시그널 
* 번호=9, SIGKILL(KILL) : 프로세스를 강제 종료 시키는 시그널  
* 번호=15, SIGTERM(TERM) : 가능한 정상 종료시키는 시그널  
* 번호=18, SIGCONT(CONT) : 시그널 등에 의해 정지된 프로세스를 다시 실행
* 번호=19, SIGSTOP(STOP) : 터미널에서 입력된 정지 시그널 
* 번호=20, SIGTSTP(TSTP) : 프로세스를 계속 실행해야 하는 상황에서 해당 프로세스를 일시 중지시키기 위해 사용되는 시그널  

3. 데몬  
주기적이고 지속적인 서비스 요청을 처리하기 위해 계속 실행되는 프로세스로 백그라운드로 실행된다 리눅스에서는 서버 역할을 하는 프로그램들이 해당하고 보통 이름뒤에 d를 붙임  

* standalone : 단독 데몬 방식. 데몬이 독자적으로 메모리에 상주 및 제거되는 형태 
* inetd : 관련 프로세스가 메모리에 항상 상주하는 것이 아니라 클라이언트의 서비스 요청이 들어왔을 때 관련 프로세스를 실행시키고 접속 종료 후에는 자동으로 프로세스를 종료시키는 방식  

4. 데몬의 실행  

* RHEL 6 이전 : System V 계열의 init 스크립트 사용
* RHEL 7 이후 : systemd 기반으로 동작  

5. 데몬 관련 유틸리티  

* ntsysv : 텍스트 환경에서 커서를 이용하여 부팅 중 자동으로 실행되는 서비스를 설정할 수 있는 유틸리티  

사용법  
ntsysv [option]  

* --level 실행레벨 : 특정 실행레벨을 편집한다.
