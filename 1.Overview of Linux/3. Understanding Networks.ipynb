### 3.1 네트워크 기초 
##### 3.1.1 네트워크 개요와 통신망의 종류  

네트워크의 정의  
컴퓨터를 상호 연결하여 정보 교환 및 처리, 자료 및 하드웨어 고유 등을 위해 구성한 통신망을 말한다. 컴퓨터 네트워크는 규모에 따라 근거리 통신망(LAN), 도시권 통신망(MAN), 광역 통신망(WAN) 등으로 구성한다.  

1. LAN  
가까운 지역을 네트워크 매체를 이용하여 하나로 묶는 근거리 통신망을 말한다. 현재는 이더넷을 주로 사용하고, 무선 방식의 발전으로 무선 랜도 보편적으로 사용하고 있다.  

* 스타형  
중앙에 위치한 중앙 컴퓨터가 각각의 컴퓨터와 통신하는 방식으로 중앙의 제어기를 중심으로 모든 기기는 Point-to-Point 방식으로 연결한다.  

장점: 고속의 대규모 네트워크에 이용하고, 일부 장애가 발생해도 전체 네트워크에 영향을 주지 않고 단말기 추가나 오류진단이 용이하다.  
단점: 중앙 컴퓨터 고장시에는 전체 네트워크가 중단되고 회선의 수가 증가하면 제어가 복잡해진다  

* 버스형  
하나의 통신 회선에 여러 컴퓨터를 연결해서 전송하는 방법으로 모든 장치들은 동등한 조건으로 경쟁한다. 한번에 한 컴퓨터만 전송할 수 있어서, 연결된 컴퓨터의 수에 따라 네트워크 성능이 좌우되며 신호 반사에 의한 상호 간섭을 막기 위해 종단에는 종단기가 존재한다. 전송 방식에는 CSMA/CD와 토큰패싱을 이용  

장점: 구조가 간단하여 단말기의 추가 및 제거가 용이하고 설치비용이 저렴하다.  
단점: 장애 발생 시에 발생지의 위치 추적이 어렵고, 버스의 연결 부위나 종단 장치에 문제가 발생하면 전체 네트워크가 중단된다.  

* 링형  
원형의 통신 회선에 컴퓨터와 단말기를 연결하는 형태로 연결된 앞의 컴퓨터로부터 수신한 내용을 다음 컴퓨터에 재전송하는 방식으로 동작하며, 토큰 패싱이라는 방법을 통해 데이터를 전송한다.  

장점: 네트워크 전송상의 충돌이 없고, 노드의 숫자가 늘더라도 전체적인 성능의 저하가 적다.  
단점: 노드의 추가가 비교적 어렵고, 노드에 문제가 발생했을 경우에 전체 네트워크가 중단될 수 있다.  

* 망형  
스타형과 링형의 혼합된 형태로 각 네트워크 장비가 여러 개의 인터페이스를 갖추고, 상호간에 그물 형태로 연결하는 방식이다. LAN을 구성할 때보다 라우터를 이용하여 LAN과 LAN을 연결하거나 백본망을 구성할 때 주로 사용된다.  

장점: 장애 발생 시에도 다른 시스템에 영향이 적고, 우회할 수 있는 방법이 존재하여 신뢰성이 높다.  
단점: 설치비용이 많이 들고, 운영이 어려우며 장애 발생 시에 고장 지점을 찾기가 쉽지 않다.  

2. LAN 전송 방식  
* 이더넷과 CSMA/CD  
이더넷은 LAN을 위해 개발된 네트워크 기술로 각 기기들이 48비트 길이의 고유한 MAC 주소를 기반으로 상호간에 데이터를 주고받을 수 있도록 만들어졌다. 전송매체로는 동축, UTP, STP 케이블을 이용하며, 각각의 기기를 상호 연결시키기 위해 허브,스위치,리피터 등의 장치를 이용한다. 이더넷은 CSMA/CD 방식을 이용하는데, 단말기가 전송로의 신호유무를 조사하고 다른 단말기가 신호를 송출하고 있는지를 확인한다.  

장점: 저렴함  
단점: 토큰 링 방식에 비해 효율성은 떨어진다.  

* 토큰링  
1980년대 초반 IBM에 의해 개발된 LAN 기술로 토큰링 네트워크는 여러 스테이션을 하나의 링에 이어져 형성되고 데이터는 한쪽 방향으로 흐르도록 하여 한 컴퓨터에서 다음 컴퓨터 순으로 전달된다. 특히 데이터 전송을 위한 네트워크 채널을 사용하려면 토큰이라는 일종의 사용권을 획득해야 한다.  

* FDDI  
전송 매체를 광섬유 케이블을 사용하여 설계된 링 구조의 통신망으로 네트워크 액세스를 제어하기 위해 토큰 패싱 방법을 사용한다. 한 빌딩 내의 간선 LAN으로 사용되거나 컴퓨터 사이를 고속으로 접속하는데 주로 사용한다. 간선 LAN의 경우에는 FDDI를 갖춘 라우터나 브리지, 집선장치로 링형 LAN을 구성한다. 보통 이중 링형을 구성하여 통신망이 한꺼번에 단절되는 경우를 방지한다.  

* MAN  
LAN과 WAN의 중간 형태로 위성 도시 등을 연결한 형태를 말한다. WAN 보다 규모는 작지만 높은 데이터 전송률을 갖는다. LAN보다 장거리이고 고석이며, 음성과 데이터 모두 전송이 가능하다. MAN은 전화선을 통해 서로 연결된 여러 가지 LAN을 포함하는 대도시 통신망이라고 할 수 있다.  DADB는 광케이블을 이용해서 두 개의 단방향 선로를 사용한다.  

* WAN  
넓은 지역을 연결하는 네트워크이다. 거리상의 제약이 없지만, 다양한 경로를 경유하여 도달하므로 LAN보다 속도가 느리고, 전송 에러율도 높은 편이다. WAN을 구성하는 방법에는 전용 회선과 교환 회선 방식이 있다.  

1. 전용회선 방식: 회선 제공하는 통신 사업자가 계약 체결한 송수신 사용자끼리만 전용 통신 선로로 연결하여 데이터를 교환하는 방식으로 안전성이 우수하나 높은 비용이 소요된다.  

2. 교환회선 방식: 공중망을 이용해 전송하는 방식이다 전용회선 방식에 비해 비용이 저렴하지만, 다수의 사용자가 선로를 공유해서 사용하므로 데이터 전송 속도가 상대적으로 느리다.  

2-1. 회선 교환 방식: 송신 호스트와 수신 호스트가 데이터를 전송하기 전에 연결 경로를 설정하여 일종의 물리적 전용선을 연결하는 방식으로 송수신 호스트간의 경로가 결정되기 때문에 모든 데이터는 같은 경로로 전달된다. 고정된 대역폭을 할당 받아 전송된다.  

2-2. 패킷 교환 방식: 패킷이라고 부르는 정보 블록을 이용해서 전송하는 방식이다. 패킷이란 전달하고자 하는 정보를 일정 크기로 분할하고 각각에 송수신 주소를 비롯한 관련 정보를 추가하여 만든 데이터 블록이다. 전송 대역폭을 동적인 방식으롤 공유하고, 고정대역을 할당하지 않는다. 단점으로는 더 많은 지연이 발생하고 각 패킷마다 오버헤드 비트가 존재한다.  

2-2-1. 데이터그램 방식: 비연결형 서비스를 이용해 전달되는 패킷들이 독립적으로 전송된다. 연결을 설정하기 위한 과정이 없고, 경로를 미리 할당하지 않는다. 전송할 정보의 양이 적거나 상대적으로 신뢰성이 중요하지 않는 환경에서 사용한다.  

2-2-2. 가상 회선 방식: 연결형 서비스를 지원하기 위한 기능으로 송수신 호스트 사이에 설정된 가상의 단일 파이프를 통해 패킷을 송수신한다. 패킷이 도착하는 순서는 전송한 순서와 같게 된다.  

2-3. 관련 프로토콜  
2-3-1. X.25 : 패킷망에서 DCE, DTE, 회선 사이에 이루어지는 상호 작용을 규정한 프로토콜  

2-3-2. 프레임 릴레이 : LAN또는 WAN 상에서 단말 지점간의 효율적인 데이터 전송을 위해 고안된 방법으로 패킷 대신에 프레임이라고 불리는 가변 길이 단위에 데이터를 넣고 재전송과 같은 오류정정기능은 단말 지점에 맡기는 방식이다.  

2-3-3. 셀 릴레이(ATM) : 셀이라고 부르는 고정 길이 패킷을 이용하여 자료를 전송하는 방식.  



##### 3.1.2 네트워크 장비  

* LAN 구성 장비  
1. 네트워크 카드  
네트워크 안에서 컴퓨터끼리 통신할때 사용하는 하드웨어로서 네트워크 어댑터, 네트워크 인터페이스 카드라고도 부른다.  

2. 케이블  
케이블은 신호를 전송하게 되는 물리적인 전송 매체로서, 이더넷을 구성하는 케이블에는 BNC, UTP, 광섬유 등이 있다.  

2-1. BNC 케이블  
보통 동축케이블 이라고 부르는데, 가정에 들어오는 케이블 TV의 선이나 안테나선을 생각하면 된다.  

2-2. UTP 케이블  
트위스티드 페어 케이블은 두 개의 선이 서로 꼬여진 형태로 만들어 졌는데 그 이유는 전자신호가 변형되는 것을 줄이기 위해서이다. STP,UTP로 나눌 수 있는데, 피복 사이에 차폐물을 입힌것이 STP, 입히지 않은 것이 UTP이다.  

2-3. 광케이블  
매우 가는 유리섬유나 플라스틱 섬유 등을 이요해 정보를 보낸다.  

3. 허브  
이더넷 네트워크에서 여러 대의 컴퓨터 및 네트워크 장비를 연결하는 장치이다. 하나의 허브를 중심으로 여러 대의 컴퓨터와 네트워크 장비가 마치 별모양으로 연결되며, 같은 허브에 연결된 컴퓨터와 네트워크 장비는 모두 상호 간에 통신할 수 있게 된다.  

* 인터네트워킹 장비  
네트워크와 네트워크의 연결을 인터네트워킹이라고 하는데 이러한 기능을 수행하는 장치를 일반적으로 게이트웨이라 부른다. 게이트웨이는 서로 다른 통신망과 프로토콜을 사용하는 네트워크 간에 통신 기능을 가능하게 해주는 장치를 뜻한다.  

1. 리피터  
물리적 신호는 전송 거리가 멀면 감쇄되기 때문에 중간에 이를 보완해야 하는데, 입력된 신호를 증폭하여 중계하는 장치가 리피터이다. OSI 모델의 첫 번째 계층인 물리 계층에서 동작한다.  

2. 브리지  
OSI 모델의 데이터 링크 계층에 있는 여러 개의 네트워크 세그먼트를 연결한다. 브리지는 물리 계층에 있는 리피터나 허브의 기능을 하면서, 특정 네트워크로부터 오는 통신량을 조정하여 트래픽을 관리할 수 있다.  

3. 라우터  
OSI 모델의 물리, 데이터 링크, 네트워크 계층의 기능을 지원하는 장치이다. 브리지 기능에 추가하여 네트워크 계층에 대한 경로 선택 기능을 제공해야 하고, 임의의 네트워크에 들어온 데이터를 어느 네트워크로 전달할지 판단할 수 있어야 한다. 이를 지원하기 위해 라우터는 자신과 연결된 네트워크 및 호스트 정보를 유지하고 관리해야 하며, 어떤 경로를 이용해야 빠르게 전송할 수 있는지를 판단해야 한다.  

##### 3.1.3 프로토콜  

개요  
컴퓨터 상호간에 통신을 위해 미리 정해놓은 특별한 통신규칙이자 통신규약. 대표적인 프로토콜로 현재 인터넷에 사용되고 있는 TCP/IP가 있다.  

프로토콜의 기본 구성요소 3가지.  
1. 구문: 데이터의 구조나 형식을 말하는 것으로 부호화, 신호 레벨 등을 규정  
2. 의미: 전송의 조작이나 오류 제어를 위한 제어 정보에 대한 규정  
3. 순서: 접속된 개체 간의 통신 속도의 조정이나 메시지의 순서 제어  

프로토콜의 기능  
1. 주소지정: 하나의 개체에서 다른 개체에 데이터를 전송할 경우에는 상대의 주소를 알아야 한다.  
2. 순서지정: 프로토콜 데이터 단위의 통신이 이루어질 때, 데이터의 전송순서를 명시하는 기능.  
3. 단편화와 재조합: 개체 사이에서 대용량의 데이터를 교환하는 프로토콜의 경우에 블록을 분할 전송하고, 수신측에서 재조합하는 기능  
4. 흐름제어: 송신속도가 수신측의 데이터 처리능력을 초과하지 않도록 하는 것.  
5. 오류제어: 데이터 교환시에 발생하는 오류를 발견하는 것으로 체크섬 기능이 대표적.  
6. 연결제어: 프로토콜의 연결 설정에서의 구문, 의미, 시간을 제어하는 것을 말한다.  
7. 동기화: 양 개체간의 데이터 전송 시 타이머값, 윈도우 크기 등의 인자값을 일치시키는 것.  
8. 멀티플렉싱: 하나의 통신 선로에서 다중 시스템이 동시에 통신할 수 있는 기법.  
9. 전송 서비스: 우선순위 결정, 서비스 등급과 보안 요구 등의 제어 서비스.  
10. 캡슐화: 송신자와 수신자의 주소, 오류검출 코드, 프로토콜 제어 정보를 덧붙이는 것.  

프로토콜 제정기관  
ISO, IEEE, ANSI, TIA, ITU-T  

TCP/IP  
1. 역할  
컴퓨터 기종에 관계없이 정보 교환이 가능하게 해주는 통신 프로토콜이다.  

2. 구조

##### 3.1.4 IP 주소와 도메인  

* IP 주소  
인터넷에서 다른 컴퓨터와 서로 인식하고 통신하기 위해 사용되는 특수한 번호로 각 컴퓨터마다 고유한 값으로 제공된다. 전체 IP 주소는 0.0.0.0 ~ 255.255.255.255 사이의 값을 가지게 된다.  

* IP 주소의 클래스  
1. A 클래스 (N.H.H.H)  
IP 주소의 첫 번째 부분의 1bit가 '0'이며, 네트워크 ID는 0~127  

2. B 클래스 (N.N.H.H)  
IP 주소의 첫 번째 부분의 2bit가 '10'이며, 네트워크 ID는 128.0 ~ 191.255  

3. C 클래스 (N.N.N.H)  
IP 주소의 첫 번째 부분의 3bit가 '110'이며, 네트워크 ID는 192.0.0 ~ 223.255.255  

4. D 클래스  
IP 주소의 첫 번째 부분의 4bit가 '1110'이며, 다중 방송 통신용으로 예약되어 있다.  

5. E 클래스  
IP 주소의 첫 번째 부분의 4bit가 '1111'이며, 연구용으로 예약되어 있다. 

* 도메인네임  
숫자로 표현된 IP 주소를 알기 쉽게 영문으로 표기한 것으로 2~5개 영역으로 나누어진다.  

* 도메인네임 시스템 (DNS)  인터넷상에서 컴퓨터 이름에 대한 주소 매핑 서비스를 제공하기 위해 개발되었다.  

* 도메인 체계  
1. com : 영리를 목적으로 하는 기업이나 회사  
2. net : 네트워크와 관련된 기관 
3. org : 비영리 목적의 기관이나 단체 
4. edu : 학위를 수여하는 교육기관 
5. gov : 미국 연방정부기관
6. mil : 미국 연방군사기관
7. int : 유앤등과 같은 국제기관  

* ICANN  
국제도메인관리기구로 부르는데, 비영리기관으로 인터넷 도메인 관리와 정책을 결정하는 국제기구  

* IPv6  
IPv4의 주소 고갈 문제를 해결하기 위해 제안된 차세대 인터넷 프로토콜이다. 주소 범위는 128비트이고, 주소 표현은 16비트 단위로 하여 8자리로 나타낸다.  


### 3.2 네트워크 환경  
##### 3.2.1 환경 설정  

네트워크 인터페이스 설정 방법  

* nmcli 명령  

사용법  
nmcli [option] [command]  

* -v: nmcli 도구의 버전을 출력  

command에 존재하는 명령  

1. g(general)명령  
* status : 네트워크 매니저의 상태 정보를 출력한다. (디폴트값)
* hostname : 호스트명을 설정하거나 확인할 때 사용  
* permissions : 네트워크 매니저의 허가권 정보를 출력  

2. n(networking)명령  
* on/off : 네트워크 매니저에 의한 네트워크 제어를 on/off 한다.  
* connectivity : 네트워크 연결 상태를 검사하는 명령이다.  

3. c(connection)명령  
* show : 네트워크 장치의 기본 프로필 정보를 출력한다. 
* up/down : 네트워크 장치를 활성/비활성한다. 장치명 이외에 uuid 등도 사용할 수 있다.  


##### 3.2.2 네트워크 관련 명령어  

* ip : 이더넷 장치, IP 주소, 라우팅 정보 등의 설정 정보를 출력하거나 변경하는 명령어  
사용법  
ip [options] 대상 [command]  

[주요 대상]  
address : IPv4, IPv6 주소를 나타낸다  
route : 라우팅 테이블 목록을 나타낸다  
link : 네트워크 장치를 의미한다  

[주요 command]  
show, list : 대상에 대한 정보를 출력한다.(디폴트)  
add : 대상에 대한 특정 정보를 추가하거나 설정  
delete, del : 대상에 대한 특정 정보를 삭제  

* ifconfig : 네트워크 인터페이스를 설정하거나 확인하는 명령 IP주소, 넷마스크 주소, MAC 주소등을 설정하고 확인할 수 있음  

사용법  
ifconfig [interface] [address] [option]  

1. flags : 네트워크 인터페이스의 상태 정보를 나타낸다.  
1-1. UP : 인터페이스가 활성화되어 있음  
1-2. BROADCAST : 브로드캐스트를 사용함  
1-3. RUNNING : 인터페이스가 동작중임을 나타냄  
1-4. MULTICAST : 멀티캐스트를 사용함  
1-5. mtu : 한번에 전송할 수 있는 최대 패킷의 크기를 나타냄  
1-6. inet : IPv4 주소를 나타냄  
1-7. netmask : 넷마스크 주소값을 나타냄  
1-8. broadcast : 브로드캐스트 주소를 나타냄  
1-9. inet6 : IPv6 주소를 나타냄  
1-10. ether : 이더넷 카드에 설정되어 있는 48비트 형식의 고유한 하드웨어 주소 보통 MAC으로 부름  
1-11. RX/TX : 수신, 송신 패킷 수를 나타냄  
1-12. collisions : 충돌이 발생한 패킷 수를 나타냄  
--------------------------------------------------------------------  
* route : 라우팅 테이블의 정보를 출력하거나 관리하는 명령  

사용법  
route [add|del] [-destination] [netmask 값] [gw 값] [dev 인터페이스]  

항목  
1. Destination : 목적지 네트워크 주소를 나타냄  
2. Gateway : 목적지 주소로 보내기 위한 게이트웨이 주소를 가리킴  
3. Genmask : 목적지의 네트워크 주소를 알아내기 위한 넷마스크  
4. Flags : 목적지 네트워크와의 연결 상태를 준다.  
4-1. U : 인터페이스가 사용되고 있음  
4-2. G : 라우트가 게이트웨이로 사용되고 있음을 나타낸다.  
4-3. H : 오직 단독 호스트만이 라우트를 거쳐서 접근할 수 있다.  
4-4. D : 테이블 엔트리가 설정된 경우, ICMP 리다이렉트 메시지로 운영된다.  
4-5. M : 테이블 엔트리가 설정된 경우, ICMP 리다이렉트 메시지에 의해 수정되고 있음을 알림  
5. Metric : 목적지 네트워크와의 거리를 나타낸다. 직접 연결되어 있으면 0이면 라우터를 하나 거칠 때마다 1씩 증가한다.  
6. Ref : 다른 라우터가 목적지로 가기 위해 참조한 횟수를 나타낸다.  
7. Use : 다른 라우터가 목적지로 가기 위해 사용한 횟수를 나타낸다.  
8. Iface : 목적지 네트워크와 연결된 인터페이스를 나타낸다.  
--------------------------------------------------------------------  
* netstat : 네트워크의 연결 상태를 출력하는 명령이다. 네트워크 연결 상태 이외에도 라우팅 테이블 정보, 네트워크 인터페이스 상태, 매스커레이드 연결 상태, 멀티캐스트 멤버 등의 정보를 출력한다.  

사용법  
netstat [option] [address_family_option]  

[option]
* -a : 모든 소켓 정보를 출력한다.  
* -n : 기호화된 호스트명이나 포트명 대신에 숫자값으로 표시한다.  
* -p : 소켓에 대한 PID와 프로그래밍을 출력한다.  
* -r : 라우팅 테이블 정보를 출력한다.  
* -I : 대기하고 있는 포트인 리슨 포트를 출력한다. 
* -i : 네트워크 인터페이스 테이블 정보를 출력한다. 
* -s : 네트워크 프로토콜에 대한 통계 정보를 출력한다. 
* -c : 네트워크 정보를 계속해서 출력한다. 
* -t : TCP 프로토콜 기반 정보만 출력한다. 
* -u : UDP 프로토콜 기반 정보만 출력한다. 
* -g : 멀티캐스트 그룹 멤버의 정보를 출력한다.  

[address_family_option]  
* --protocol=값 : 특정 프로토콜과 연관된 정보를 출력한다. 
* --inet, --ip : IP 주소 기반으로 연결된 정보를 출력한다.  
* --unix : UNIX Domain Socket에 대한 정보를 출력한다.  

[netstat의 State 결과]  
* LISTEN : 서버에서 클라이언트로부터 들어오는 패킷을 위해 소켓을 열고 기다리는 상태이다.  
* SYS_SENT : 로컬 시스템의 클라이언트 애플리케이션이 원격 호스트에 연결을 요청한 상태이다.  
* SYN_RECV : 서버 시스템이 원격 클라이언트로부터 접속 요구를 받아 클라이언트에게 응답을 하였지만, 아직 클라이언트에게 확인 메시지를 받지 못한 상태이다. 
* ESTABLISHED : 3Way-Handshaking이 완료된 후 서버와 클라이언트가 서로 연결된 상태  
* FIN_WAIT1 : 소켓이 닫히고 연결이 종료되는 상태  
* FIN_WAIT2 : 연결은 종료된 상태이고, 소켓이 원격 호스트로부터 종료되었다는 정보를 기다리는 상태이다.  
* CLOSE_WAIT : 원격 호스트는 종료된 상태이고 소켓을 종료하기 위해 기다리는 상태  
* CLOSING : 흔하지 않지만 주로 확인 메시지가 전송도중 분실된 상태  
* TIME-WAIT : 패킷 처리는 끝났지만 분실되었을지 모를 느린 세그먼트를 위해 당분간 소켓을 열어놓은 상태  
* CLOSED : 소켓 연결이 종료된 상태  
* LAST_ACK : 원격 호스트가 종료되고 소켓도 닫힌 상태에서 마지막 ACK 패킷을 기다리는 상태.  
* UNKNOWN : 소켓 상태를 확인할 수 없는 상태이다.  
--------------------------------------------------------------------  
* ss : 소켓 상태를 출력하는 명령으로 netstat 명령과 유사  

사용법  
ss [option] [ 필터 ]  

* -a : 소켓의 모든 정보를 출력 
* -n : 서비스명으로 리졸빙하지 않고 숫자값으로 출력한다.  
* -r : 숫자값 형태의 주소나 포트 정보를 호스트명이나 서비스명으로 리졸빙해서 출력한다.  
* -I : 서비스 연결을 위해 대기 중인 소켓 정보를 출력  
* -o : 시간 관련 정보를 출력  
* -e : 소켓 정보를 확장해서 자세히 출력한다.  
* -p : 소켓을 사용중인 프로세스 정보를 출력한다.  
* -i : 시스템 내부의 TCP 정보를 출력  
* -s : 소켓 관련 상태 정보를 요약해서 출력한다.  
* -4 : IPv4 관련 정보만 출력한다. 
* -6 : IPv6 관련 정보만 출력한다.  
* -O : PACKET 소켓 정보를 출력한다.  
* -t : TCP 관련 소켓 정보만 출력한다.  
* -u : UDP 관련 소켓 정보만 출력한다.  
* -x : 유닉스 도메인 소켓 정보만 출력한다.  
* -f 패밀리명 : 소켓의 특정 타입을 지정해서 출력한다.  
--------------------------------------------------------------------  
* ifstat : 네트워크 인터페이스의 상태 정보를 출력하는 명령  

사용법  
ifstat [options]  

* -d : 지정한 초마다 발생하는 통계 정보를 샘플링한다.  
* -e : 에러 정보를 출력한다.  
* -a : 히스토리 파일 정보를 무시한다.  
--------------------------------------------------------------------  
* arp : ARP 캐시를 관리하는 명령이다. 

사용법  
arp [option]  

* -a : ARP 캐시에 기록된 정보 중에 특정 호스트에 대한 정보를 출력한다. 
* -d : ARP 캐시에서 해당 호스트 정보를 제거한다. 
* -n : 출력 정보를 도메인 주소 대신에 IP 주소로 출력한다. 
* -v : 관련 정보를 자세하게 출력한다.  
--------------------------------------------------------------------  
* mii-too : 네트워크 인터페이스의 상태를 점검하고 설정하는 유틸리티  

사용법  
mii-too [option] 네트워크_장치명  

* -r : 네트워크 인터페이스 자동 인식을 위해 재시작한다.  
* -F : 강제로 설정한다.  
* -v : 관련 정보를 자세하게 출력한다.  
--------------------------------------------------------------------  
* ethtool : 이더넷 카드 설정 정보를 출력하거나 변경하는 명령이다.  

사용법  
ethtool [option] 네트워크_장치명  

* -s : 이더넷 카드의 설정을 변경할 때 사용한다.  
* -P : 이더넷 카드의 하드웨어 주소를 출력한다. 
--------------------------------------------------------------------  
* ping : 인터넷 제어 메시지 프로토콜인 ICMP를 이용하여 네트워크 연결을 확인할 수 있는 명령  

사용법  
ping [option] Hostname 또는 IP_Address  

* -c 개수 : 몇 개의 패킷을 보낼 것인지를 설정 
* -i 초 : 몇 초 간격으로 패킷을 보낼 것인지를 설정  
* -s 패킷크기 : 전송하는 패킷의 크기를 바이트 단위로 설정  
* -w : 패킷을 보내고 몇 초 후에 실행을 멈출 것인지를 설정한다. 
--------------------------------------------------------------------  
* traceroute : 패킷이 특정 호스트까지 라우팅되는 과정을 출력하는 명령으로 라우팅과정에 어떤 장애가 있는 경우에 위치를 파악할 수 있다.  

사용법  
traceroute Hostname 또는 IP_Address  

--------------------------------------------------------------------  
* nslookup : DNS를 이용하여 도메인이나 IP 주소를 조회하는 명령이다. 셸 상에서 직접 조회할 수도 있고, nslookup 명령 실행 후에 대화형으로 조회할 수도 있다.  

사용법  
nslookup [-option] Hostname 또는 IP_Address  

--------------------------------------------------------------------  
* -type='서버타입' : A=IP주소 및 도메인, mx=메일서버, ns=네임서버, txt=TXT레코드  

[대화형 모드 주요 명령어]
* 호스트명 : 보통 도메인명이나 IP 주소를 입력하면 기본 설정된 네임 서버를 이용해서 관련 정보를 출력한다. 참고로 DNS 서버 설정의 A 또는 PTR에 대한 질의를 진행한다.  
* server : 도메인 또는 IP 주소를 조회하는 네임 서버를 변경한다.  
* set 키워드 [=값] : all,type 등의 키워드를 이용해서 조회에 영향을 주는 상태 정보를 변경한다. 
* exit : nslookup 의 대화형 모드를 종료한다.  

[대화형 set 명령어 주요 키워드]
* all : 자주 사용하는 옵션의 현재값과 현재 기본 네임 서버 및 호스트에 대한 정보 출력  
* type=값 : 기본값은 A고 mx=메일서버, ns=네임서버, txt=TXT 레코드  
--------------------------------------------------------------------  
* dig : 도메인명으로 정보를 조회하는 명령  

사용법  
dig [option] 도메인명 [type]  

* -t : 질의 타입을 지정하는 옵션으로 기본값은 IP주소를 조회하는 A
--------------------------------------------------------------------  
* host: 도메인명으로 정보를 조회하는 명령  

사용법  
host [option] 도메인명  

* -t : 질의 타입을 지정하는 옵션으로 기본값은 IP를 조회하는 A 
* -v : 지정한 도메인에 대한 자세한 정보를 출력  
* -a : -v 옵션과 같지만 MX,NS,TXT 등의 타입값 위주로 출력  
--------------------------------------------------------------------  
* hostname : 시스템에 설정된 호스트네임을 출력하거나 설정하는 명령  

사용법  
hostname [option] [hostname]  

* -d : 도메인명만 출력 
* -f : 완전한 호스트명을 출력 
* -a : 호스트명에 대한 Alias 명을 출력  
* -i : 호스트명에 설정된 IP 주소를 출력  
--------------------------------------------------------------------  
* whois : 등록된 도메인에 대한 정보를 확인하는 명령 (dnf install whois로 설치)  

사용법  
whois [option] URL  

* -h : 지정한 DNS 서버에서 관련 정보를 조회해서 출력  
--------------------------------------------------------------------  
* wget : 웹 URL을 이용해서 다운로드할 때 사용하는 명령이다. HTTP, HTTPS, FTP 프로토콜 기반의 URL을 통해 다운로드 할 수 있다.

사용법  
wget [option] URL  

* -r : 하위 디렉터리까지 다운로드 할 때 사용하는 옵션  
* -I : 하위 디렉터리의 depth 레벨을 지정할 때 사용  
--------------------------------------------------------------------  
* curl : 다양한 네트워크 프로토콜을 이용해 URL 기반으로 원격지 서버에 데이터를 보내거나 받는 명령이다.  

사용법  
curl [option] URL  

* -I : 서버의 헤더 정보만을 출력  
* -i : 서버의 헤더 정보 및 응답 정보도 출력하낟. 
* -f : 서버의 HTTP 요청 후 오류 발생시 자동으로 실패하고, 별도의 출력은 발생하지 않는다.  
* -s : 진행 상황이나 에러 메시지를 출력하지 않는다  
* -S : -s의 추가 옵션으로 사용하며 진행 상황은 출력하지 않지만 에러 메시지는 출력한다  
* -L : 서버로부터 HTTP 300번대 응답을 받는 경우에 리다이렉션 URL을 따라간다. 
* -o : 서버로부터 받아온 내용을 표준 출력 대신에 지정한 파일로 저장한다. 
* -O : 서버에 저장된 파일명과 동일한 이름으로 파일을 저장한다 
* -X : HTTP 서버에 요청할 때 사용한 메소드를 지정한다. GET, POST, PUT등을 지정해서 사용할 수 있으며 curl 명령 사용 시 아무 옵션 없이 지정하면 GET이 사용된다.  
* -v : 서버로부터 가져오는 정보를 더욱 자세하게 출력한다.  

* 네트워크 관련 파일 및 디렉터리  
1. /etc/sysconfig/network-scripts 디렉터리  
네트워크 인터페이스 환경 설정과 관련된 파일 및 스크립트가 저장되는 곳이다.  

2. /etc/resolv.conf  
시스템에서 사용하는 네임서버 (DNS 서버)를 설정하는 파일이다.  

3. /etc/hosts  
IP 주소와 호스트명을 매핑시켜 데이터베이스 처럼 사용하는 파일이다. 별칭 설정 가능
